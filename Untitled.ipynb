```python
def clean_corpus(text):
    text = text.replace("\t", "")
    text = text.replace("\n", "")
    text = text.replace("\r", "")
    text = text.replace("\u2028", "")
    text = text.replace("\u2029", "")
    text = text.replace("  ", " ")
    text = text.replace(". ", ".")
    text = text.replace("? ", "?")
    text = text.replace("! ", "!")
    text = text.replace('”', '')
    text = text.replace('"', '')
    text = text.replace('.)', ').')
    text = text.replace('?)', ')?')
    text = text.replace('!)', ')!')
    text = text.replace("et.", "et")
    text = text.replace("al.", "al")

    
    split_text = re.split(r"([.?!])", text)
    punct = ".?!"
    sentences_w_punctuation = list()
    for sentenceid, sentence in enumerate(split_text):
        if "www" in sentence: #makes sure URLs are not included in the corpera
            continue
        if sentenceid == len(split_text) - 1:
            break
        if (sentence + split_text[sentenceid+1]) in sentences_w_punctuation:
            continue
        if (split_text[sentenceid+1] in punct and sentence not in punct):
            sentences_w_punctuation.append(sentence + split_text[sentenceid+1])
    output = ""
    for sentence in list(sentences_w_punctuation):
        if len(sentence) <= 16:
            continue
        output = output + sentence + "\n"

    output = output[:-1]
    
    while "\n\n" in output:
        output = output.replace("\n\n", "\n")

    while "  " in output:
        output = output.replace("  ", " ")

    output = output.replace("\n ", "\n")
    return output

with open("corpus.txt", "w") as outfile:
    outfile.write(clean_corpus(text))
    outfile.close
```

```python
text = """
Hello, my name is Jacob Rosen, and I just want to say that it is very crazy, that it is the time to culminate out of 5th grade. It seems like on the first day of kindergarten, I started with an empty tool belt. Every lesson, grade, thing I learned, and test was like getting a new tool in my belt. I have so many tools in my belt, I can build a bridge. That bridge will take me to middle school, and with all of these tools, I’m very prepared. Willow has taught me a lot, and I have been through kindergarten, 5th grade, and all of the grades in between. The bridge I am going to build is very large, but Willow got me tools to go on the bridge.
	It all started when I walked into the classroom in kindergarten. So much has changed from that day, academic and not. I barely knew anything, I didn’t have braces or glasses, and I wasn’t that mature. I was very lucky, I came to Willow the last year there was the kinder campout, the last year of reciting the “Finding Dracula” poem too. But we are talking about kindergarten, not how lucky I was. The kindergarten campout taught me a lot, like how to be brave in circumstances that you have never experienced. With that in my toolkit, I will use that to actually be brave enough to build that bridge, and cross the gap to middle school.
To start a new experience, when I went to 1st grade, I got to go to the big yard, with the 2nd graders and giant 3rd graders, which taught me about kids that were older than I. What gave me a better tool was Turkey-Tea. Turkey-Tea is when the 1st graders go up on stage near the time of Thanksgiving, and sing some songs. That was one of the first times I went on stage. When you are that age, it takes a lot of guts to do that. I guess that got me a tool of perseverance, which will let me keep building the bridge, even when it is embarrassing, or not going well.
Incredibly, 2nd grade taught me a big lesson. I was in Mrs. Fersht’s class, and we were reading a book called Strega Nona. Now I don't remember much about the book, but what I do remember is there was a character named “Big Anthony”. One thing I remember about Big Anthony is he went to this magic pot, said a spell to make pasta, said the spell to stop it, but what he did not know is that he had to give the pot 3 kisses to stop. So eventually, Strega Nona came back and stopped the pasta from coming out. But Big Anthony had to eat the somewhat 5 tons of spaghetti, and you could guess what happened. So the tool is to follow directions, and don't be a Big Anthony, I repeat Don't Be A Big Anthony!                                        
Next up, there is 3rd grade. 3rd grade has a big event, Ellis Island. Ellis Island is when you go to the M.P.R and you dress up and pretend you are a different person. The person that you are is an immigrant from another country. You have to get through security, not be sick, and you can buy stuff. You can also write letters to your family back in your old country. That gave me an important tool, to be thankful of where I live, and that I am lucky to be born into the family that I have, and not to live in another person's shoes.
Now we will talk about 4th grade. Probably the biggest thing you do in 4th grade is the 4th Grade Fiesta. In the 4th Grade Fiesta, you sing some songs, dance, and have Mexican food(Yummy Yummy)! Probably most 4th graders agree that they did not like the dancing. I in fact did not too. The tool I earned from that is that things that you don’t like, endure through them and you will have rewards(Ex. Mexican food).
The end of the line is 5th grade. Probably most kids would talk about Outdoor Ed, but I don't get what tool you would get from that. Surviving in the wild? Learn about the outdoors? No. I will talk about the State Report. In this report, you have to learn about a state, and write down interesting facts about the state on a trifold poster, and memorize a speech you made. It is hard work to glue everything on your trifold, get the facts, make it look good, and make and memorize a 3 minute speech. But if you work a lot on it, it will look spectacular. That is a good tool. Get done with everything first, and then have time to do what you want.
Thank you, siblings and parents and aunts and uncles and grandparents or whatever relationship to your culminating 5th graders, to come out, change your schedules, work around your everyday life, and thank you teachers for helping us get tools to build a bridge to middle school, and I hope you enjoy the rest of your day. 5th graders, yay! We’re culminating!  
The bear who lived in the forest heard a loud strange call. Who could it be he thought to himself. He realized it was the Jade Emperor. 

“To all animals in the forest. The year has shifted in an unknown way, and the Zodiac is ruined. We will have to do it again, or else there will be no time. We will have a new animal, the Bear.” The bear was celebrating that he was chosen. He was the special one.

The next day, the bear got to the start line. The ice on the ground was shining like the stars. You could feel the cold breeze. You could see the bear zoom of at the start. The race had started. Just then, the horse caught up with the bear, and the bear attacked the horse.                          

“Nobody passes me,” the bear howled to the fallen horse. The bear was as vicious as a black widow.

“Nay!” yelled the horse.
Back a bit, the ox saw the fallen horse, and helped him up. He asked who did it, and he said the bear. The ox sprinted up to the newly accounted zodiac, and said, “Not cool. I am stopping you.”
The bear rammed into him, and that’s when everything changed. The ox bit him.

“OWW! Why did you bite me? If that’s what you want to do, here is a taste of your medicine,”

The bear bit into the ox’s leg, and blood came dripping out like water from a river, and the bear's teeth, like razors, pierced the poor ox. As the raven once quoted, the bear said, 

“Nevermore. You will now not pass me.” 

As the bear gracefully swam through the tough river, and sprinted to the finish line, a flash busted in front of him.  Blending in with his brown fur, the rat jumped off of the bears back, and beat him with the trick with the ox that was now bleeding, thousands of years ago. 

With all of the viciousness of the bear, and the insanity of the race, the bear got tricked at the end,.
He was announced as 2nd place.
To fit in, wear the clothes that the other girls are wearing. Hang out with girls, and go over and play hopscotch. You should not hang out with the boys and run.
Native Americans use natural resources in their everyday lives. 
My chosen tribe was the Apache, who live in the Southwest USA.
Today, the Apache would be in Arizona and New Mexico, which is very warm.
In this report, you will learn about the Apache’s, traditions, food, tools, clothing, shelter, and 3 facts.
Natural resources are key for Apache food. Food is important for survival and they get it provided by hunting and gathering. One way the Apache get food is hunting. Buffalo is something they hunt. People hunt other things though. Also, Apache gather acorns and berries, which are a staple. They can cook in their wikiups and in their tepees.The Apache made great use of their resources food.  
The Apache built their homes with natural materials. 
One type of home was a wikiup. A wikiup was also called a wigwam. The wikiup is shaped like a dome. Another type of home was a tepee. A tepee was a temporary building. Also, a tepee was lightweight. The Apache were nomads, which means no home, and nomads usually were in search of food. That is how they used natural resources to make homes, were nomads, and what their homes were like.
The Apache used natural resources to make clothing and make artifacts.
Apache clothes were sometimes leather. The shoes Apache wore are called moccasins. Tools were helpful for Apache life. One tool they used was a travois, a tool used by nomads. They also used horses, which were important because they helped them get around. The weapons they used were arrowheads and a bow and arrow.
These things were important to everyday life, hunting, and transportation.
My tribe had some interesting customs, traditions, and artifacts.
One tradition they had is to have roasted agave. Agave comes from cacti. Agave was a food they enjoyed very much.  Also, the Apache women stayed at home. They made artifacts. One of these artifacts are sewn baskets.  The men were warriors. Lastly, the extended family stayed together. These facts make the Apache fun to study.         
I am about to tell you 3 interesting facts about the Apache. When an Apache woman marry a man, the man joins the woman’s family. Also, the Apache word for corn is maize. Finally, a famous Apache was Geronimo. The Apache had unique customs and facts about them.
All Native American tribes use the natural resources that they have in their everyday lives. The Apache natural resources were used in a lot of things, the Apache have an interesting culture, and have interesting lives. There are  a lot of things to learn about the Apache.  They are a special tribe with great characteristics.
Cling! Cling! Cling! You hear the bell ringing.  You just got out of school. It is a cold winter day, and you hear the pitter patter of the rain outside. When you get home and you want to play your new FIFA game, and have hot chocolate. You get on the couch, and right before the Xbox turns on, you realize you have two hours of homework.  You just were learning a subject for an hour in school, and you will remember it. How would you forget it? Homework takes free time that could be family time out of people's lives, wastes free time, and reduces sleep hours, because kids are up all night doing homework. Lastly, homework can raise stress levels on kids. Homework should be banned. 

It seems illogical that someone was in school and they learned something for an hour, and that they expect that they forgot everything, and they give them homework to remember everything. According to a website, apa.org, “It barely affects them. Their scores barely improve with or without homework, so it shows that homework doesn't help” (APA). What this means is that someone is wasting time on something that might get them a 96% on a test, and doing no homework would get them a 95.5%. It's not that it’s 5 minutes of homework, it is a great amount of homework. A PBS writer stated, “When asked how many minutes of nightly homework students receive, the average was 81 minutes. Yes, 1 hour and 21 minutes. Average” (PBS).John Buell from the Huffington Post explained that some research has actually proven that homework makes your grades worse than without homework (Buell).  One hour and Twenty one minutes is a massive amount of homework. Somebody could be hanging out with your family, but it takes massively too long. If you had the same amount of homework for all 180 days of school, it would be 180 x 81 or 14580 min or 243 hours. 243 hours could mean a tremendous amount of time with one's family. In 2017, everybody is so busy that our only free time is taken up by homework.  It is a massive amount of work to not make major a change. An hour might get a person one more question correct. 

Homework wastes our time, not just personal, but family time too. Scholastic wrote about freetime and homework. Something they illustrated was, “Opponents of excessive homework argue that it forces parents to sacrifice already scarce family time for the sake of completing assignments, not because they are necessary for students to grasp concepts but because the schools have decided in advance that children must do something for each subject every night whether it's necessary or not” (Scholastic). What that means is that free time is snared and grasped by homework. The time is scarce, and the scarce time is being spent on homework. A New York Times writer, Kristen Weir, wrote about how homework uses up a lot of time. She also pointed out how there has been more homework, and it negatively affects somebody's grades (Weir). Peter Druker, a Huffington Post writer, wrote about how homework is a waste of time, and a student could could be doing better things with their time than doing homework. Homework is something that wastes your time, and sometimes negatively affects your grades, and usually does not affect your grades at all.   		

	Most people know how stressful homework is. Studies have shown that having homework is more stressing than not having homework. John Buell, a writer from a source that gets checked a lot, Alternet, wrote that if homework was a prescription drug, FDA would have banned it a long time ago. It is so unhealthy for you that if it was prescribed, you would get unprescribed, because is is bad for you(John Buell, Alternet). The Huffington Post remarked, “If bombarded with countless lessons at school and at home, students may feel stress and anxiety should they fail  to complete the assignment on time”. If someone get a lot of homework, they are stressed. It is because teachers give too much homework.

	Some may say that homework helps you with school, teaches you time management, and helps you be independent(e-skool), but many studies have shown that homework negatively affects you. Some think that “Homework is great,” but the evidence shows that that is not true. There are few positives and a tremendous amount of negatives with homework. It takes off free time. An unknown author called Memphis Parent stated, “ Students complete their assignments or projects on time”. Well they might learn time management, but they don't have enough time. The people who are saying that homework is amazing are companies like Common Core and the company that makes CAASPP, SBAC, Big Ideas Math, and GO Math. It is like saying that fast food is incredible for you if you are the C.E.O(Chief Executive Officer) of McDonalds®. You just want your product to make money. Evidence backs up that homework is terrible.
	
	Overall, homework is a useless. It wastes free time, it is not necessary, and it stresses kids a large amount. That is why everybody should team up and end homework. It is just a torture device, something that is unnecessary and makes people feel uncomfortable. Without homework, the world would be a better place. That is why we should not have homework.
Greetings,
Hello, nice to meet you. My name is Walle.
I am from Timbuktu, and I am 49 years old
If you don’t know what Timbuktu is, Timbuktu is the great city that is the capital of Mansa Musa’s Mali Kingdom. As my job, I am a brewer, I make beer, I create something for people to drink. It is very enjoyable. I taught myself how to be a brewer after I retired from being a ferba because it sounded like more fun, and I am glad I decided to make the switch.

I live in the brewery, and as a home, it is pretty big. When I am not doing my job, I love to cook. The food I eat consists mostly of salted meat and vegetables. I also pray to Allah, as I am Islamic. Unlike other civilizations, men and women have basically the same rights.

Our Mansa, or ruler has complete power. His name is Mansa Musa, and he is the most wealthy ruler currently. The Hajj is an important event that affected Mali, as Mansa Musa goes on it. I went on Mansa Musa’s Hajj. I met Mansa Musa because I once was a government official, and went on the hajj for that reason, but about 15 years ago, I decided to become a brewer. He treated me well besides the time when I overslept, and made a delay. He got very mad at me, and when Mansa Musa gets mad, he gets very mad.

Thanks for reading this letter. I hope you learned something about my culture. Hope to see you at the Faire.

Sincerely,  Walle.

        	In the novel Counting by 7s, the character who changed the most is Dell Duke, the counselor that does not do what he is supposed to do.
At the beginning of the book, Dell didn’t run because he was not in shape, he didn’t have a desire to run, he was a slob, and he didn’t appreciate the world like he did at the end of the book.
When the book started, the unathletic Dell didn’t run, and at the end, Dell ran, and got in shape. Dell wasn't in shape when he first met Willow, and got inspired later. Dell didn’t need to run at the beginning, because he wasn’t changing,  and didn’t have a desire to. Dell also didn’t know how to pace yourself when running. He probably spirited for 5 seconds, and then got super tired. When he ran for fourteen minutes, Dell looked like he was about to pass out, and probably didn’t even run a mile..
Next, Dell was a slob, lazy, and only ate microwavable meatloaf that was for a family when the book started.
He didn’t have real furniture, because he could just hose off the plastic garden chairs.
He didn’t see anyone, so he didn’t have to make it formal. That is what Dell thought.
When Pattie Nguyen opened up the closet, they saw the Leaning Tower of Pisa made out of underwear.
I don’t mean that literally, but he had a LOT of unwashed boxers. 
He only ate meatloaf, because he didn’t have to eat anything else.
Nobody saw him, so he could eat whatever he wanted.
	Lastly, he didn’t appreciate the world.
He just laid around, and didn’t care about other people.
When he met Willow, and her parents died, he thought of the world differently.
He left Cheddar at the beginning of the book.
He saw Cheddar at the end of the book beat up, and took him to his house.
	So, Dell is now a much different man. He is not a slob, because of the Nguyens moving in, and is more healthy, because he doesn't eat the meatloaf.
He also looks at the world differently. He would still be the slob he was if not for Willow, but he changed more.
I wore  a fancy outfit. I wore that because my parents told me to. What I wanted to wear was a Neff shirt, and athletic pants. Why did I want to wear that. I wanted to get a good impression, and I liked the shirt. I don’t know why I had to wear the shirt. My parents said that “I should ‘look good’ on the first day of middle school.” But really, what I wanted to wear would of made me “Look good.”

	Willow did not cheat because of a few reasons. She probably didn’t know that she was going to take a test, and didn’t know of any sources. If you want to see if she is cheating, or is just a human calculator, give her a random test. If she got a 100 on the 6th grade test, make another one, with the same style problems, but different problems. For example, if the problem is 2x/3z√2
      3x/2z*4 = 2 + 3x / (2z * 2)^2,
change the values. Then you will see the difference. If she Passes (100% - 70%), she didn’t cheat. If she was below 70%, she cheated. 
Describe a topic, idea, or concept you find so engaging that it makes you lose all track of time. Why does it captivate you? What or who do you turn to when you want to learn more?
	August 21, 2019, 11:15 a.m.: I stepped into my Spanish 1 class for the first time. Like my friends, I was not particularly enthusiastic about this new class. What I would later come to realize is that I really enjoyed not only Spanish, but also learning about how different languages are structured. In the end, I participated more often in Spanish class than my peers. I started reading additional material outside of class and watching television shows in Spanish to expand my vocabulary. 
Although the Covid-19 pandemic made learning Spanish in person more difficult, I think I have still made great strides during the pandemic, and my passion for learning new languages  continues. I plan on continuing to study Spanish and other languages in university and beyond. I have even joined forums dedicated to language learning and have had very interesting conversations with those who speak primary languages other than English. Studying languages inspires me, and I hope to become a polyglot (somebody who speaks many languages fluently).
Localization, or the process of taking content produced in one language, and adapting text, images, references, and audio to better suit it to a different audience would be a possible avenue that I would like to pursue. Although I do not consider myself proficient enough in another language to start localizing content, I am deeply interested in that, and have read articles and watched interesting videos about the localization process. What I find most interesting in localization is converting or rewriting puns and jokes for another language; an example being “antelopes eating cantaloupes” is a pun that would only work in English.
I have spent hundreds of hours reading about linguistics and studying other languages’ grammar, through e-books and articles. Just a few years ago, I assumed that different languages were just like English but with different words; the more I dove down the linguistic rabbit hole, the more I realized that was simply not the case. I decided to start learning Finnish because the grammar is notoriously complex, and it has 15 grammatical cases. While I have not ‘Finnished’ mastering this language as of yet, I am proud of how much I have learned on my own and continue to work on my goal of C2 Finnish fluency.
Although some consider the two year language requirement in high school mundane, I enjoyed language classes and am thrilled to be in Spanish 4 Honors this year. Linguistics and language learning have helped me connect with people around me, and has sparked a love of learning that goes well beyond my high school Spanish classrooms.
Change is inevitable and a slow process that can take decades to fully transpire; however, under extreme circumstances change can almost be immediate. The adventure novel 20,000 Leagues Under the Sea, by French author Jules Verne introduces Pierre Aronnax and his limited crew as they are taken hostage.  Pierre is known as a famous explorer and his reputation becomes a point of interest to the elusive Captain Nemo and his submarine of horrors, the Nautilus, as it needs his expertise. Unfortunately for Pierre Arronax,  he, along with his servant and a Canadian harpooner, get cast overboard from a voyage they were traveling on, and find themselves being held hostage on this mysterious submarine. Although initially intrigued, his captivation shifts to horror as he learns that this ship will be where he lives for the rest of his life, as the ship’s captain, Captain Nemo, forbids the trio to leave the ship, as to maintain the secrecy of the ship. Arronax initially starts out as an adventurer, eager to explore the world, and ends the story wanting the opposite. The various forms of trauma which Arronax endures is the agent of change which turns his enthusiasm for adventure into hatred towards the Nautilus and his eventual escape.
		Arronax enthusiastically starts this story with a bright future ahead of him. He is invited on a ship known as the Abraham Lincoln, on an expedition to hunt down a mysterious sea monster. He, along with his servant, Conseil, befriends Canadian harpooner Ned Land, who will use his expertise to help bring this mystifying creature to its end. Everybody on board believed that this monster truly was a monster, as even the captain stated that the creature “is evidently a gigantic narwhal” (28). This plan, however, goes awry when Land’s harpoon cannot penetrate through the creature's steel-like skin, and after the ship gets attacked by the monster, the trio has to jump overboard to save their lives. This would have been the end of their lives if not for the ‘monster,’ as what they landed on was actually a submarine. At this point, Arronax still is enthusiastic about adventuring, as when he comes face to face with Captain Nemo, he eagerly tries to explain the trio’s situation to Nemo in multiple languages in hopes that he would allow them on board. He even tried to “narrate [their] adventures in Latin” in hopes of Nemo’s understanding (41).  When Nemo eventually responds, the three are shocked to learn that they would travel with the crew, but would never be let off the submarine. Although this is initially off-putting to Arronax, his inner explorer ends up engrossed in all of the intricacies on the ship, such as the organ, the advanced equipment, and the library which contains books in many of the world’s languages. As he and his companions are still explorers at heart, this new environment is extremely exciting for the three, and at this point in the novel, they still have high hopes for their futures. 
	Repeated exposure to traumatizing events will cause even the strongest willed people’s perspectives to shift, and Aronnax is no exception.Captain Nemo decides to adventure to where no man had ever gone before, the South Pole. When he reaches the South Pole, he plants his flag, and turns back, only then to come face-to-face with a massive iceberg. They try to venture through the iceberg, but the iceberg shifts and they become trapped. As such, they have limited food, water, and air, and need to use it all to attempt an escape, and they narrowly escape certain death. The stress and lack of fresh air causes Aronnax to be in immense pain, as he states that yawning “dislocate[s] [his] jaws” (371). The immense pain and stress caused by almost encountering death itself fabricates a sense of wanting to escape the Nautilus, not only for Arronax, but also for Conseil, and Ned Land. This is not an isolated incident though, there are many similar situations that have made Aronnax want to never embark on any further adventures with Nemo, and return home immediately.  Stress builds up, and there are still events to come that are far worse than this; however, the incident with the iceberg still has caused Aronnax’s perspective on travel to shift. Even after basking in the glory of fresh air, he still is shaken from this occurrence, and trauma takes years to recover from, years that Aronnax soon will not be spending on the Nautilus.  If he did, however, the Nautilus is not a safe space for Aronnax, and would likely cause his trauma to increase. Another example of an extremely traumatizing situation that Aronnax endures is the fight with the squid. The nautilus encounters a large squid that would be impossible to kill simply by ramming the Nautilus into its flesh, so everybody on the ship had to go to the deck in order to fight it. Their weapons seemed to do nothing against the squid but anger it further. The squid decides to pick up one of the crewmates, and drags him off deck to his eventual death. His blood curdling screams ring through the air, further scarring Aronnax mentally.  Occurrences like escaping from the South Pole and the giant squid are creating a wind of despair within Aronnax to blow, which will eventually blow out the flame that is his ambition to explore.
		After experiencing trauma, change in one's opinions will transpire almost immediately. After each brush with death, Aronnax becomes further displeased with being on the Nautilus and becomes more eager to escape the ship. When he encounters one final near-death experience, it finally motivates him to escape the Nautilus for good, ending his adventures. This ultimate event is a maelstrom, which is an extremely powerful whirlpool. Aronnax, his companions, the ship’s crew, and Nemo all become extremely frightened when they realize the dire situation they are in. This extreme stress that the crew is under is too much for Nemo to handle; he goes to his organ, plays haunting chords, and shouts “Almighty God! enough! Enough!” (276). This situation allows Land, Conseil, and Aronnax to make a run for the dinghy, and potentially be free men once again. The stress that Nemo had been feeling was also affecting all of them. Escaping on a small dinghy during a violent whirlpool is a rather senseless thing for them to do; however, their plan ended up being successful, as they landed on an island off the shores of Norway. These men were no longer adventurers, that time had passed. They risked their lives to get away from adventure, as the immense stress and trauma had caused all of them to change forever.
		Impactful events have caused, cause, and will continue to cause change in people rapidly. Arronax, who was originally an ambitious explorer, put his and his companions' lives at risk to escape from exploration. He upended everything he had, because changes in his life dictated that he no longer feel interested by exploration, and as a matter of fact, he felt repulsed by this very facet of his previous life. This theme can also be reflected in the real world, for the better or the worse (think about how 9/11 both helped increase airline security, but also created widespread xenophobia against Muslims). His journey can also be put into the context of mankind itself; as major changes, like industrialization, evolve us as humans at a rapid pace, not biologically, but informationally. If humans did not have this generational evolution, our modernized world would not make sense to anybody, as society would have never changed to suit anything beyond the neolithic era; however, we are able to undergo this rapid change, whether as an individual, or as a whole society. These major events have shaped our lives in the past, and will continue to do so in the future.
Luke Skywalker accepts personal responsibility for R2D2 after R2D2 ran away to try and find Obi-Wan Kanobe. He realises he will get in a lot of trouble if he is irresponsible and doesn't find R2D2, so he accepts the task of finding him. He also accepts personal responsibility when his remaining family dies, and he has to become a Jedi, and this shows that he accepts personal responsibility, as he knew that he had to do this to help the rebel army, as that was his responsibility. One last time he accepted personal responsibility was when he helped the princess escape. He realised it was his duty and responsibility to save her, as if he didn’t help her escape, she would have been killed by the Empire.
Luke Skywalker discover self motivation by going back to Obi-Wan and wanting to become a Jedi, even after storm troopers destroyed his home, and killed his aunt and uncle. He could have given up, which he might have done saying everything he had was gone and the last of his family (that he knew of) was dead, but instead he went back to Obi-Wan, and ended up doing great things, because he was motivated to avenge his aunt and uncle.
Han Solo is a character who is very self managed. He knows what to do and when. We know that because he knows what to do in almost every situation, and multiple times has gotten out of trouble because of how well managed he is, like with his encounter with Greedo, and when they get sucked into the Death Star.
A time when people employed interdependence was when Luke, Han, and Chewbacka were able to get into the main control room. They worked together as a team, Luke and Han pretending to be stormtroopers, and Chewbacka attacking everybody. Without each other, there is no way that they would have been able to get into the main control room. Another time they employ interdependence is when the same group, Chewy, Luke, and Han, rescue the princess by pretending that Chewy is their prisoner, and that they are taking him to a cell. That cell happens to be the princesses cell, and they help her escape.
Han Solo developed emotional intelligence when he wasn’t even intimidated by Greedo pointing a gun at him and asking for money that he didn’t have. He learned how to control all of his emotions, and isn’t affected by stress at all. He also doesn’t freak out when the stormtroopers attack him. Instead, he gets on to the Millenium Falcon, and tells Chewbacka to fly the ship out of Mos Eisley.
Luke Skywalker believed in himself when Obi-Wan put a helmet over his head, and he had to fend off attacks from a blaster ball with a lightsaber. If he doubted himself at all, he would have gotten hit with the attacks. But he trusted himself and the force, and was able to fend for himself.
After R2D2 and C3PO get out of the escape pod, C3PO doesn’t want to go with R2D2 in the direction R2D2 wants to go, and says he is making the wrong choice. Later, C3PO ends up on a less than optimal path, and he blames R2D2, saying that he was trying to trick him, when really, C3PO was the one who made a mistake and not listen to R2D2.
Ever wonder about black holes? Of course you have! They’re the most popular topic in space for non-astronomists! The fact that they are just a theory makes the common person wonder even more. Well, there’s now an even bigger possibility that black holes don’t just exist… but they also have an opposite.
White holes are theoretical opposites of black holes. Like a black hole, if you go near, space time would be warped. The force is so strong that if you are near the white hole, you would get destroyed.
White holes are formed by the rotation of a black hole. If a white hole connects to a black hole, you have a wormhole.
Despite just being a theory, the idea of white holes is actually over 50 years old. A Russian cosmologist named Igor Novikov came up with the idea in 1964, “as part of a solution to the Einstein field equations known as the maximally extended version of the Schwarzschild metric describing an eternal black hole with no charge and no rotation (White Holes Wikipedia.)”
To teach further about the discovery of white holes, as well as white holes in general, Brit Bendavid and Jacob Rosen, the people behind Space Corp., are creating their first documentary on the subject of white holes. This documentary short is already in pre-production and set for the release date of Friday, April 28, 2017.
Although it is a direct-to-video release, you can still watch it at home. We are not releasing this in IMAX, but instead are selling copies at Space Corp. Our address is 29276 Circle Drive, Malibou Lake, California, 91301. You can also purchase a copy through phone at (278) - 766 - 6691. The cost of our product is $19.99 or a three-day rent of $4.99. Finally, you can contact us through email at asavagenamedbrit@gmail.com, jacob@rosen4.com or spacecrop@gmail.com.
The first words, “Who’s there?” are not simply there as a question; this foreshadows many events that will transpire in the future. “Who’s there?” represents two key questions that are addressed throughout the book, who the ghost is, and who killed King Hamlet. The many encounters with the ghost throughout the first act, and the very end of the first act, are testament to this fact.
What has happened around me since I returned from Wittenberg is quite crazy. I first got asked to see if there was a ghost (to which I had my doubts). Turns out, there was a ghost! I could not believe myself. After this, the state of Denmark has been getting wilder and wilder. King Claudius announced to all that he would be getting married to his late brother’s wife’s sister, which caused much turmoil among us Danes. After that point, poor Hamlet feels much despair, and does not want to continue his life; to him, his life is now futile, however, all of this changes when his temperament suddenly changes, and he proclaims that he is fine right now, but that his actions in the future will seem to be those of someone who has gone mad. This has been the craziest time of my life!
Highland climate is the climate of 'high' 'land'.  So, this climate is found in high mountain areas.
There are no seasons in Highland climate. Any seasonal differences would only be felt at low elevations, near the bottom of a mountain.
The reason mountains need their own climate type is because the "climate" changes as you move up the mountain.  At the base (bottom) of a mountain it might be 80 degrees and sunny, but as you climb the mountain it will get colder and be rainy.  As you keep climbing, it might be snowy and freezing cold.  In fact, the temperature drops about 3 degrees every 1000 feet in elevation as you move up a mountain.  So, the temperatures in Highland depends on the elevation.
The amount of precipitation in Highland climate depends on the elevation.  Sometimes the land around the base of a mountain is dry, but snow may cover the top of a mountain.  This happens because high mountains force warm air to rise, where it cools and creates precipitation.  The level of precipitation in a Highland climate depends on the elevation you want to measure.
The type of vegetation that grows in Highland also depends on the elevation.  At the base (bottom) of the mountain the vegetation will be the same as the surrounding 
climate type.  So, a mountain in the rainforest will have rainforest at the base.  As you move up in elevation it will change to plants that can survive colder weather, eventually there will be no vegetation at very high elevations.  In fact, on high mountains you can see what is called a tree line or timberline.  A tree line is a clear elevation line that divides where trees are able to survive from areas that are too harsh for them to live.  Most mountains have different tree lines depending on temperature, soil, and moisture.  
Again, the types of animals will depend on the elevation and the surrounding area.  Typically, different types of goats and sheep are common in Highland climate.  Mountain Goats are common in the Rocky Mountains--their feet are adapted to walking and jumping on steep cliffs.  Large cats such as Mountain Lions and Snow Leopards are also found here. Snow Leopards, found around the Himalayas, have large feet and big thick tails to give them better balance in mountain terrain. 
Highland climates are cool to cold, found in mountains and high plateaus. Climates change rapidly on mountains, becoming colder the higher the altitude gets. The climate of a highland area is closely related to the climate of the surrounding biome.
Highland climates are cool to cold, varing by altitude. Usually cooler and wetter than nearby lowlands. 
Recycling, in concept, is a greatly beneficial concept, as society can properly reuse disposed goods and turn them into new products, and for a time, that is what society as a collective accomplishments. Nowadays; however, this practice of recycling is no longer widely practiced, as China no longer accepts trash from the United States because of health concerns and fewer economic incentives.
	Historically, China had ships bound from the United States to their harbors coming home empty, which wastes fuel and money. This waste changed when China decided to buy the United States’ trash, to recycle it into new plastics. When this practice first began there were many economic advantages to recycling the United States’ garbage, but as this practice began to evolve China no longer needed this excess plastic, as it was causing more problems than it was fixing.
	The end of the American-Chinese garbage exchange is unfortunate, as not only does it cause more trash to enter the landfills, but it also causes more pollution in the world, and helps fuel oil miners to destroy the crust further. If a domestic recycling solution were to have been implemented instead of relying on a foreign nation, this problem would have been ameliorated; however, as this course of action was not taken, our dependence on China caused what would have been the inevitable future of green plastic to plummet overnight.
	The future is bleak, but if society can conjure up a solution to stop wasting useful, non-renewable, used materials, society can prosper. The United States should continue to implement larger and larger tax cuts to those who recycle in order to incentivize people to make plastic more ‘green’, rather than purchased, filthy, mined plastic.
Change is inevitable, it happens to everything. It cannot be avoided, but what can be is how something changes and why it changes. Jem changes his attitude throughout the book, starting as a rebellious young boy, to a mature young man. He starts out young and immature, thinking that he shouldn’t listen to any rules, and create his own. He didn’t respect his father, Atticus, and those around him, doing whatever pleases his and makes him happy. For example, Dill dares Jem to touch the Radley house. Instead of doing the mature thing, Lee wrote that Jem, “threw open the gate and sped to the side of the house, slapped it with his palm and ran back past us, not waiting to see if his foray was successful.” (Lee, 18). Even when Jem had previously asked people about Boo, and they told him to not disturb him, he still thought that he knew better, and didn’t do the right thing. He could’ve done the correct thing, but instead, he tried to impress Dill, showing who his true character was. His rebellious attitude could’ve ended up going wrong, luckily though, it didn’t. Jem just isn’t the most mature he could be, and it shows. However, he does change, for the better, throughout the rest of the story.

Change does not just happen, as something needs to cause it. Jem isn’t static; for he changes with the environment around him. He starts out not knowing what is right and what is wrong. He changes though, standing up for what is right. He does this when Mrs. Dubose is being toxic about Atticus’s case. Jem shows her that she is a terrible racist when he cuts her hedges. This is because she said that Atticus was, “No better than the [n-word]s and trash he works for” (Lee, 135). Was this a reactive decision? Yes. But it is the motives that drove him that shows that he is changing. He believes in his father, and instead of disobeying him, he stands up for him. This change in attitude is no accident; it is him adapting to changes in his environment, as his father's stress about the upcoming case is reflected in Jem’s attitude. He changes how he acts because he moves with the tide. He gets very affected by his environment.
There are many inventions that were made in Ancient Greece, but there are few that we still use. Some of the ones we use are, democracy, coins, and boats. 
Democracy back in the Greek days was run by the assembly. The assembly voted on laws, and voted on a ruler. Now, in the U.S. government, we vote on either a Democratic or Republican candidate.  There are three branches of the government, the Legislative, Judicial, and Supreme. The president and vice president are in the supreme court. The legislative has the Congress, which has the House of representatives, and the senate. The senate writes the laws. The judiciary has the supreme court. 


Another invention is coins.
In the olden days of Greece, there were three main types of currency: Talent, Mina, and Drachma. We now have pennies, nickels, dimes, quarter, half dollar, and dollar coin, as well as paper dollar bills, 1, 5, 10, 20, 50, 100, 1000, and 10000 dollar bill. There is now inflation, and different countries have different currency.


Last of all, we have boats.
The old boats were not simple, but they were much simpler than they are now. When now boats are run by steam, pistons, or motors, back then they were had a sail, and they were run by oars. Olden boats were usually made of wood, and were hand made, when boats now are made from metal, an extremely strong plastic, or an alloy of many strong materials. Greek boats did not have a large capacity, but now boats(mostly cruise ships) don’t just have a extremely large capacity, they can have things like theatres, pools, restaurants, gift shops, and even suites.
Gravity can be pulling an object even when it is not falling.
The Force of gravity of an object is different when the object is in different places in the solar system
??? What determines the strength of gravity?
Mass: The amount of matter within a object.(takes up space)
Weight: The force of gravity acting on your mass. (scale is squeezed)
He makes the connection between the audience by introducing himself as an African American student. This establishes ethos because it shows the reader that this is somebody who has experienced prejudice firsthand. He uses pathos, logos, and ethos throughout, one example of pathos is giving anecdotes, and another example would be his use of allusions. He changes the pronouns he uses to create a sense of disconnection, something that an African American student faces often. His use of the n-slur makes this part of the piece emphasized, although some formality, unfortunately, is lost.
Manhattan, New York City, New York State. Felix and Antonio, the Amigo Brothers, were walking with each other, and were talking about the fight that they were going to fight against each other. They decided to split up to balance their mind sets, and make the fight fair. They didn’t want to fight like they were friends.

Felix decided to pass the time by training, and he also watched a boxing movie. He tried to not think about Antonio. Antonio passed the time by training, and also went to his Aunt’s house in Bronx. While he made the journey, there were some gang members he avoided to prevent conflict.

It was time for the fight. They saw each other when they dressed and waved to each other. When they went into the ring, the crowd was roaring and cheering with an excited atmosphere. BONG! Round One began. They fought hard, but they were fighting with care. They fought not hard enough, and fought with care.

That changed in round two. BONG! They fought with viciousness, like two starving bears fighting for food. But they fought each other way too hard. The round ended, and they felt as if they had just climbed a mountain with somebody punching them while they climbed. They were bleeding, and when they went over to their coaches, they said to not fight as hard. 

BONG! Final round. The last two rounds tied, so this was sudden death. This was the moment that would change them. They fought out their hearts, and fought with all of their energy(that was remaining). BONG! The round was over and the fight was too. This was the climax, with everything rising up until now.  The announcer was attempting to talk in multiple dialects. “Ladies and gentlemen, Senoras y senores” The announcer was announcing who the winner was. “The winner is…”. And that is how the story ends. That was the end of a development, and the end of the story. The Amigo Brothers had left the ring, and would never know the winner.
Charlie wants to become a true father to his daughter again. He leaves in order to spend time with her and to try to convince Honoria that he can be a sufficient father and should be given back.
He used to have a toxic relationship with his daughter and left her, and his wife, because his wife kissed another man at a party.
He means that at every party that he and Helen got more distant from each other, and that the constant partying was pulling him away from her.
Marion resents Charlie and therefore puts all of her negative feelings and anger on him.
He likes to believe that he is a good person, but what he did when he abandoned Helen obviously has taken a toll on him, and he regrets his previous actions.
She becomes sickened because of what she saw.
Charlie does not want to think that he is a weak, lazy, and bad man, and paints a picture of Helen in his mind that is unrealistic. He does this because he needs to find comfort in an otherwise extremely bad situation.
The difference between a company doing “good” and a company doing “well” is that the company doing good is contributing positively to society and making, or at least trying to make a positive impact while the company doing “well” is doing whatever it must in order to make a large profit and nothing more. Mattel diversifying Barbie dolls is Mattel doing “good” by making sure that anyone can find a Barbie doll that looks like them, and promotes a broader sense of what it means to be fashionable rather than the generic blonde Barbie doll.
This topic is more suitable to the daily rather than the monthly as although it is a feel-good piece, not much change will occur from this. Yes, many people will be able to buy Barbies that more accurately reflect themselves, but this will not make that big of a difference in the grand scheme of things.
The two authors’ points of view are similar in the fact that they both despise how a cookie-cutter model is cast among so many young people in an unhealthy way that promotes unhealthy bodily standards. The value that these toys bring is outweighed by the harm they may cause, and luckily, the diversity helps with this by making Barbies more fit for everyone.
He uses a serious tone, as this subject is no laughing matter. The voice he uses in each passage is one that has undergone or is undergoing a considerable amount of pain. He could have used an omniscient third person point of view, which would give the same plausibility, but also could put one in the shoes of the waterboarders too, rather than just the waterboarded.
Hitchen means that the interrogators, although questioning this man and torturing him, still had sympathy for what he was undergoing. The effect of putting this towards the end of the essay changes one's view of the interrogators, from being soulless tormentors to being real people who were trying to get an answer, but could see the person on the other side. Although this was just a simulation, people went through this unsimulated, and probably got sympathy from the interrogators.
He looks like he is screaming out in pain, and cannot escape. This angle shows his face and body foremost, although you cannot see his legs, which are probably flailing out of pain.
The difference is that Hachiya uses tone to convey grimace, while Hitchen uses tone to convey pain and suffering. They do both; however, use serious language, and go into vast quantities of details. Hachiyas essay makes me sad, while this essay makes me feel angry.
What he gains from this is that he makes it clear that everywhere on Earth, there is racial discrimination, but what he risks is people not taking his work seriously because of a rather simple analogy. His use of figurative language, in my opinion, makes the text more understandable and vivid and makes the reader more engaged. 
By addressing his son, you realize that this is a problem that is multigenerational, has lasted a long time, and may not end anytime soon, unfortunately. By mentioning those two who got shot, it puts perspective into the average day of an African American, and the reader contrasts their lives with that of an African American and begins to understand, and maybe even feel passionate about, the issues of African Americans.
The issue that Coates gets by assuming that schools are always against him is that there are definitely schools that are supportive of African American students. The educational system is unfair, but it is to everybody, not just one specific group, as it was designed to make students break.
When the author states that “The West isn’t a place that gives up easily”, she means that the west is not a place where people often sit back, and people do stand up for what they believe in.
The narrator describes Homer’s job to show that people who moved to the west had an affinity towards nature.
She describes the way that Homer and her make love in-depth and with a lot of detail. This is to make the reader understand that she and Homer did not have a very close relationship and a strong connection to one another and that she loved others.
He was the man she truly loved. She never cared for Homer, she only cared for his body, but she did love Monte.
Homer was soft and would not hunt animals, and cared for nature. The narrator views this as a weakness.
She realizes that Homer is who she really loved and that she shouldn’t find love in people’s body.
She isn’t happy with the outcome of the events that unfolded.
If I were Dr. Heidegger, I would believe that this water causes more harm than good to society. The men become extremely immature and ended up destroying the vase that contained the water in which they cherished due to its youthful properties.
These characteristics describe characters who although are annoying, will be gone in the next 5 years. These characters are greedy, as they all try to take the woman for themselves, gluttonous, as they all devour the water, and lustful, also pertaining to their actions caused by the woman.
The part where the flower suddenly youthened revealed the science fictonious nature of this work.
The motivation was to create a cure for aging and becoming an old geezer.
The mirror represents the fear that might be caused by people and speculation of what something like this means for the greater society.
The guests first hand learned what greed can cause. I know this because they saw themselves shift into better versions of themselves, only to be reverted shortly thereafter. 
Coronavirus came out of nothingness and changed everything. Suddenly we had to adapt to these new circumstances that were cast upon us. For school, we started by doing “Distance Learning”, and all classes were held virtually. There have been many positive experiences in this new world of learning, such as the ability to more reflect our sleeping schedules to how our bodies seek fit, as we can now wake up 30 seconds before class starts, as opposed to the 30 minutes required prior. There also is much more free time to get homework done during nutrition and lunch that won't have to be done after school is dismissed. Everything isn't perfect though, as many students miss seeing their friends in person, which can no longer be done. That being said, the students can communicate through chatting apps like Discord, so not all hope is lost. Overall, Distance learning is very good, and even though there is the elephant in the room that is the social aspect to school, most of this program has gone swimmingly.  
Sentiment analysis is a powerful tool that can be used to evaluate if a given sample has a positive, negative, or neutral sentiment. Within this paper, a bidirectional RNN trained to analyze sentiment was given sentences  from Wikipedia article revisions over a certain time period corresponding to before and after a major current event which involved the subject in question. Based on the findings, analysis suggests that political issues were more likely to experience a negative shift in sentiment. This sentiment change was often sharp and sudden.
The use of language on the internet has been tied to people’s sentiments since the very early days; however, in today’s world, with people more than ever connected to the internet, this phenomenon has real world ramifications. One of the largest websites, Wikipedia, a crowdsourced online encyclopedia, gets updated constantly. While Wikipedia intends for its articles to be politically neutral, articles can oftentimes exhibit political skew (Greenstein & Zhu, 2012). By evaluating Wikipedia articles over time, and tracking the sentiment for each article section, the correlations between current events and the way language is used can be observed. 

Tracking these articles over time can reveal several details about the use of language on Wikipedia. Wikipedia was specifically chosen as it has a somewhat consistent writing style, and is one of the largest worldwide sources of information. Additionally, it has been posited that articles on Wikipedia undergo high levels of change during newsworthy times, which create articles that are noticeably different from the prior versions (Ferron & Massa, 2011; Keegan et al., 2013). Such findings can include which topics are the most contentious, which types of topics seem to be affected the most by current events, and if shifts in the type of language used happen suddenly or gradually. Researching these shifts in language use reveals information about both Wikipedia as a platform, and the people who use it.

As such, this paper aims to address three main research questions: 
-RQ1: To what extent does sentiment change occur in relation to current events on Wikipedia?
-RQ2: Is the aforementioned sentiment change a universal phenomenon on Wikipedia, or does it only occur on certain issues? 
-RQ3: Is the change in sentiment gradual or sudden?

In order to obtain such data, it is important that a computational model be applied, as manually sifting through (in some cases) over a thousand article revisions per article to search for subtle differences in tone would be an infeasible task. To achieve this, a bidirectional recurrent neural network (BiRNN) was trained on labeled tweets from Twitter, and then was subsequently fed sentences from multiple article revisions of a single article to track the change over time. Although domain shift of this type has the possibility of introducing noise into the model, a qualitative test inputting Wikipedia sentences into the model suggests that the model performs appropriately on sentences from Wikipedia.

To examine this data, computational sentiment analysis was performed. Sentiment analysis, or opinion mining, is the computational process of using natural language processing to determine sentiment within a given text (Pang & Lee, 2008). Rather than utilizing a simple sentiment analysis via cosine similarities (which can be used as a tool to determine similarity and therefore sentiment for individual words), this approach uses a BiRNN to learn more complexity about sentences and account for context.

Learning about language use over time through revisions as is done here is important as it shows how the information which people consume affects the sentiment of online writing. It has been proven that the information that people intake can affect their sentiment on a topic (Mutz et al., 1997). Since the way Wikipedia is written both reflects and affects how people feel, understanding these changes can show how people feel and can help us predict how people will feel in the future.

A corpus of tweets marked as positive, negative, neutral, or irrelevant was selected for this process. This dataset was chosen due to the wide variety of topics discussed on Twitter. Many corpora use online reviews as topics, but since those usually just talk about commercial products (whereas Tweets discuss a larger variety of topics), they would be less appropriate for classifying Wikipedia text. All tweets marked as irrelevant were removed. Each tweet was made lower-case and then it was embedded with spaCy’s pretrained embeddings  (en_core_web_lg) with an embedding dimension of 300. The labels were mapped to 0 for “Positive,” 1 for “Negative,” and 2 for “Neutral.”  About 17% of the data was removed due to being marked as irrelevant, and the remaining data was well distributed, with 22,542 negative samples, 20,832 positive samples, and 18,318 neutral samples. Sequences were padded by the model at batch time, not during preprocessing. No additional preprocessing (such as removing mentions or URLs) was applied.

A bidirectional RNN was trained on these embeddings. An RNN was selected for this task due to its relative speed and effectiveness. Training data was batched, with a batch size of 32 that was padded and packed. The model consisted of two hidden layers, both with a size of 48 in each direction (or 96 per each timestep). From there, this output was passed into two feedforward layers, the first of which had 128 features and the second of which had 64. Dropout was used on the training with a value of 0.3 in the BiRNN, and 0.5 in the feedforward part of the model. Training used cross-entropy loss, and Adam was used as the optimizer with a learning rate of 10^-4. During training, the model was trained on batches of 32 shuffled tweets. The model was trained for 144 epochs (at which the training started to plateau).
The model, after 144 epochs, ended up with a training loss rate of 0.26, and a validation accuracy of 0.84. Overall, the model performed with an F1 score of 0.84. The validation set contained 1000 samples. The validation samples were provided within the dataset used for training as a separate file. Overall, due to the variability of the dataset, and the subjective nature of sentiment analysis, an accuracy/F1 of 0.84 was deemed to be sufficient.
In order to test a wide variety of concepts, both pop culture and political topics were chosen for evaluation. This choice was made, as politics and pop culture are among some of the most talked about aspects of modern society, and these topics were likewise edited frequently on Wikipedia. These will provide data that shows many edits on ever-changing topics, which is optimal for this analysis

Due to the varying amount of edits made to each individual article in comparison to the other articles, no uniform date range around a given topic was chosen, and instead a case-by-case evaluation was made to ensure that there were enough edits made before the event (at least a month), and enough time and edits made after the event (at least half a year and at least several hundred edits).

 The chosen articles were as follows: Charli XCX, Dave Grohl, Russia, Ukraine, Zionism, and the Wikipedia entry about apples as a control. Apple (the fruit, not the company) was chosen as a control due to apples rarely appearing in the news, and it not being a charged topic within current discourse. All of these following topics were in the news and relevant in society one way or another, hence their inclusion.

British musician Charlotte Aitchison’s (Charli XCX) 2024 album Brat was a cultural phenomenon in the English speaking world, making her music and her fame grow. This album was so popular that the summer of 2024 was retroactively named “Brat Summer” by some (Planey, 2024). As such, a period of time spanning approximately half a year before the release of the album to about a year later was chosen for this article.

Foo Fighters singer and former Nirvana drummer Dave Grohl is an important figure in the music world, but to the dismay of his fans, he announced on Instagram in September of 2024 that he had fathered a child outside his marriage (Grohl, 2024). Because this was a contentious time for his image, a period of time spanning about two months before this incident to ten months after the incident was chosen to see if and how this controversy would affect his main Wikipedia article.

Following the Russian invasion of Ukraine on February 24, 2022, both Russia and Ukraine were in the news. These articles can help illustrate the differences in views between these two sides, and how the war has affected them. For the Russian article, the span of dates goes from about a month prior to the invasion to a year after, and the Ukrainian data was collected from about two months before the invasion to a year afterwards.

The topic of Zionism has been controversial following the attacks on Israel on 2023-10-07 and the subsequent war. Since it has been such a debated topic, this article can provide insight about how articles with contentious topics are edited. The date range chosen for this begins three months before the October Seventh attacks and ends slightly over a year afterwards.  Overall, all of these topics provide ample grounds for evaluating the types of topics most associated with sentiment shift on Wikipedia, and provide insight into how these edits are accumulated, and how they last over time.
Graphs were constructed for each as a line graph, with the option of displaying all the sections in the article as a single graph. Sections that did not exist in every revision tested were not used to ensure all sections existed for the whole graph. Due to the sheer number of revisions used, a parameter was set on each graph to average a fixed number of revisions per data point. Additionally, to ensure that the graphs show data over time elapsed instead of revision number, the averaged datapoints were placed according to the date of the last revision in the batch of averaged revisions.

Since (excluding the control) all the articles have relevant dates in which a current event occurred, there is an option to place one or multiple vertical lines on specific dates. This approach, rather than putting a dot on the graph at the location, helps the multi-section graphs show where on the graph the event occurred, and if or how long after any change in sentiment occurred.

The control (apple) found that within an article that was not related to current events, sentiment tended to stay relatively consistent throughout the duration of the data, and stay around 0.5 on the graph, ± approx 0.1. One important thing to note about the graph for apple is that the etymology section tends to spike and dip substantially more than other sections — this is due to the length of the section, which is substantially shorter than other sections. Even very subtle changes within the section can have drastic effects on the graph. Therefore, while using this approach, very short sections are not good candidates for evaluating language change.


Within the article for Russia, most of the sections looked consistent, with two notable exceptions. The first exception was the main contents of the article, at the top. Additionally, the section for government and politics was also affected. Both of the aforementioned sections showed declines in sentiment, although the way in which they declined was different. While the main contents had a more gradual decline that seemed to start from the beginning of the data (prior to the invasion), the government and politics section had a sharp and sudden decline. Additionally, the main contents tended to vary more throughout the trial, suggesting that subsections tend to show stronger sentiment trends than the main section.


In contrast to the article for Russia, the article  sections for Ukraine tended to show no or positive change in sentiment, with the exception of etymology and orthography. Similarly to the article of apple, the short length for this section causes even small changes in language to have large effects on the graph, explaining this change. The rise in the contents section can be attributed to subtle changes in tone between revisions — for example, the model classifies the sentence "The country is a member of the United Nations, the Council of Europe, the OSCE, the GUAM organization, and the Lublin Triangle, and is one of the founding states of the CIS, even though it never became a member of the organization” (Wikipedia Contributors, January 2022) to be negative, due to the sentence ending with a negative implication. In the newer revision, this sentence was changed to say “It is a member of the United Nations, the Council of Europe, the OSCE, the GUAM organization, the Association        

   Trio, and the Lublin Triangle.” (Wikipedia Contributors, March 2022), which the model classifies as neutral (due to there no longer being a negative connotation at the end). Revisions similar to this were made to the article by Wikipedia editors following the start of the war, causing the sentiment within this section to become more positive.

In addition to the war between Ukraine and Russia, another major conflict was the war between Israel and Palestine. Rather than looking at a country in this scenario, a philosophy was looked at instead, namely Zionism. Certain sections, like the contents and the terminology tended to fluctuate, while most other sections remained relatively stable, with the notable exception of history and beliefs. History was steadily declining ever since the start of the war, and beliefs tended to fluctuate, until September 2024 on the graph (which was when Israel launched attacks on Lebanon). Unlike other articles where after a major dip in sentiment, the sentiment starts to steadily rise to a steady level, the sentiment stayed relatively stagnant for the remainder of the graph.

The graph for Dave Grohl shows relatively consistent sentiment over time (with the exception of Musicianship and equipment, which is too short to suggest stable estimates). The dip in all sections about 40 days in was due to the formatting shifting within the article, and was deemed as meaningful due to the short length.
     
The graph for Charli XCX shows a spike in approximately October of 2024, which corresponds to an addition in the section about her heritage, specifically that she has pride in her Indian heritage. The decrease in the early life section corresponds  to an addition in the section about racism she experienced as a child being half-Indian and living among white people. The sections for career and the main section tended to stay relatively stable, with a slight dip around the beginning of November 2024 making it more neutral. This was because the size of the section expanded, which increased the amount of neutral sentences that were compared, lowering the               average that was above neutral. For the same reason as previous etymology sections, the tours section is ignored due to insufficient length causing too much noise. 
Another pattern that is observed within the graphs was that when significant changes occurred, most often the sentiment ended up returning near the level at which it was prior to it changing drastically. The notable exception to this is the main contents of the Ukraine article, which went up, and did not return to the original levels. The two mentioned sections in the Zionism article did not return to normal levels, but this was because the dips happened very late in the collected range of revisions and therefore there was insufficient time for them to return to their mean position.

Overall, the findings indicated that political topics tended to experience the effects of current events in terms of sentiment change the most, whereas pop culture topics seemed to experience less sentiment shift. The time taken for this change in sentiment to occur varied, but the change typically was pronounced.

The use of a BiRNN was effective for creating data, as the model was able to understand context within a sentence, however, a major limitation to this approach involved the dataset: there is a mismatch between what was used to train (Tweets) and what was evaluated (Wikipedia sentences). While at first this may seem alarming, this did not appear to be a limitation or an issue based on qualitative analysis, as the samples tended to average around neutral sentiment (which a quick read through several Wikipedia articles matches), suggesting that the use of training data that is less formal than Wikipedia has a negligible effect on the quality of sentiment analysis performed by the model. Future research can improve on the methodology used by this paper by ensuring that the model is trained on data from the source that is being evaluated.

These findings indicate the type of edits made to Wikipedia articles, specifically in regards to current events. As mentioned previously, the issues that tended to receive the most attention within the sampled articles in terms of edits within were issues that were mainly ongoing political topics. As such, it can be suggested that topics related to current day political events tend to see strong emotional shifts during current events on Wikipedia.

According to Wikipedia themselves, Wikipedia has an “internal policy which states that articles must be written from a neutral point of view, which has the goal of representing fairly, proportionately, and without bias, the significant points of view that have been verifiably published by reliable sources on a topic” (Wikipedia contributors, October 2025). This is in line with what was noticed by one of the Wikipedia cofounders, who suggested that Wikipedia is no longer a neutral source and may now skew left politically (Sabur, 2021). While this subjective and anecdotal claim cannot be proven based on the testing that occurred (in regards to the political leanings of Wikipedia), more emotional writing has been correlated to stronger influence on ones view of an issue (Baum and Rahman, 2020), and therefore the effect that emotional writing has on Wikipedia readers is a genuine question. This paper does not aim to address political leanings and therefore this paper cannot prove or disprove the co-founder’s statement; however, it can provide insight as to why he might believe this.

It has been proven that when discussing political topics, people tend to use more emotional speech on issues relevant to their beliefs (Gennaro and Ash, 2021).  This is consistent with the data,  as edits that change the overall sentiment of political topics tend to cluster soon after a major event. An example of this can be seen in the Zionism article: while the section for the main contents seems to vary, it overall has fallen since the start of the war, and sentences like “Zionists wanted to create a Jewish state in Palestine with as much land, as many Jews, and as few Palestinian Arabs as possible” (Wikipedia Contributors, October 2025), which was not present in the article prior to the war. Prior to the war, this section appeared to contain analytical and descriptive information, rather than emotionally based claims about what Zionists historically wanted.

The role of moderation is very important to this factor, whereas while the sentiment could change rather quickly, it usually returned to the ordinary sentiment level of the section prior. The fact that the sole exception to this rule was a positive change in sentiment raises the possibility that Wikipedia editors may be more likely to moderate lexical shift in the negative direction than the positive direction. Unfortunately, the only other similar datapoint (the personal life section in the Charli XCX article) did follow the aforementioned trend, that being said, it did level off at a similar yet higher mean sentiment, therefore this trend involving the moderation is inconclusive, and more research should be done involving this subject.

Due to the fact that current events are shown to impact the sentiment of Wikipedia articles, some may question the accuracy of Wikipedia as a source of information. As previously mentioned, when people intake information, it does affect their perception of issues (Mutz et. al., 1997), and Wikipedia has shown sentiment shift in areas relevant to online political discourse, which can suggest the type of edits made on Wikipedia. That being said though, every editor of any news source will perceive the issues they are writing about and will in turn take a stance, which can be reflected in their writing, meaning that all news sources are susceptible to this phenomenon. Despite these findings, Wikipedia still remains a reliable source of data (Mannix, 2022), despite the emotional writing style of political topics. This is because sentiment and accuracy are different measures, whereas sentiment reflects emotion, accuracy reflects truthfulness. This is an important and nuanced distinction, as even though it is safe to say the information provided on Wikipedia is likely true, that does not mean that people will be swayed one way or the other by the way that said information is portrayed on Wikipedia. 

Overall, the findings from the data show several trends about Wikipedia (specifically these set of articles). First, section that received sentiment shift tended to be topics that are political. While these types of edits were made, typically the large changes in sentiment tended to return back to a steady level after some time, suggesting that the moderation on Wikipedia is effective within keeping overall sentiment consistent. Additionally, Wikipedia has been proven to be an accurate source of information, although accurate does not always equate to unemotional, and the way political topics are discussed tends to be more emotional, despite the fact that Wikipedia has an internal policy of neutrality (Wikipedia contributors, 2025). Despite this, Wikipedia is usually a good source for obtaining information, although caution should be taken when approaching political topics on Wikipedia.

Summertime is almost here! The weather is great (finally!), the birds are chirping, and yet, you are still eating your disgusting plain instant ramen. How sad. However, I offer to you a solution — here are two delicious and medium effort recipes that you can make for yourself and share with your friends (this assumes you have them). These recipes are healthy, delicious, and taste MUCH better than whatever chemicals you are ingesting with your ramen from its dissolving styrofoam container.

Recipe 1: Poke Bowls (serves 3)
Did you know that Norwegian Farm raised salmon is safe to eat raw? According to FDA Food Code References: 3-402.11, “Aquaculture Fish, such as Salmon, that are served raw or undercooked are exempt from [flash freezing requirements],” given that they are served feed that does not contain live parasites, which Norwegian salmon feed is free of due to heat treated feed. Additionally, farm raised salmon (at least bought on the west coast) is almost always flash frozen as well due to traveling across the country. In short, this means that Norwegian farm raised salmon is safe to eat raw.

Recipe 2: Hearty Pasta Salad (Serves 6)

I know many a college student who eats pasta plain, no creativity, no sauce (or even worse, canned sauce), and no fun. A great way to make your past more interesting and healthier is to turn it into a delicious, hearty salad.

Summertime is wonderful, and with two brand new and delicious recipes for you, it will be even more wonderful! Hopefully, those of you who are hesitant to cooking due to the amount of effort required now can see that healthy and delicious food can be made for very low effort. Enjoy!

My artistic point of view about this production is characterized by struggle through contrast — struggle in adjusting to life in the US, struggling to retain ones culture, and struggle against oppression. The acting style for this production should generally be loud and full of emotion but should at times also be reserved — themes such as violence and abuse should have one side’s being portrayed as strong and full of power (the abuser), and one side small and lacking power (the abused), symbolizing the dissonance between the two sides. A good example about this would be when the two men grab the young girl. When they grab the girl, the lighting should cast shadows that make them look bigger than they are, and the girl should be screaming and struggling, but not with full effort (to symbolize the lack of power she has in this situation). This mindset of expressing struggle through dichotomy can be applied to other aspects of the production. For example, the background can be representative of Medea’s struggle to assimilate to American culture, having the American elements of the Victorian household look neglected while the Mexican elements look colorful and full of life, representing her view about life in America. This mentality can also be applied to the costume design, for example, when Medea kills Acan, his American outfit can look faded and dull, the only color being the blood. This contrast once again represents the struggle for Medea to adapt, but also how Medea is losing Acan. This contrast could also be shown through the music — within the same scene, there could be a medley of strong uptempo Mexican music when Medea is talking, which switches to slow,  American music when Acan is talking. Overall, my artistic point of view is that this production needs to emphasize struggle through contrast in the acting, set design, clothing, and music.

This production should be for a wide demographic but is especially important to happen now. In such a tumultuous time, especially with the hostility to immigrants, it is important to watch a production like this to remind the audience both about the importance of keeping ones culture, the importance of assimilation, and the importance of accepting both of these. This conflict is what leads to the play, and in a time where people are starting to be hostile to those who are different I believe this play is a good reminder of acceptance and empathy.

I want the audience to gain an understanding of the importance of kindness and understanding — the tragedy of this play (the killing of Acan), and many other misfortunes that the characters endure due to lack of kindness and understanding. Many people in today’s world do not accept others and treat each other without kindness, and I want the production to show what outcomes such behaviors cause.

I hope that the audience walks away feeling bad for the characters, especially those who lost their lives, like Armida and Acan. I believe that this is important, as the aforementioned importances of kindness and understanding play directly into the theme of struggle, and when characters don’t receive kindness and understanding. A character such as Armida, despite being in the grey area within this story due to her influence, was still an active victim, just like Acan, whose death is equally tragic, showing that approaching struggle (in this case between Mexican and American culture) without kindness and understanding can have major consequences.

My central image is the border between the United States and Mexico. I believe that this represents the transition and struggle between Mexican and American culture in Medea’s life, and also the trauma from getting across the border. I chose this specific image, as I am fascinated in man made structures and their effects on people. This would also be seen in the costumes — those in the story who had to pass through the dusty border, especially those who were negatively affected by this experience like Medea, would wear dusty clothes. I would motivate the lighting to be bright — this image shows the day in full light, but I would also include shadows (since the wall creates a large shadow), placing characters who are facing negativity in the dark. This ties back to the artistic point of view, as it really highlights the characters undergoing struggle, and accentuating their differences. I would use this scenery directly when the characters are crossing the border, as it shows the true size of the wall, symbolizing the contrast between Mexican and American lives, and the struggle to incorporate both together for Medea. I would also incorporate this into sound design, because even though the house in Boyle heights is in the city (a loud place), the border represents the inner struggles of the story, which is a quiet place, meaning when conflict arises in the production, the background noise should be reduced to a minimum. In these quiet moments, sounds that might be heard at the border (such as screeching of birds, or the rustling of wind) can be heard, timing. This represents both the dirtiness of the border, and the emotional aspect of it. To the characters in the story These all help me convey my artistic point of view by accentuating differences between characters, showing how the differences between them cause struggle, providing contrast between the sides of the border, both figuratively and literally. 

I plan on using the elements of design to construct sets which portray environments in a fashion that accentuate certain characteristics that show contrast. Some examples would be using lighting to make the difference between bright spots and dark spots much greater than typical, or making sure that characters that are in conflict are closer to each other, but making sure that the person who is portrayed as stronger is physically elevated to show the power dynamic. Choices like these show how differences between characters cause struggle. 

Line: I would use lines for positioning, but also some backdrop elements, such as individual lines for the fence at the border. These lines shouldn’t be straight though, to promote disorder (which represents the disorder in the characters lives).

Shape: I would use shapes for shadows in the backgrounds, and for props in front of the backdrop that are supposed to receive less attention from the viewer

Form: I would use geometric objects to make objects within a scene that should get more attention from the viewer, as opposed to the 2 dimensional ones that will be further back.

Value/Contrast/Color: I will use this in full force, I will make sure that things full of life are brighter and have more vibrancy, and things that are on the receiving end of struggle are void of color and are darker.

Texture: I would use texture mainly within costume design, making the clothes of those receiving struggle to look battered and unkempt, showing how it is affecting their life.

Scale/Proportion: I would make characters who are facing oppression or struggle look smaller by placing them lower down on the set or by placing them further away from the audience.

Motion: I will make scenes that involve movement have the characters constantly moving, and the scenes that do not involve movement having the characters stay in place to increase the contrast between the two. 

Emphasis: As previously mentioned, done through form, and additionally through light.

Balance: There will be unbalanced sets, especially in scenes with conflict.

Rhythm: Like with balance, there will be irregular rhythm, as it shows the disorder in the characters life.

Space: I will put important things closer to the stage and less important things further away (as mentioned in form)

Unity: The scenes should be slightly chaotic, but especially chaotic whenever conflict arises.

Overall, most of the production decisions fit within my analysis of the script. The big decision I feel needs to be rethought is spacing — I think that since there is a lot of motion, having characters in conflict be closer together would be very difficult to choreograph, so perhaps placing less emphasis on the movement, and more emphasis on other forms of contrast.
Throughout the last couple of years, the Jewish community has faced many positive and negative experiences, and we as a community have grown stronger and changed in many ways. Although there were many challenges in the past couple of years, one benefit is the creation of a Jewish student-led magazine — Fig and Vine. This year, many members of the Jewish community were able to contribute and write some phenomenal articles. One such member is Shani Levy-Richards, who has just graduated this year. I had the pleasure of being able to interview her and to share her thoughts to the rest of the community.

The first question I asked Shani was how she felt being a Jewish student at UCSB. She said that overall, she liked it (due to the large Jewish community). By percentage, UCSB has the largest Jewish community of any UC, leading to our community to be large and strong. She also mentioned that she liked the various events that happen within the UCSB Jewish community. Having such a great Hillel and Chabad here at UCSB is very helpful because these events are fun, engaging, and a great way to meet friends. Some of her favorite memories that made her feel like a part of the community have been Pesach at Hillel, and various events at Chabad. She says that she feels like the community here is more welcoming than other UCs.

When I asked her about what brought her to Fig and Vine, Shani told me that she is a history major, and that Jewish history has always interested her. She told me that there is so much to talk about and learn within Jewish history, hence her fascination. I agree — we have existed as a people for the last 3000 years, and within that time so many events have occurred. She specifically told me that one interesting part was how holidays evolved over time. 

I followed up with Shani and asked her if she had taken any Judaic study classes at UCSB. I have taken many, and have learned how Jewish practice and laws have evolved over time. She told me that unfortunately she had not had the time to take any classes despite the fact that she wanted to, but she had many friends who had taken Judaic studies classes, and who had enjoyed them. Personally, I have taken several of these classes with Professor Medina, and have learned a lot about our history, so I am very sad for her that she did not have the opportunity to take those classes.

After that point, I inquired about her experiences as being an author with Fig and Vine. Shani told me that she enjoys documenting the history — to her, it felt similar to a diary. She also mentioned that she feels more confident as an author than as a speaker, making it easier for her to express her thoughts. She told me that homework does not feel empowering to create, but since she started writing on her own volition, she feels very empowered in her writing.

Following, I asked Shani about her plans post college. She told me that she plans on taking a gap year between college and grad school. She wants to do a masters in English (potentially with history). Furthermore, she told me that although her time is done at UCSB, she still has a twin sister here, so she will be visiting.

The penultimate question which I asked her involved how the Jewish community has evolved at UCSB. She told me that over time, the community, through events such as October 7th, the rise in anti-semitism, and incidents of hate directed at specific individuals, has gotten more tight knit. The Jewish community at UCSB, and within the greater world has been through a lot, so in order to ensure that we are able to prosper, our communities have gotten stronger.

I then asked her if she had anything else to share, and she told me to let every know that she wishes them good luck. Overall, Shani has contributed greatly to the community by teaching us all about the history, and she will be dearly missed. Lets all wish her a successful time at graduate school, and hope that she visits Santa Barbara very soon!

Throughout history, there have been many examples of indigenous groups that face oppression from the government which is supposed to help them. Oftentimes, when this happens, the oppressed people do not take kindly to being oppressed and rebel, sometimes gaining de facto independence from their oppressors. One such example are the Zapatistas, which are formed by several indigenous groups within Chiapas, Mexico. They historically faced oppression, such as having diminished land rights, and other racist policies. This caused the indigenous people within Chiapas to revolt, and the day that the NAFTA agreement was being signed, the Zapatistas declared themselves as an independent state. This caused conflict within the region, leading to a massive military response from the Mexican government, a governor getting kidnapped by the Zapatistas and being forced to do hard labor (Delpama), failed and disingenuous attempts of peace being proposed by the Mexican government, a massacre by the Mexican government (Fisher), and many more such tragic and horrific events. The reason for these forms of violence by the Mexican government was because they wanted to quell any sort of rebellion put upon them. This is because they wanted to appear to the broader global stage as a strong, capable country, being that his was a time of great growth for Mexico, and they were finally no longer considered a 3rd world country. Even though efforts have been made to ensure peace in the region over the last three decades, no such peace has been formed, due mostly in part to the Mexican government’s refusal to address the issues that caused the Zapatistas to rebel in the first place. Such a rebellion is not unique — there are many such examples an or many indigenous groups banding together in order to fight back on an oppressor, leading credence to the validity of such movements. Overall, the Zapatista’s struggle for peace and security only exists and is exasperated by the actions of the Mexican government.

For over 2000 years, the Mayan population has thrived within Chiapas, keeping the land healthy and their home. In 1523, however, their peace and prosperity was destroyed when a colonizer, Luis Marín, started conquering the area. He was able to take over the land of many of the indigenous people, but he ran into issues fighting the Tzotzil indigenous people (Schmal). He fortunately was not able to conquer all of the land — however, the Spaniards would not give up their goal of conquering all of the Mayan’s land, so they kept fighting. By 1528, all of the land had been conquered, and the last of the indigenous people that were free were now under control of the Spanish crown. The Spanish instituted a rule of slavery for the indigenous people, further diminishing the rights of those who belonged to the land, until it was abolished. Even though this rule of slavery had ended, the Spaniards (and later the Mexicans) were always at odds with the indigenous people. Historically, “Chiapas’s indigenous people have been excluded from the governmental decision-making process as well as from enjoying basic human rights and services such as education and healthcare” (Godelmann). Additionally, the government wanted to gain land rights, so it fabricated a situations where it appeared as if “government simply was resolving land disputes between different indigenous groups” (Krinsky), and seized all of the land, despite the fact that there were many other indigenous groups that were not the “Lacandron Community.” (Krinsky) As can be seen, both the Spanish and Mexican governments have historically treated the people within the Chiapas state extraordinarily unfairly, leading to deep resentment within the people, and an eventual liberation movement.

This aforementioned tension within the region in regards to the indigenous people boiled over on New Years Day of 1994 (Mora). This day was going to be the day that Mexico joined the NAFTA (North American Free Trade Agreement), and therefore would be considered in the eyes of the greater world no longer a third world country. This came about through the Ejército Zapatista de Liberación Nacional’s (EZLN), known in English as the Zapatista National Liberation Army’s declaration of war against the state of Mexico. This was not simply preformative, as the Zapatistas “briefly occupied six towns in the state’s central highlands,” before retreating to the jungle to avoid further conflict (Harvey, Arnson, 129). This fighting continued, and further tensions with the Mexican military required the Zapatistas to resort to more brutal means of enforcing their rule, such as kidnapping “Absalon Castellanos Dominguez, the 70-year-old former Governor of Chiapas” (Depalma). Despite peace talks occurring, this tension still exists to this day (albeit less severe), as the Mexican government does not want to listen to the demand of the Zapatistas. These demands are as follows: “truly free and democratic elections,” “the incumbent federal executive and the incumbent stare executives… must resign,” “Recognition of the EZLN as a belligerent force,” the end of neoliberal policy, “General elections for the whole state of Chiapas,” among many others (CCRI-CG of the EZLN). Since these demands have never been met, the EZLN still holds itself as a de facto state, and is and continues to be at odds with the leadership and military of the Mexican government. It is important that this be known, as these demands exist due to the abuse of both the Spanish and Mexican governments, and until the Mexican government rectifies its past wrongdoings, the people of Chiapas will never accept them as a valid state and will continue to posit their own independence.  The reason that all of this conflict exists lies solely on the actions (and the lack of actions) of the Mexican government. The actions that the Spanish government did in the past were horrible — enslaving indigenous people, running them off of their land, and many other things. When Mexico gained independence, they finally had the opportunity to right the wrongs that the Spanish crown had done to the indigenous people — a fresh start. This fresh start, however, was (as previously mentioned) not such a fresh start, and in fact worsened the lives of the indigenous people. Moving towards the modern day, the modern version of the government has acted against these people. Even though their anger is rightful and justified, the government treats them as a bunch of rebellious hoodlums, and treats their demands as irrelevant and unjustified. This government treatment of indigenous people is not unique to the state of Mexico — the United States, Canada, Australia, and so many other countries have treated their indigenous people unfairly. That being said, many wrongs do not make a right, and Mexico treating their actions as justifiable just because others have been able to get away only serves to worsen the situation in Chiapas. Had the Mexican government treated their demands as they should have, not only would there be peace within the region, countless lives would have been saved, but this would only help show the rest of the world that they in fact were a competent power, and worthy of being considered a first world country. Their treatment of the indigenous Mayans whom inhabit this land has shown and continues to show the world that the government is cruel, and only serves to give more attention and positive press to the EZLN. Overall, had the Mexican government treated these indigenous people fairly, such a situation would have never occurred in Chiapas, and had they chosen to listen to the reason provided by the EZLN, this crisis for the government would be a situation of the past and not of the present.

As previously mentioned, the reasoning for the Mexican government’s oppressive rule over the Mayans and their Zapatista movement was to make their country appear more competent and powerful. While Mexico had been seen as somewhat strong during the founding of their nation, throughout the 20th century, due in part to drugs, gang activity, and weak economics, Mexico was seen as a Third World country, and was not taken seriously on the global stage. Because of this, they felt that they needed to look ‘ready’ and ‘presentable.’ The obvious humane solution to this would be to get in talks with indigenous people and figure out what would help them feel like they could peacefully live within Mexico; however, as has been seen previously, this was not the approach taken by the government. Likely the reason why this was was because of public sentiment towards indigenous people — at the time (due to previous propaganda), many Mexicans viewed the indigenous people as inferior, so taking such an approach would risk the non-indigenous Mexicans getting angry and starting an uproar. This would obviously look bad for the Mexican government as well, and as the indigenous people had fewer resources, less power, and a weaker voice, they chose to oppress the indigenous people. As has been mentioned, this oppression was the basis of the creation of the Zapatista movement. At this point, the government had had their reputation smeared — instead of showing the world that they were a competent first world country, the Zapatista movement instead showed a darker and more sinister side to the government. In response, the government retaliated against the Zapatistas, furthering the point that the Zapatistas were making about the government. At any time, the Mexican government could have accepted defeat, which would have made the current government look better, and would have better the lives of both the Mayans and the general Mexican population, but the tyrannical leaders at the time chose violence, hurting the image of Mexico, which in turn led to a spiral of more violence and tyranny.

This type of struggle, while having unique facets, unfortunately is not unique to just this localized instance. A great example of this comes from geographically close to Mexico,  Guatemala. The indigenous Mayans of Guatemala had also faced many hardships for generation upon generation, and despite the changing world of the 21st century, this trend continued and continues into the new millennium. For example, there is currently a “fourth invasion” going on, which is distinguished from three previous invasions: first, the Spanish invasion and colonization; second, the creation of the plantation economy in the late nineteenth and early twentieth centuries; and, third, the state-sponsored genocide that occurred during the Guatemalan Civil War (1960–96)” (Batz, 3). This invasion is due to the government deciding that they are going to, without the permission of the indigenous people whom live there, build a hydroelectric plant. This is very similar to the aforementioned example of the Mexican government seizing land for their own purposes — just because a country is within a certain area does not mean that it has full autonomy and privilege to force people off the land for its own gains. After this, the government sent in the military to destroy the protests that started because of this declaration of construction. This too is very similar to the aforementioned example in Mexico, where in 1994, the government sent in the military to attempt to (and not successfully) end the defect state that the Zapatistas created. This is a common theme worldwide — governments believe that they are always righteous, and therefore possess the right to do whatever they want to quell rebellion and dissenters because they always must be ‘wrong.’ Unfortunately, this line of thinking has been implanted in the minds of many, so one must fight against it rather than waiting for this line of reasoning to disappear.

Although situations like the two that were previously mentioned are grim and dire, it is important to remember that fighting for rights is important, and has, does, and will continue to positively benefit the people who are fighting. A great example of this comes from the United States of America, specifically Colorado. Many indigenous tribes lived along the Colorado river for millennia upon millennia, until the descendants of white Europeans, specifically members of the Church of Latter-Day Saints, colonized this area and forced the native people out (Sakas & Bures).  This river was essential to these indigenous people — their farming and tradition was built upon it. In 1922, the river was divided among many states — though this did not give the indigenous people any rights to the river. This left the indigenous people of that area in a very tough spot — they now had even less access to the river they barely had access to for the past couple of centuries. They then protested this, and fought tirelessly for decades to ensure that they got access to the river which they called home. Finally, in the 1990s, all the indigenous tribes of this region finally got what they so desperately needed and deserved. This example shows the efficacy of fighting against the government to get what is needed. In this situation, the time required to get the government to meet the demands of the indigenous people exceeds the time which the independent Zapatista movement has existed. While people may look at the Zapatista movement and believe that it is just a waste of energy for the people there, it is important to remember that the United States government is just as unwilling to help indigenous as the Mexican government, but even for a self centered government like Mexico, eventually, they will give in, hence the importance of fighting on continually.

In conclusion, indigenous people all over the world face oppression by their governments, and the indigenous groups that form the Zapatistas are no exception. The indigenous people of Chiapas during the rule of the Spanish crown fell victim to oppression, slavery, and mistreatment. This mistreatment did not stop during Mexico’s independence from Spain, and continues on into the modern day. Because of this, the indigenous groups of Chiapas banded together, and declared independence from Mexico on January 1st, 1994. Rather than listen to the demands of the Zapatistas, the Mexican government sent the military in and unleashed violence upon the people who were living there. The reason that the Mexican government unleashed such a strong response was because they wanted to be seen as a strong, 1st world nation, and by being oppressive to its subjects, ended up doing the opposite. The whole reason that there is conflict in this region in the first place is entirely the fault of the Mexican government and its inability to listen to the demands of the people who they oppress, and is the responsibility of the Mexican government to fix. Had previously the Mexican government listen to the needs of the indigenous people of Chiapas, no such de facto state would have been formed, and had they listen to the needs of the indigenous people after the de facto state had been formed, the ensuing violence, death, and embarrassment for the government would have never happened. This situation is not just unique to the Zapatistas — many movements like this exist within the world (in smaller scale or larger grandeur), and many movements by indigenous people have been successful (such as the aforementioned example with the Colorado river). Overall, the existence of the Zapatista state and the violence that has ensued due to it are the fault of the Mexican government, and until they address the issues that the Zapatistas ask, no positive change will occur.

1. There are many features of Rabbinic literature that are very different from the Hebrew Bible. First of all, whereas revelation was something that occurred through prophets in the HB, within Rabbinic literature, it was carried out through knowledge and interpretation of Jewish texts. Two additional features that distinguish Rabbinic literature are makhloket and safek. Makhloket refers to disagreements, where multiple Rabbis would have opposing opinions on an issue and would have them debated, and safek refers to doubt or uncertainty, which was not discussed as often within the HB.

2. The emergence of the Torah scholar contributed to this phenomenon of disagreement because rather than just understand the Torah, Torah scholars now both understood and interpreted the Hebrew Bible. Since the Hebrew Bible was not ultra specific, and oftentimes vague, this could lead to disagreements in interpretation.

3. The sages disallow this. Even when a divine voice spoke out to the sages, they still stated that it was required to show proof about a matter in order to have it count for debate.

4. The Tannaim and Amoraim have very different views about Torah Scholars. Whereas the Tannaim believed it best for one to understand the Torah, the Amoraim placed the importance of Torah study on debate and interpretation. Additionally, the Tannaim view Torah study as something that is done in addition to a job, whereas the Amoraim see it as a job within itself. There are two very good examples of this, one of which is the gathering of manna (or specifically when manna should be gathered), and the story of Rav Shimon Bar Yochai in the cave.

The first example comes from the gathering of Manna. It was stated by the Tannaim that it should be done once a day, every day. While this might seem cut and dry, the Amoraim suggested an interpretation about this that it is permissible to collect manna for the day and for the day that follows, as is done for Shabbat. This shows that the Amoraim view a Torah scholar as one who can interpret information from Jewish texts, whereas the Tannaim are just recounting events and laws.

The second example comes from the Babylonian Talmud. In this story, Rav Shimon Bar Yochai emerges from a cave, and his son in law sees that he is in very poor physical condition and starts to cry. Rav Shimon Bar Yochai states that he is happy, because although he is low in physical health, his knowledge and understanding of the Torah is much higher. This is important, because it shows how the Amoraim view Torah study as work (as he locked himself away in a cave to study Torah), rather than a Tannaitic Torah Scholar who would have lived within their town and had an additional job.

5. Makhloket is a very important part of Rabbinic law, as it serves to show differing opinions of Rabbis. The jug-breaking case is a great illustration of this, as the original law simply states that an owner is liable for damages for a jug left within a public area. The makhloket surrounding this discusses in which cases this law is considered and several opinions are presented, two examples of which are this applying when the jugs completely block the road and when it is around the corner. This reveals that makhloket  presents differing opinions of Rabbis.

The Mishnah preserves minority opinions even if they aren’t the ones being followed because in the future, a court might change its view. If in the future what once was a minority opinion becomes the majority and it is viewed as being of a higher spiritual status, then this opinion is the one that is followed rather than the previous majority opinion. This shows that the Rabbi’s held specific situations and common views as important facets of laws.

A contemporary example of this that is important is the discussion surrounding Roe V. Wade. While this ruling had been previously followed, the supreme court in 2022 decided to overrule the previous decision of the court. The minority opinion was stated in the original ruling, which was cited in the new ruling, showing a modern example of the importance of recounting minority views.

6. This principle is important in declaring a husband dead because the consequence of a wife remarrying does not change based on how long she waits (should she be wrong about the death), but the certainty of the husband being dead goes up over time. For this reason, the Rabbi’s decided that a 3 day waiting period would be sufficient for determining a husband dead. Even though an example of somebody being alive after 3 days was brought up, the Rabbis still held that there was high enough certainty after 3 days for a wife to remarry.

The contrary example to this is saving a life on Shabbat. If somebody were to be covered in rubble, the Rabbis deemed it to be permissible to remove rubble so long as they did not find that the person was dead (at which point they should stop). This is for two reasons — unlike the previous example, this example is time dependent, for which the Rabbis allow more leniency. Additionally, the mitzvah of Pikuach Nefesh supersedes all other mitzvot, hence why the Rabbis allow this example.

7. Within this passage, the majority view a vessel as being complete when it is acts as a functional vessel, whereas Rav Eliezer views it as complete when it appears as a vessel (with the caveat of something being repaired with sand not being considered as appearing as a vessel). This is important, as it shows two different legal approaches — one where the law is strictly defined — absolutism (as in Rav Eliezer’s case), and one where it is defined by the view of the majority. 

A modern example that illustrates the difference between these views this would be the speed limit on the freeway. Somebody who helped a strict interpretation of the law would state that anybody going 66mph or faster would be breaking the law and should be fined, whereas the majority view (and what the majority of people do) Is follow the speed that traffic is moving on the freeway rather that driving at exactly 65 mph.

8. Throughout its history, Jewish law evolved a great deal. In the beginning, the Torah was simply a partially known tool that was used for teach — however during the second temple period it became both fragmented (due to the sects) and statutory. After the temple was destroyed, the law became much more interpretive and united as compared to the 2nd temple period.
 
Jewish law began during the biblical period (First Temple). In this time period, the law was viewed as pedagogical rather than being statutory. In other words, this meant that the laws were meant to teach, rather than to be strict ways of life that must be followed. Additionally, for much of the first temple period, the Torah was hidden, so the exact laws were not followed (a good example of this would be the King David and the spoils of battle). 

Within the 2nd temple period, a sharp difference emerged based on how the law was interpreted. Laws were followed much more strictly, and were now seen as a legal code rather than teaching. Additionally, sects (such as the Pharisees and Sadducees) forms, each having their own interpretation of the law. Within the Qumran community for example, they did not view the 2nd temple as being valid, and therefore set up their own community near the Dead Sea. These changes occurred due to the Torah being better known, and questions about the validity of the Maccabees as leaders of the Kingdom of Israel.

After the fall of the 2nd temple, the Rabbinic period began. The Rabbis did not want a repeat of the sects from the 2nd temple period, so they allowed more flexibility within interpretation of the law. To be more specific, when the Amoraim constructed the Gemara, they recounted multiple views so that future courts and different communities could have flexibility under the law. Additionally, within this time period, the law evolved too — starting at the beginning with the Tannaim, the Mishnah was created which was a recount of the oral law, and later during the period, the Amoraim made the Gemara, which provided further interpretation of the Mishnah, both of which make up the Talmud.

Overall, throughout Jewish history, the use of the law, and how the law was used has changed significantly. In the beginning, the law was a tool for teaching and was partially unknown, after which (during the 2nd temple period) became a strict statutory code while also being fragmented by the multitudes of sects, before Judaism became united again during the Rabbinic period, where interpretation of the law and court rulings became important.

Bonus. During this course, we went much more into depth about various interpretations and discussion/debate about the law (makhloket) than we did during the Judaism class during the fall. My takeaway from this is that multiple opinions can be correct because of the many context — this relates to how multiple opinions were recounted so that future courts could change the interpretation of the law if they saw fit. This is important in life because although the world may seem very black and white, this simple fact shows that it in fact is not so black and white, and in many cases, a different perspective is not an incorrect perspective, but simply just a different perspective.

Journal 1:
Reflecting on this journal, I feel that I have learned a lot about how people are educated about indigenous people. Going into this class, I had very little experience studying the lives of indigenous people (specifically how it relates to the state), and having learned that the phenomena of indigenous people’s interactions with the dominant power of an area often being misconstrued is a very common occurrence. I feel that it is very important for people not only to learn that specific examples are wrong (such as the pilgrims being good guys example), but that this practice is very common, and understanding why it happens.

Journal 2:
I think that what I wrote in this journal rings very true. As a linguistics student, I can also think of many examples about colonization and power dynamics affecting language (think about how the large bed room where a slave owner could have slept is called a “master bedroom”). While these racist and prejudiced pasts to certain parts of various languages lexicons may not be apparent in every day speech anymore (for example the greeting “long time no see” originally was a mimicry of what English speakers thought that Chinese speakers speaking English sounded like), but should be taught about.

Journal 3:
I feel that this was the journal I forgot most about throughout the course, so revisiting it was very nice to do. As a linguistics major, I have learned about how certain words can evoke certain emotional responses, and an otherwise trivial difference in word choice can be the difference between the message that you want to have sent being sent and not being sent. I believe that the choice of words being “rage” rather than “fear” can evoke an extraordinary difference in emotional response, and changes the role of the women from sounding like weak and helpless to strong and powerful — which they are.

Journal 4:  
I consider myself to be an environmentalist — and as somebody who considers himself an environmentalist and cares about how indigenous people are treated, this journal hits home on two different fronts. For example, the anecdote from Batz about the dam constructed on Mayan land without the Mayan’s consent bothers me both because this dam can potentially harm the environment in drastic ways, but also because it was an infringement on their land. I believe that both of these factors are very important when holding governments accountable — oftentimes people like to focus on one or the other, but both are meaningful and important issues.

Journal 5:
I think that blood quantum are a very important topic to be discussed. I did this journal entry before we watch the video about the Indigenous people in Canada, and how they were not treated as a part of their community, and I believe that that is horrible, and the culture that you grow up in determines your ethnicity and what community you belong to, not your genetics. Additionally, oftentimes these blood quantum are only a problem for women who intermarry rather than men, not just making society more xenophobic but also more misogynistic. Overall this is an important issue that I hope more people will start to discuss.

Journal 6:  I think assuming that western education, and the systems that surround and govern it are the default and nest way to educate is a very dangerous assumption to make, and a myth that needs to be distilled. For example, Finland has some of the highest test scores in the world, and they spend significantly less time learning. Additionally, many forms of western education are competitive rather than collaborative, which can lead to heightened anxiety and an overall worse outcome for science as compared to those who are not stressed. It is important that these myths get distilled, and that we all learn the truth.
Journal 7:
This journal was speaking about how certain indigenous communities in Latin America have been fighting for their rights, in regards specifically to their culture. This is a very important part of learning about how indigenous people get affected by the state — indigenous people are not idle ducks — they will fight a gains unfair laws and orders put upon them. It is very important to remember that learning about these fights against tyrannical laws are just as important as learning about the tyrannical laws themselves — having information about both helps us better understand the context of ever indigenous struggle in the world.

Journal 8:
I really enjoyed this talk — I am a very big dan of music (I have been playing piano since age 5, I love listening to music and making music), so I can appreciate the importance of music to both individuals and to a larger culture. I enjoyed learning about all the ways that the indigenous communities in Oaxaca preserved their culture, but the music part specifically stood out to me for the aforementioned reason. Overall, my main takeaway from this week’s journal entry is that there are many important facets to one’s culture, and finding ways to preserve them is very important (such as with the music)

Journal 9:
I really resonated with this journal entry because I really care about nature and I do my best to be in harmony with it. I love going on hikes with my friends, and loving nature for what it is and has to offer shows me that the oftentimes overly urbanized capitalist society does not appreciate where we came from. I always do my best to keep nature clean and safe — there only is one Earth upon which we live, and taking care of it is very important. That being said, the Earth (as discussed in the reading) can benefit sometimes from being taken from, so it is important to be a part of the planet and to love the planet.

Journal 10:
I believe that state intervention is a major problem when it comes to indigenous issues. Although this is not necessarily about indigenous rights, what has been currently going on with ICE and the Trump administration is a good demonstration about the problems with state intervention. By the state not listening to the demands of the protestors and instead sending in the ICE and making the situation more dangerous for everybody, they end up aggravating people and showing people that this aforementioned kind of intervention (rather than listening) is a major problem in our society today, in regards to the state.

Abstract

Introduction. Patient evaluation and diagnosis is important in cases of aphasia to provide good treatment and support. Case Study. A YouTube video of an aphasic is watched and a diagnosis is provided from said video. Results. A transcription was made for the video, and it was found that the patient had difficulty forming long sentences, and oftentimes fell back on familiar language. Diagnosis. It was determined that the patient in question has Broca’s Aphasia. Discussion. Correctly diagnosing and treating Aphasias is important to help the patient, and give closure to them and their family.

Introduction

Learning about aphasias and the type of symptoms that they produce is important as it can help patients receive the care they need, and help those around them understand what is happening and why. Additionally, since aphasics can get better over time, it is important to diagnose it so that they be in an environment conducive to brain recovery, such as language rehabilitation. While many people might rush to diagnose the aphasia quickly, getting language rehabilitation at any point will help about the same amount (Moss & Nicholas, 2006). One of the major types of aphasia, Broca’s Aphasia, is caused by damage in BA 44 and 45 and patients often have issues with speaking fluently, finding the correct words, articulating well, repetition, and the production of complex sentences (Dronkers et. al., 2007). Because other language tasks, such as semantics, and comprehension, occur in other parts of the brain, even though they are closely intertwined with tasks associated with Broca’s Aphasia, they are not affected in Broca’s Aphasics (Schwartz et. al., 2009).

Case Study

The person being discussed is an elderly woman, Jan, who is having a discussion with her daughter, Carly, on a YouTube channel they share dedicated to informing people about aphasia. The specific video is about how to have a good conversation with an aphasiac.

Results

Within the conversation, she had difficulty creating complex sentences, and saying the name of a number (although she could write it down). Additionally, her daughter (who also is her caretaker) already had a sheet on hand with many activities on it, further indicating that she had problems expressing her thoughts. Jan additionally had multiple phonemic paraphasias. Patient also extensively used the word “yeah,” as this was the only interjection used by Jan (at least during this recorded snippet). Additionally, almost every utterance she said was very short, lending to the fact that it was difficult for her to form long utterances.

Diagnosis

Because Jan’s speech was at times telegraphic, and she had word finding difficulty, the  diagnosis that is given is Broca’s Aphasia. Jan showed a high level of comprehension (ruling out Wernicke’s Aphasia and Global Aphasia). There were also many signs of Broca’s Aphasia, such as a common word or phase being often used (in this case “yeah”), and the knowledge of a number, but not being able to say the name of it.

Discussion

As previously discussed, the diagnosis and understanding of diseases like Broca’s Aphasia are important as they can help get patients the best treatment that they deserve and give closure to both them and those close to the patients. Since certain aphasias are caused by certain areas of the brain (Schwartz et. al., 2009), being able to know what is and what is not affected for patients like Jan is essential because treating her like a Wernicke’s Aphasiac will not help her in any meaningful way, and would be mentally taxing for her. Giving her the correct treatment could, as previously discussed, bring back language functionality to her brain (Moss & Nicholas, 2006), further stressing the importance of correctly identifying and treating aphasias. Additionally, knowing a rough idea of the area in the brain in which the the stroke occurred is helpful because it can predict other symptoms of the stroke. Since Broca’s area is near BA 44 and 45 (Dronkers et. al., 2007), other related processes that occur in the frontal lobe nearby will likely be affected. Overall, correctly diagnosing and treating aphasias is vital to the efficacy of short term care of patients, the potentiality of long term restoration of language skills, and the closure that the patient and their family deserves.

Judaism and food have a long and complex relationship — we have entire holidays dedicated to abstaining from eating. We have certain foods we can’t eat, and certain foods we must eat. And finally, lets not forget, gefilte fish. Jewish culinary culture is formed by these very laws, making Kashrut (kosher laws) an important facet of our food and our culture.

Growing up, I kept no form of kosher, and up to going to school here, I continued this trend. If you were to ask me what my all time favorite food was, I would have enthusiastically said “shrimp!” Having come to college, I formed a deep connection with my Judaism, and in turn, with Kashrut. While I might not be fully orthodox kosher (I will still eat at non-kosher restaurants as a “pescatarian” with a “shellfish allergy”), I do keep a somewhat strict degree of Kashrut. I may be a Jew in IV who keeps kosher, but I am not the only Jew in IV that does, so I wanted to see what others’ experiences with Kashrut in IV and at UCSB were.

The first person to whom I posed the question of Kashrut in IV was Nicole Zar. She told me that she grew up keeping kosher, and that this was important to her, so she kept keeping kosher in college. To her, it is totally doable to keep kosher here, despite the many obstacles. She said that she often relies on Chabad for meals (especially on Shabbat). “Although it is not simple,” she told me, “it is a very rewarding part of Judaism that [she] takes pride in and a mitzvah that [she] enjoys keeping!”

As Nicole said, UCSB Chabad besides being a keystone to the Jewish community is also one of the best sources of Kosher food in Isla Vista. Miri Klein, the Rebbetzin at UCSB Chabad, is the mastermind behind all that is food at UCSB Chabad and works tirelessly to ensure that Jewish mouths in IV are fed. She told me that the biggest challenge for providing so many people food is that there are many difficulties getting large quantities of food — either the foods need to be bought from a kosher market, or comes in very small quantities. She suggests that anyone who wants to start keeping kosher contact her with any questions, and go to Trader Joes.

I was able to get another perspective from Jake Nguyen, another student who keeps Kosher. He said that personally, he has not had to make any concessions with Kashrut when coming here, and that the only downside to a kosher diet is that it is more expensive. All of his friends are also supportive of his diet, making the social aspect of living in IV easy. He also recommends Trader Joes to anybody who wants to keep kosher, as it is the only place nearby with kosher meat.

Like for Jake, Trader Joes is also an important part of keeping kosher in IV. Most of the groceries I buy are produce (which don’t require kosher certification), meat, fish, bread (which is more expensive), rice, and pasta. Thankfully (besides produce), Trader Joes has all of this, so if anybody reading this wants to keep kosher, Trader Joes is a great place to start!

Another student, Idan Fierstein, shared his experiences with me. He grew up in New York City, where there were many kosher restaurants, and now living here is very difficult in that regard (the closest kosher restaurant is over 70 miles away). As such, he eats out in restaurants (similarly to how I do), but he says that he appreciates kosher meals more than he used to due to the complexities involved with eating kosher in IV. He hopes that we can get a kosher food cart in IV like they have in SLO.

Eating out “kosher style” in non-kosher restaurants is something I understand — even if there were kosher restaurants in IV, it would be very hard to stay social especially with how often social plans revolve around a restaurant. A way I have found to bridge the gap between keeping kosher and staying social is having plans revolve around cooking in my apartment — even though my apartment isn’t kashered, everything I use for cooking is kosher (including hard-to-come-by kosher items like meat and balsamic vinegar), and as of right now, that’s kosher enough for me!

I do try my best to keep kosher, even if I am not following the laws to the dot of how they are in the book. Others, however, take a more liberal approach to Kashrut. When asked about his favorite kosher dish to cook, one Max Bloom responded: “I’ve gotta go with the cheeseburger.” For those not in the know, cheeseburgers have not been, are not, and will never be kosher, unless you are Max Bloom. More power to him! 


As you can see, keeping Kosher in Isla Vista is no easy feat — the odds are stacked against you, and unless you live exclusively with others that keep kosher, there will be some difficulties. While not every Jew keeps kosher, those who do feel a strong connection to the food they eat, and are willing to make sacrifices to keep their diet how they wish. Overall, Kashrut in Isla Vista comes in many colors, shapes, and forms, and is an important part to the Jewish community in IV.

While “The 13 Colonies” have not legally been colonies for 100s of years, and do not operate the same way they did when they were considered colonies, the effects of colonialism still affect our everyday lives in many ways. A good example of this is the perception of race. Within the reading for this week, it is stated that "Cherokee identity is socially and politically constructed,” stating that much of what and who is considered Cherokee is largely not decided by the Cherokee  (Strum, 2). These harmful labels affect every ethnic group in America — anybody of European descent is clumped together as “white,” or anybody of African ancestry is labeled “black,” even though Africans who migrated here in the past couple of years have completely different culture and struggles than African-Americans whose lineage has been in the United States since the slave trade.
Another example of colonialism has to do with who lives where. I go to school at UCSB, and although it is a nice institution, 1000 years ago the Euroamericans who constructed this institution did not even know about the land on which this university sits. This is not exclusive just to UCSB, as according to the textbook, “between 1808 and 1835, Cherokees were increasingly confronted with U.S. expansion into their territory in the Southeast” (Strum, 52). Both the Chumash people who lived in the land which now holds UCSB, and the land in which the Cherokee people lived is now land owned by wealthy descendants of colonizers, even though in both example the indigenous people were here first. This is yet another example of how colonialism still affects our lives today.
In conclusion, while the US is not actively expanding its territory, and many colonial practices are illegal (such as slavery), the scars which these activities left on society many years ago are still visible today. Race labels are oftentimes arbitrary, and don’t reflect the actual experiences of ethnic groups, and yet they are still used today. Additionally, the land which many of us inhabit belonged to indigenous people for thousands of years before it belonged to us, and yet we are the ones who live there, not the indigenous people who were forcibly removed. These examples illustrate how the effects of colonization still remain with us today.

My first memory about “Indians” was from 1st grade, when we were learning about Thanksgiving. The holiday was introduced to us, and we learned the typical story — Christopher Columbus accidentally discovered North and South America when he tried to get to India by sailing west. He called the people there Indians (which was the explanation as to why they were called “Indians” even though they were not from India. After Europeans learned about these new continents, they wanted to move there. One of those groups of people from England, the Pilgrims, ended up caught in a storm and it took too long to get to America, and they ended up landing in the wrong place. By the time they got off the ship, they were all starving, and there was a nice group of Indians there who helped them, and taught them how to keep the land. They ended up having a great big feast, which is why we celebrate Thanksgiving today.

This narrative went unchallenged until middle school. We first learned that no, not only was Columbus not the first person to discover the Americas, he wasn’t even the first European to discover the Americas, just the most famous. Additionally, we were told another important part about him — mainly that he did horrible things to the indigenous people that he encountered.

We also didn’t learn about the fact that both Indigenous people and Indians do not like when Indigenous people are referred to as “Indians,” and how that diminishes their identity. Calling them Indians while mentioning that they aren’t actually Indian implies that they are somehow not worthy of having an endonym to refer to them.

Additionally, the story of the first Thanksgiving is a gross misrepresentation of what actually happened. While the Indigenous people did help the Pilgrims, there was no feast, and the second that the Europeans were able to kick the Indigenous people out in order to colonize the land.
Architecture has been an important part of history, being with the human race ever since the beginning of civilization, and continuing to be an important part of humanity to today. As society started to progress, architecture started to prioritize form over function; however, recently, this trend has started to reverse. A prime example of this is Manzanita Village at UCSB, which was designed to house 1100 UCSB students. (UCSB Campus Housing, 2024). Being built in 2001, it is relatively recent, and features a minimalistic approach to its housing, using its space to maximize occupants rather than feature complicated designs. Simple shapes such as rectangles and triangles are used, as these shapes are cheaper and more space efficient than more ornate designs. In addition, architectural features that provide little to no functionality to the building such as complicated pillars or statues are not used, opting instead to make every architectural piece of Manzanita Village to contribute to the practicality rather than the aesthetics of the site. Overall, Manzanita Village is a great example of the trend within contemporary architecture to prioritize function over form known as functionalism, saving both money and time for contemporary architectural works.

Manzanita Village’s architectural designs were implemented to maximize space, as having more space per building means that more students can be housed. Rather than using just wood or steel, Manzanita Village is built out of primarily reinforced concrete, while incorporating wood and steel as a minority material. Concrete as a material has many advantages, such as low cost, durability, and the ability for the rooms to cool themselves at night (Bell and Rand). Advantages such as those can help maximize occupancy — for example, a low cost means more buildings can be built, and the rooms being able to cool themselves at night removes the need for air conditioning infrastructure to be build, further maximizing space. Additionally, the buildings of Manzanita Village are built with several buildings in the shape of a ‘U,’ allowing for three buildings to share a communal lawn, rather than all the buildings sharing a lawn. With one large communal lawn, a significant amount of money would need to be allocated to watering the grass, which is money that cannot be spent on domiciles. These are examples of functional contemporary architecture, as rather than allocating money towards unnecessary additions such as a complicated cooling system, or a large, luxurious lawn, simplicity and efficiency were paramount while still allowing social functionality.

Additionally, Manzanita Village features a minimal design, opting to use basic shapes rather than intricate designs. Functionalism is an important part of contemporary architecture, with contemporary architecture rejecting the notion that complicated intricacies are what make beautiful architecture. Since the industrial revolution, emphasis has been placed on function over form, leading to simpler designs meant to be functional rather than be flashy (Giedion). The most utilized shape within the construction of Manzanita Village is the rectangle, which is the shape used for the buildings itself, the windows, window shades, and the rooms within. Triangles are also put to use, as they are the primary shape of the roofs. These shapes provide important structural integrity, as do more elegant designs, but rather than be complicated, the beauty lies within the simplicity of the design. When designing this housing complex, intricate designs could have been used, but the designer opted for simple geometry, choosing a functional approach over a flashy approach. This trend is very common in contemporary architecture, where the beauty comes from simple shapes laid out elegantly (ie a decorated shed), rather than using intricate shapes that are both cost and shape inefficient.

Furthermore, every part of Manzanita Village provides some sort of functionality in addition to visual appeal, which is a defining trait of contemporary architecture. Many older architectural works contain details such as fountains and statues that do not add any functionality to the building, which is a characteristic which seldom exists within decorated-sheds, which is the type of functional buildings that are used for Manzanita village. As previously mentioned, Manzanita Village is made up of several buildings for housing, several lawns for students in which students fraternize, and paths connecting the buildings to each other and to the rest of UCSB. Every part of each individual building, and the village as a whole, contribute to the functionality in some way, such as the roof slant being pointed towards the grass to redirect precipitation to water the crops, or the colored awnings shading the windows as to not let in too much light. Within older styles of architecture, many ornaments or architectural features may have contributed to the allure of the work, and yet not contribute to the functionality of the constructed object. Many of the features of Manzanita Village provide visual appeal, and yet they still contribute to the functionality of the housing complex. As such, Manzanita Village is a great example of the ability of contemporary architecture to provide both visual appeal and functionality to a constructed object. 

In conclusion, within the long history of architecture, there have been many trends, and within contemporary architecture, one important trend is functionalism, the minimalist trend of prioritizing functionality over aesthetics (while still taking those into account) within architectural works. Manzanita Village at UCSB, being built in the 21st century, is a contemporary architectural work that is reminiscent of this functionality of form mindset. Manzanita was built in such a way that occupancy (the main focus of the building) was maximized by minimizing cost of materials, and using the environment, not AC, to cool the building. Additionally, simple geometric shapes were preferred over more complicated shapes that require more space and more money to construct. Furthermore, every part of the architecture contributes to the functionality of the buildings somehow, with none of the architecture solely being built for aesthetic purposes. While previous trends in architecture aimed to add ornate details to add beauty, contemporary architecture (such as Manzanita) shows that beauty can exist in solely functional architecture. Overall, Manzanita Village shows the trend of functionality within contemporary architecture, opting for designs that optimize cost and provide to the building without sacrificing the aesthetic beauty of the building.

Sociolects are a part of everyday usage of language that very few people besides linguists are aware of —the way we speak around others is shaped by the interactions and relations that we have with said people. Within a romantic relationship, a sociolect forms very quickly, with speakers having mannerisms, slang, and types of language use unique to the relationship. This sociolect is unique to the couple, as it reflects their shared identity within the relationship, and their history.

Within a relationship, sociolects can be used to relate information faster — a common understanding between two people means that less information needs to be transmitted. An experiment was performed by Herbert Clark and Deanna Wilkes-Gibbs in their 1986 paper, where two participants had to solve a puzzle involving tangrams without seeing what the other person was doing. Clark and Wilkes-Gibbs’s paper was the basis for the experiment in this paper. In the experiment done for this paper, participants are given the same puzzle, and the time taken for them to correctly place each tangram was noted. The test was given to see if throughout the experiment, the overall time required to determine a tangram decreased (after the initial go where the puzzle was new to both participants), or if the time remained consistent. The time taken is an indicator for sociolectal development, as a shared sociolect will be used to consolidate dialogue between participants, shortening speech.

For testing, two people in a couple (A and M) were given either a picture with tangram shapes on it in a specific order, or were given tangram pictures individually to sort. There were three trials, and for each one, the person who gave instructions and who sorted pictures switched — for the first trial for example, A gave instructions and M sorted pictures, and they switched for the second trial. All the puzzle pieces were the same for each one, just in different orientations. Participants’ views were blocked, so that they could not see what the other person was seeing, forcing them to rely only on vocalizations to communicate how to orient the puzzle pieces. Participants were told that they were going to be recorded, but were not told for what the recording was being used. After the trial, the recording was transcribed, and times for each puzzle piece were noted for each trial. While times for the first pass were noted, they were not used in the results, as the purpose of the experiment was to evaluate if further development occurred between turn two and turn three.

The results of the aforementioned experiment were as follows: between trials two and three, the average time taken for the describer to correctly explain to the matcher the location of each tangram more than halved. Additionally, the maximum time taken for a single picture also decreased between trial two and three (0:37 vs 0:26). Throughout all trials, the matcher always correctly placed every single tangram. While in trial 2, several of the shapes had been given impromptu names from trial 1 such as “eroding IV cliff,” by trial 3, most tangrams had been given names and/or identities that were shared among both participants (there was little to no confusion from the matcher as to whom was being referred by the describer).

In conclusion, the findings of this experiment show that even after having had seen and heard all pictures described once, further development of the sociolect of A and M occurred. It is already known that the number of turns after each trial gets lowered (Clark and Wilkes-Gibbs), showing that this affect exists within the general population, but after conducting this experiment, it is shown that already fleshed out sociolect (between a couple) can still be developed due to new circumstances. This supports the hypothesis that sociolects continuously evolve, as a non-evolving, already developed sociolect would likely have very similar average times for pictures, especially after all the pictures had been described once and the participants were comfortable with both the pictures and the experiment. Although these are interesting results, there is a major limitation with this study, as only evaluating one couple is much too small a sample size to have conclusive data. That being said, a similar experiment that should be performed would be to evaluate how strong this effect is on multiple different types of relationships (ie have co-workers, a boss and employee, a couple, and two random strangers), and see which one’s average time decreased the most between trial two and trial three. Overall, these findings show that sociolects, even in individuals whom are well acquainted, continuously evolve over time due to new circumstances and conversation topics.

The memorial I will be constructing will be in Agoura Hills, CA, and will memorialize everybody who unfortunately lost their lives in the Woolsey Fire during late 2018. Specifically, this memorial will be in Chumash Park, as this is a very frequented area with Agoura Hills (it is right next to Agoura High School, and many events, such as the Great Race of Agoura take place here). It will be built out of wood and concrete, and it will be approx 4ft x 2ft x 7ft. The bottom 3 feet will be the base, whereas the top four feet will be the statue itself. The concrete will be smooth, and there will be a small notice on the bottom that says “Please do not touch”. The monument will be a concrete statue, the left half is a heart design, while the right half will an anatomical heart, representing the physical life and emotional connections that were lost respectively. Below that, there will be a placard with all of the names of those who lost their lives. The names on the placard will be in the following format: “In loving memory of:” as the title, followed by: “(Birth Year) - 2018.” The wood holding up the memorial will be light birch wood, and the placard will be black with golden writing (there will be braille below each line as well), with a golden border. The platform on which the statue will be and that contains the placard will be a trapezoid, with 3 sides being rectangular, and the last (the one that contains the placard) being slanted. The monument will be about 250 ft away from Argos street, to ensure that drivers passing by don’t look at the monument too intently, as to cause an accident, and there will be a concrete path leading to the monument itself.

There are many memorials in the world, most of them focusing on either just one person, or an enormous amount of people (think wars, terrorist attacks). My memorial will fit somewhere in the middle, focusing on those who lost their lives during the Woolsey fire. The message of this memorial is two fold — to honor those who lost their lives, and to educate those who see it about the dangers of fire and fire prevention. This location is a strong location for the memorial, as it covers the area affected by the fire. In addition, the form would be a great way to convey the meaning of the monument, as the statue in conjunction with the placard will convey the importance of the memorial. In addition, this monument serves as a counter-memorial, as it serves to remind people about the lives that were lost due to the negligence of Southern California Edison.

My monument is going to convey two messages: first, this monument is going to act as a memorial for all of those who died tragically during the Woolsey fire. While many people know about the Woolsey fire and that people died during it, having the names written on a placard will remind people that families got destroyed and souls who should be with us are no more. While the more recent LA fires may have been more destructive, the Woolsey Fire impacted Agoura Hills specifically, hence the choice to commemorate that event. The second message that this is going to convey is a warning about the dangers of fire. This fire was caused because Southern California Edison (a power company) failed to shut off the power during extraordinary high winds, causing a major fire in SoCal’s history. In addition, many people chose to reside in their dwellings despite evacuation notices, causing them to unfortunately pass away. This monument will remind people of both of those, that they should be appreciative and not angry when the power company shuts off the power during high winds, and that listening to evacuation notices, no matter the circumstances, is crucial and lifesaving.

In addition, this monument is going to be located at Chumash Park because that location serves as a meeting point of the community in which it resides, and also happens to lay in the wake of where the fire passed. Within this park, many events take place, and besides organized events, this park serves as a wonderful picnic spot for many within the community. Because this is such a frequented area, having the monument located here allows many more within the community view the monument, furthering the spread of the message of the monument. Additionally, Agoura Hills was one of the locations that was evacuated during the Woolsey fire, hence why it should be located here. Many people within the community know somebody who lost their home during the fire, therefore having this monument here can serve as a reminder to those people of the importance of preventing fires.

Additionally, constructing the monument with the aforementioned design is a great way to convey the meaning of the memorial, as the placard serves to honor those who have passed, with the statue further reenforcing this meaning. Although on a conceptual level, everybody knows that people have passed, many people (especially after so much time) have forgotten what that truly means. Having the names of the people who passed during the Woolsey fire serves as an important reminder that those who we lost were people with names, families, and interests, and that they will never get to experience what life has to offer anymore. Additionally, the imagery of the statue serves as a reminder to what was actually lost: the anatomical heart shows that people who were once living are no longer living, and the heart design represents the emotional connections and love that were lost during this horrific fire. 

Furthermore, this monument serves as a counter-memorial rather than a memorial, as it both serves to commemorate those who were lost, while acknowledging the negligence of the power-companies. A memorial is defined as something that celebrates life and glorifies an event, whereas a counter-memorial is defined as a monument that recognize darker events and serve to teach, rather than glorify (Kerby et al.). The purpose of this monument is to teach about the dangers of fire, and the importance of power companies shutting off the electricity due to high winds. If this monument were to be constructed as a memorial, it would glorify rather than villainize Southern California Edison. Since the purpose of this monument is to hopefully prevent such a disaster from ever occurring again, it is important that this monument take the form of a counter-memorial rather than a memorial.

In conclusion, the monument which I planned servers as a way to remember those who regrettably passed during the Woolsey fire, to teach about fire safety, and to inform about the role that SCE played in this disaster. The selected location, Chumash Park in Agoura Hills, CA, acts as a frequented locale that exists within the affected areas of the fire, reminding residents of Agoura Hills and visitors alike the importance of fire safety. Additionally, the design of the monument reminds viewers that there were people with real names that died during this fire, and connections they had with those around them will never be able to be replaced. Furthermore, this monument being a counter-memorial serves to explain the role of SCE in this disaster, as having the monument be a standard memorial would not hold them responsible. This monument is an important way to teach about the dangers of fire, and commemorate those who were lost.
The Salem witch trials had taken place in 1692, only thirty five years earlier.  Identify five details describing the setting that suggest something sinister and supernatural.
The reason that Dritto does not explain anything is to get the reader to want to read more of this passage. Dritto not explaining anything is a hook because it doesn’t give away any plot details, and makes the reader curious as to what will happen next.
Uora-Uora was put on lookout duty because he was well dressed and would look less suspicious than anybody else.
Baby had not eaten a piece of cake in a long while, perhaps even since the war. 
He felt a desire so strong to eat those cakes that he could not focus on the task on hand, and he also feared the inevitable end of this cake feasting extravaganza, whether that be due to getting caught or because of becoming full.
The manner in which they were eating the cake was very animalistic and non-human-like, which is why the author chose to use animal similies for this animal-like feast.
Gluttony
“Eating Cake” represents sexual intercourse. “Licking cream” has strong sexual connotations, and the manner and place in which this occurred (on a bed) gives the impression that something sexual, or sex itself, occured.
We feel sorry for her, and Katie pities her. 
No
He invites them because he is attracted to Katie.
Yes, as he likes Katie more, and therefore would like Martha to be more like her. Furthermore, being with Martha would radically simplify things for Martin.
Martha puts up with Martin because she has a small child, but she cries when Jenny has her first period because she realizes that Jenny isn’t young anymore.
The reason that Dritto does not explain anything is to get the reader to want to read more of this passage. Dritto not explaining anything is a hook because it doesn’t give away any plot details, and makes the reader curious as to what will happen next.
Uora-Uora was put on lookout duty because he was well dressed and would look less suspicious than anybody else.
Baby had not eaten a piece of cake in a long while, perhaps even since the war. 
He felt a desire so strong to eat those cakes that he could not focus on the task on hand, and he also feared the inevitable end of this cake feasting extravaganza, whether that be due to getting caught or because of becoming full.
The manner in which they were eating the cake was very animalistic and non-human-like, which is why the author chose to use animal similies for this animal-like feast.
Gluttony
“Eating Cake” represents sexual intercourse. “Licking cream” has strong sexual connotations, and the manner and place in which this occurred (on a bed) gives the impression that something sexual, or sex itself, occured.
The features on computers and the internet that make for what the world is today are ease of use and speed. Paired with a good operating system, anyone can use a computer, phone, or any computer powered device easily. Having something that is fast means you can get work done quickly and efficiently, without having to wait (although the internet in its early days suffered this fate). These two inventions have changed the way my life is, from entertainment, to how I am writing this assignment.
Ordinary folks were the dreamers, the people who sought change and had brilliant minds and ideas. Isaacson states that the “great connectors” are usually the inventors, and their inventions. A CEO might invent something that makes themselves a lot of money, but a dreamer will invent something that changes the world.
The difference between these two works is how they frame social media as a tool. While Isaacson states that technology will change the world and revolutionize people, Gladwell thinks that since people still had revolutionary ideas and spread them without Twitter, that social media will play little to no role in politics.
The candles, the darkroom, the pit, the pendulum, the robed judges, and the compacting walls.
The pit symbolizes the fear of falling in, the pendulum symbolizes that time is running out for the main character, who has little time to escape his destined fate. Even though the protagonist is still alive when the rats crawl over him, these rats symbolize that the man has already reached a certain end, and time was just ticking for when it would arrive, however, he was eventually saved.
Poe is condemning the overpowered nature of the Spanish inquisition, and how someone's life could be completely upturned just because of their religion. He paints the inquisition to be bad, as he wants to warn people of what would happen if people were discriminated against by religion.
Yes, and I don’t like Disney-esque ‘happy endings.’ This ruined the story for be because it made the message seem less valuable, if all of that happened, just for none of it to matter because he was saved, for some reason.
He is not in a good state of mind and he is slowly devolving into insanity. The entire story shows his rapid demise, and this further proves it.
He made them twins as it would further contrast them; with nothing different between them, any slight difference is that much more noticeable.
The way he acts changes throughout the entire story, and this is proof enlight that he is going crazy.
I think that he is hallucinating, and his insanity is slowly killing him. There is no way he should be seeing her, but he does because his mind is deteriorating.
The title “Some Dreamers of the Golden Dream” embodies Lucille’s struggle for happiness. This struggle occurs because of her distance from her husband, and the pain and shame she dealt with for being a cheater, and a murderer. While other people lived happy lives: “The Golden Dream,” she did not.
I think that hearing Arthwell’s story puts perspective on what it is like to be in a contributor's shoes. His story shows that he got a lot of shame, and guilt, and although many people think he didn’t suffer, I know he did.
Lucille’s “Golden Dream” was to live a happy life with a man she loved. This never happened, which caused this drama.
Writing a non-fiction story often has less bias, as one does not slather their opinion throughout the work. This changed how people read this story, because they’re opinions were not swayed by the author.
He offers to support his judgment of considering his efforts a failure because he realizes that his efforts were not successful, and he should not pretend that they are. He uses the phrase “self-annihilation” to express the process of creating a mistake that makes you worse, and everybody around you worse. The significance of the quote is that it represents who he is.
Tizon looks up to his father, as any son does. This is evident throughout the essay, and his choice to mainly include his fathers examples does a good job at explaining this, and Tizon himself struggled too if he was close enough to his father to have this many stories. This also works to show the difficulties of being an immigrant (in a figurative and/or literal sense), and not fitting in.
Whiteness affects both Authors by making their culture feel different from a perceived American ‘Normal’. There is, however, no normal lifestyle, as people's lives vary greatly. The approach taken differers from this essay to that essay in that this essay is much more cynical while that essay was much more open.
This provides detail, because it justifies his point, and the way he says it makes one want to read more of this written work. 
He now means that instead of somebody getting labeled on what they say, they get labeled on the platform they say it. I disagree with this, as people criticize whatever the other side is, even if they are using the same platform as themselves.
He states this to emphasize his point that tweeting behind a screen isn’t put yourself in any harm, and is almost cowardly.
He describes “high risk activism” as any form of activism that puts you, your reputation, and your integrity at risk. The only risk you put yourself at when constantly sitting behind your computer 24/7 and tweeting is the risk of a heart attack.
I think this point was to show that the author does think that social media is good, and has its use cases; however, trying to be an activist solely on social media won’t make the cut. I agree with him here.
He makes the distinction in order to establish that being motivated by a cause, and actually participating in it are two separate occurrences altogether. The distinction between the two is that the motivation is something that most people can get behind, while the participation is oftentimes more limited.
His first example is of MLK. I think that his choice of this example is very smart, as many people know about it, and know that he simply preached, he never incited violence, and most importantly, he never twittered. Another example was Ivanna, which shows that social media is powerful, but that you shouldn’t use it for protests.
Activism:
He grants the benefit of helping people in need.I think that saying “digital evangelism” is a little extreme, and it really is just a hub.
He uses modern example to show that people very similar to us didn’t need the internet to make their points go through.
I think they are similar because they both use simple language to understand, and showed how people could be judgemental, and helpful.
The narrator wants to get validation from the couple as a way to make him feel like he is living a better life than he is. He is crippled, and not living a great life, but wants to think he is extremely important in the grand scheme of things.
Helping him will shine a light on the fact that he is crippled and will make him feel worse about his current condition. The wife does not want the illusion that her husband sees to be broken.
Hubert means that he exists only behind the scenes and does not want to be front and center as to upset the man.
The author is shining light on the fact tha
Having no names included makes the reader try to fill in gaps about these people, which engages the reader further. This rhetorical choice is a smart strategy because it ensures that the reader pays attention.
She would do that to try and understand Hubert’s perspective of life, and in turn, felt herself shift into that. This embodies greed.
A White Russian is an alcoholic beverage. Alcohol is addictive, and people can be greedy to obtain alcohol. 
The “gray zone” she speaks of are the Muslims who are genuinely good people, and also happen to be Muslim. These people are not terrorists, but often get clumped together with them, just because of their race. The gray zone not only resists hate and terror, something that the minority of Muslims who are terrorists feast on, but also the notion that they are terrorists.
The group’s religious message is that they believe that the only way to worship Allah is to kill any non-believers and try to scare those who can be changed into preaching Islam the same way they do. This parallels what most Americans think terrorists are: destructive religious extremists. It can be inferred that Lalami is emphasizing the point that a notable chunk of the American population believes that anybody who is Muslim is automatically a terrorist, and Muslims get treated unfairly because of that. Her choosing to put this essay as a personal narrative helps the points get through because this essay appeals to pathos.
Both essays share the fact that they want to ensure that their daughter(s) will know that their life will not be normal, as women who are in a minority. This emphasizes that judging people based on their religion or ethnicity doesn’t affect just the terrorists who happen to be of that faith, it doesn’t just affect adults who have to deal with this unwarranted prejudice, but also children, who will have to grow up like this.
He was feeling extreme anger and harsh resentment. This anger kindled him.
This shows the innocence of a child. This is important to understanding the plot and everything about the story.
Deb, who was once put into this awkward situation, gained someone who trusted her and felt comfortable around her.
The narrator’s anger ended up hurting their relationship with Deb. 
She points out the fact that these people were taken from their land, and their kids were not allowed to go back. This was continued, for an extremely long time. She uses pathos to make readers feel as if they are her and experiencing this racism themselves. These tactics make the reader feel resentment towards racists, and her explanation of the “microaggressions” makes the police, and the average citizen seems unknowing and ignorant to the issues that these people face.
She conveys this message this way to show that there is hope, and another way, ie to move away. The strategy that is employed most is pathos, which is used to tie examples together and make the reader be put in the shoes of those who were forced to deal with these horrific racial incidents.
She makes tragedy seem sad, while he makes it feel bad. This difference in tone makes both pieces have very different overall feels, while this one was one that made you feel bad for the African Americans, his made you want to do something about the issues.
He loves gazing at La Grande Bretèche for he loves the beauty it possesses. This, unfortunately, is frowned upon by everyone else as he has trespassed there. Furthermore, the late owner of the establishment forbade anybody from entering.
He wants to explain the severity of his passion for making sure that nobody enters the manor. This provides the reader with insight as to the nature of this house, and the lore that surrounds it.
She did, as she felt guilty about how the events, or rather, the tragedy that occurred impacted the Spaniard. Although cheating is highly immoral, suffering that fate because of cheating is a bit extreme.
Even though she was implicated in his death, she was never tried nor found guilty.
This story takes place in France, and the allusion to the biggest city in that country can be an easy place to compare from. Using a well-known city with well-known tropes as a tool for comparison is an effective rhetorical device because it gives the reader context.
She is too afraid that her husband will get mad, and call her a sinner for lying under the oath of a cross.
She believes that the household became cursed after her negligent decision-making processes were tested, and wants to ensure that the household is no longer cursed. She probably also wants the rotting smell of a dead body to not be in the house, for obvious reasons.
The word “victim” is used in the opening sentence as Staples is trying to show that occurrences like the one mentioned at the beginning of the story are so common that he has resorted to satirizing them in order to cope with them. A reader might interpret the first sentence as a shock, but after reading further, one will understand why Staples decided to use such wording.
He shifts his point of view when the story is explained from the perspective of the “victim,” to show his point of view and his intentions. His purpose in doing so is to show the reader that he is not doing anything wrong and that he is getting blamed for the color of his skin.
Staples behaves in a mature and safe manner, and although annoyed by the woman’s fear, does not react to it in any way that would escalate the situation. This is similar to Muñoz’s essay, by the fact that people, especially minorities, are just supposed to accept what somebody who is alike in every way except for race and/or gender are going to define the status quo for them, and they will get no say.
I do not think this is true, as, for years, many teachers of mine have tried to get me to enjoy reading. I have started to enjoy it more in my high school years, but I was always inspired to read something at any age. I cannot say for certain if others have the same experience as I do, but I assume that most people are similar to me, as I was never extraordinary, nor was I falling behind.
She uses such strong language to get a reaction out of those who agree with her. This technique, while effective, will turn off non-like-minded individuals from reading what was written. I found that she also puts her opinion front and center, rather than making this an unbiased piece. This, once again, will attract those who agree and detract those who do not agree with her.
Prose means that by limiting the reader’s exposure to other sources in the class, you are stifling their imagination, and that constriction leads to them forming an unnuanced opinion. I tend to disagree, as students aren’t trapped inside of a small room, and can see the outside world, and form their own opinions from that, not just from what is taught in the classroom.
She is implying that everyone in the classroom forms their opinions based on only what is taught in the classroom.
They loathe it because it feels forced and unnecessary. From my personal experience, I find:  
She suggests reading more books and using fewer lesson plans.
The origin of the name is based on the main character, who is a dwarf who is often mocked. The title is ironic because he was the one who got the last laugh, as he engulfs the cabinet in flames and murders all of them.
The only name that the main character gets called is “Hop-Frog,” as the king, and everybody else did not take him seriously whatsoever. This reveals that he was often thought of as unimportant, which can be very taxing on somebody mentally, which may cause them to act in such a manner as he did in the end. 
The author, Poe, assumes that the royalty suddenly changes their stance on Hop-Frog from being a joke to being somebody for whom they could ask. 
Poe portrays Trippetta as being short, similar to Hop-Frog. This is significant as it forms a rhetorical connection between Trippetta and Hop-Frog. Furthermore, this connection stretches into her actions, as when the king tries to get Hop-Frog to become extremely drunk, Trippetta protests, showing that, unlike the king, she actually has heart.
Since Trippetta is held in high esteem by the king, one might assume that she is like-minded; however, she ends up going against the king.
The irony in the story is that Hop-Frog, who was oftentimes mocked and harassed by the king, ended up becoming victorious and got revenge on the king for all of his misdoings to him.
The chains and the fire symbolize how the king and his men treated Hop-Frog: with disrespect and fuego.
The phrasing “basal beasts” paints the men out to be relentless, unforgiving, greedy men.
The characters engage in a lot of fantasy. They are not happy with whom they are with, and have lustful fantasies about other people.
Yes
The narrator is a victim of their own doing.
Yes, never having enough is definitely a characteristic of gluttony.
Yes.
Yes.
No.
Hubert and his wife like to live a so-called ‘eloquent’ life, and having each other’s way minimizes discourse that could get in the way of that dream.
No.
No, neither.
Yes.
No, he is irresponsible and a bad parental figure.
Women in society are often seen only as mothers.
If staying together benefits the child more than it hurts the parents, then yes, otherwise, no.
He feels as if Deb is an impostor.
Yes.
Society weighs heavily towards couples staying together, even if it negatively affects somebody or both people in the relationship.
Men, as traditionally, were the ones to ask out the women, and the ones to propose.
Yes.
Balzac tells his story from an outsider’s perspective as that would be the most relevant vantage point for the audience.
These are intermingled in the story through the main character’s strong desire to visit a forbidden place, even if somebody specifically requested not to.
Even though he was no longer creating music, his death made his popularity rise, as the death of such an influential figure made waves in the music community, causing many people to discover his music.
We then moved about a mile away, up a hill and across our street, there is a hiking trail. I live in my house with my brother, mom, dad, and my cute dog, Suki. I went to Willow, from kindergarten to 5th grade, I was at the GATE program at A.E. Wright in 6th grade, and of course, in 7th grade, I attend Lindero. 
I think the turning point in the poem is line 20: “But yet sufficient for us left”
Line 45: “With glory richly furnished”. People today are much more materialistic and less intrinsic as the past, so we would probably take longer to get over the loss while being less affected by it. 
The house is nature, it is made by God (hence the capitalized “Him”), and it is better because to her because it was a gift.
In comparison to what she now has, what she once had is gross and unwanted.
What she wrote was lost. This was her life, this was her. Losing this was like losing everything about you. She had nothing left of what she had written.
She doesn't want to move and does not want to accept defeat.
 He said that he had sold the house and he suggests that everyone should just take the money.
He would essentially be accepting defeat. White and black people should have the same opportunities, and by giving up hope, he is pushing that away.
Walter tells Linder that he isn’t going to move, and Linder appeals to mama in an attempt to reason with her. She gives the same response that Walter had given to Linder.
The importance is that everything is coming full circle and she has accepted that it is time to move on and live life.
His arrival changes the mood because it prompts Bertha to remove her headdress, revealing a nearly bald head, and all of those who were around her were appalled to see her head in this condition.
Walter looks down upon George and thinks that he is lesser in class and in intelligence compared to himself.
She has put a down payment on the house.
Ruth rejoices while Walter feels despair and anger.
George wants to marry somebody who is simple yet intelligent, and he refuses to listen to her ideas because he thinks that he is more important than her.
She helps end the argument and stands behind her daughter.
She gets called, and he was wandering into clubs instead of working.
He envisions a future where his family lives in a utopia with only happiness and smiles to go around.
They aren’t white and the neighborhood around them does not appreciate their presence.
They are black.
He tells him no and tells him to leave.
Gardening tools, as she was at the time attending to her plant.
His money had been stolen, and he was angry.
The apartment itself is very small, unfurnished, and old. These lead to the living conditions for the Younger family to be sub par at best, and horrid at worst.
Ruth is reluctant to give money because she does not see the value of spending money unless it is absolutely necessary, while Walter gives the dollar because he understands the importance of this to Travis.
Walter thinks that women should employ roles in society traditionally held by women, and should act “womanly.”
Bertha wants to study medicine, which goes against Walters views on a woman’s role in society. 
She takes her plant, which symbolizes her nature as one who cares and doesn’t take, unlike everyone else.
Ruth returns from the doctor, for whom had told her she was pregnant. 
He brings traditional Nigerian clothing and music as gifts for her.
He made Ruth so mad that she locked herself in the bathroom, and he carelessly spends their money.
Mama influenced him by telling him the truth, that Ruth is pregnant.
Fair. He was the only person to vote innocent right off the bat, wanting to give the defendant a fair chance.
Sympathetic. He feels for the defendant and understands that he had a rough life and sometimes a second chance will make you better.
Smart. He knew exactly what to say in order to try to sway some of the other jurors.
Impation. He is grumpy about how long the trial is, even before it starts.
Mean. He was abusive, even to his own son.
Non-sympathetic. He says that he would be happy to perform the execution.

"""
```

```python
texta = """
Hello, my name is Jacob Rosen, and I just want to say that it is very crazy, that it is the time to culminate out of 5th grade.
It seems like on the first day of kindergarten, I started with an empty tool belt.
Every lesson, grade, thing I learned, and test was like getting a new tool in my belt.
I have so many tools in my belt, I can build a bridge.
That bridge will take me to middle school, and with all of these tools, I’m very prepared.
Willow has taught me a lot, and I have been through kindergarten, 5th grade, and all of the grades in between.
The bridge I am going to build is very large, but Willow got me tools to go on the bridge.
It all started when I walked into the classroom in kindergarten.
So much has changed from that day, academic and not.
I barely knew anything, I didn’t have braces or glasses, and I wasn’t that mature.
I was very lucky, I came to Willow the last year there was the kinder campout, the last year of reciting the Finding Dracula poem too.
But we are talking about kindergarten, not how lucky I was.
The kindergarten campout taught me a lot, like how to be brave in circumstances that you have never experienced.
With that in my toolkit, I will use that to actually be brave enough to build that bridge, and cross the gap to middle school.
To start a new experience, when I went to 1st grade, I got to go to the big yard, with the 2nd graders and giant 3rd graders, which taught me about kids that were older than I.
What gave me a better tool was Turkey-Tea.
Turkey-Tea is when the 1st graders go up on stage near the time of Thanksgiving, and sing some songs.
That was one of the first times I went on stage.
When you are that age, it takes a lot of guts to do that.
I guess that got me a tool of perseverance, which will let me keep building the bridge, even when it is embarrassing, or not going well.
Incredibly, 2nd grade taught me a big lesson.
Fersht’s class, and we were reading a book called Strega Nona.
Now I don't remember much about the book, but what I do remember is there was a character named Big Anthony.
One thing I remember about Big Anthony is he went to this magic pot, said a spell to make pasta, said the spell to stop it, but what he did not know is that he had to give the pot 3 kisses to stop.
So eventually, Strega Nona came back and stopped the pasta from coming out.
But Big Anthony had to eat the somewhat 5 tons of spaghetti, and you could guess what happened.
So the tool is to follow directions, and don't be a Big Anthony, I repeat Don't Be A Big Anthony!
Next up, there is 3rd grade.
3rd grade has a big event, Ellis Island.
Ellis Island is when you go to the M.P.R. and you dress up and pretend you are a different person.
The person that you are is an immigrant from another country.
You have to get through security, not be sick, and you can buy stuff.
You can also write letters to your family back in your old country.
That gave me an important tool, to be thankful of where I live, and that I am lucky to be born into the family that I have, and not to live in another person's shoes.
Now we will talk about 4th grade.
Probably the biggest thing you do in 4th grade is the 4th Grade Fiesta.
In the 4th Grade Fiesta, you sing some songs, dance, and have Mexican food(Yummy Yummy)!
Probably most 4th graders agree that they did not like the dancing.
I in fact did not too.
The tool I earned from that is that things that you don’t like, endure through them and you will have rewards(Ex. The end of the line is 5th grade).
Probably most kids would talk about Outdoor Ed, but I don't get what tool you would get from that.
Surviving in the wild?
Learn about the outdoors?
I will talk about the State Report.
In this report, you have to learn about a state, and write down interesting facts about the state on a trifold poster, and memorize a speech you made.
It is hard work to glue everything on your trifold, get the facts, make it look good, and make and memorize a 3 minute speech.
But if you work a lot on it, it will look spectacular.
That is a good tool.
Get done with everything first, and then have time to do what you want.
Thank you, siblings and parents and aunts and uncles and grandparents or whatever relationship to your culminating 5th graders, to come out, change your schedules, work around your everyday life, and thank you teachers for helping us get tools to build a bridge to middle school, and I hope you enjoy the rest of your day.
5th graders, yay!
We’re culminating!
The bear who lived in the forest heard a loud strange call.
Who could it be he thought to himself.
He realized it was the Jade Emperor.
To all animals in the forest.
The year has shifted in an unknown way, and the Zodiac is ruined.
We will have to do it again, or else there will be no time.
We will have a new animal, the Bear.
The bear was celebrating that he was chosen.
He was the special one.
The next day, the bear got to the start line.
The ice on the ground was shining like the stars.
You could feel the cold breeze.
You could see the bear zoom of at the start.
The race had started.
Just then, the horse caught up with the bear, and the bear attacked the horse.
Nobody passes me, the bear howled to the fallen horse.
The bear was as vicious as a black widow.
yelled the horse.
Back a bit, the ox saw the fallen horse, and helped him up.
He asked who did it, and he said the bear.
The ox sprinted up to the newly accounted zodiac, and said, Not cool.
I am stopping you.
The bear rammed into him, and that’s when everything changed.
Why did you bite me?
If that’s what you want to do, here is a taste of your medicine,The bear bit into the ox’s leg, and blood came dripping out like water from a river, and the bear's teeth, like razors, pierced the poor ox.
As the raven once quoted, the bear said, Nevermore.
You will now not pass me.
As the bear gracefully swam through the tough river, and sprinted to the finish line, a flash busted in front of him.
Blending in with his brown fur, the rat jumped off of the bears back, and beat him with the trick with the ox that was now bleeding, thousands of years ago.
With all of the viciousness of the bear, and the insanity of the race, the bear got tricked at the end,.
He was announced as 2nd place.
To fit in, wear the clothes that the other girls are wearing.
Hang out with girls, and go over and play hopscotch.
You should not hang out with the boys and run.
Native Americans use natural resources in their everyday lives.
My chosen tribe was the Apache, who live in the Southwest USA.
Today, the Apache would be in Arizona and New Mexico, which is very warm.
In this report, you will learn about the Apache’s, traditions, food, tools, clothing, shelter, and 3 facts.
Natural resources are key for Apache food.
Food is important for survival and they get it provided by hunting and gathering.
One way the Apache get food is hunting.
Buffalo is something they hunt.
People hunt other things though.
Also, Apache gather acorns and berries, which are a staple.
They can cook in their wikiups and in their tepees.
The Apache made great use of their resources food.
The Apache built their homes with natural materials.
One type of home was a wikiup.
A wikiup was also called a wigwam.
The wikiup is shaped like a dome.
Another type of home was a tepee.
A tepee was a temporary building.
Also, a tepee was lightweight.
The Apache were nomads, which means no home, and nomads usually were in search of food.
That is how they used natural resources to make homes, were nomads, and what their homes were like.
The Apache used natural resources to make clothing and make artifacts.
Apache clothes were sometimes leather.
The shoes Apache wore are called moccasins.
Tools were helpful for Apache life.
One tool they used was a travois, a tool used by nomads.
They also used horses, which were important because they helped them get around.
The weapons they used were arrowheads and a bow and arrow.
These things were important to everyday life, hunting, and transportation.
My tribe had some interesting customs, traditions, and artifacts.
One tradition they had is to have roasted agave.
Agave comes from cacti.
Agave was a food they enjoyed very much.
Also, the Apache women stayed at home.
They made artifacts.
One of these artifacts are sewn baskets.
The men were warriors.
Lastly, the extended family stayed together.
These facts make the Apache fun to study.
I am about to tell you 3 interesting facts about the Apache.
When an Apache woman marry a man, the man joins the woman’s family.
Also, the Apache word for corn is maize.
Finally, a famous Apache was Geronimo.
The Apache had unique customs and facts about them.
All Native American tribes use the natural resources that they have in their everyday lives.
The Apache natural resources were used in a lot of things, the Apache have an interesting culture, and have interesting lives.
There are a lot of things to learn about the Apache.
They are a special tribe with great characteristics.
You hear the bell ringing.
You just got out of school.
It is a cold winter day, and you hear the pitter patter of the rain outside.
When you get home and you want to play your new FIFA game, and have hot chocolate.
You get on the couch, and right before the Xbox turns on, you realize you have two hours of homework.
You just were learning a subject for an hour in school, and you will remember it.
How would you forget it?
Homework takes free time that could be family time out of people's lives, wastes free time, and reduces sleep hours, because kids are up all night doing homework.
Lastly, homework can raise stress levels on kids.
Homework should be banned.
It seems illogical that someone was in school and they learned something for an hour, and that they expect that they forgot everything, and they give them homework to remember everything.
According to a website, apa.org, It barely affects them.
Their scores barely improve with or without homework, so it shows that homework doesn't help (APA).
What this means is that someone is wasting time on something that might get them a 96% on a test, and doing no homework would get them a 95.
It's not that it’s 5 minutes of homework, it is a great amount of homework.
A PBS writer stated, When asked how many minutes of nightly homework students receive, the average was 81 minutes.
Yes, 1 hour and 21 minutes.
John Buell from the Huffington Post explained that some research has actually proven that homework makes your grades worse than without homework (Buell).
One hour and Twenty one minutes is a massive amount of homework.
Somebody could be hanging out with your family, but it takes massively too long.
If you had the same amount of homework for all 180 days of school, it would be 180 x 81 or 14580 min or 243 hours.
243 hours could mean a tremendous amount of time with one's family.
In 2017, everybody is so busy that our only free time is taken up by homework.
It is a massive amount of work to not make major a change.
An hour might get a person one more question correct.
Homework wastes our time, not just personal, but family time too.
Scholastic wrote about freetime and homework.
Something they illustrated was, Opponents of excessive homework argue that it forces parents to sacrifice already scarce family time for the sake of completing assignments, not because they are necessary for students to grasp concepts but because the schools have decided in advance that children must do something for each subject every night whether it's necessary or not (Scholastic).
What that means is that free time is snared and grasped by homework.
The time is scarce, and the scarce time is being spent on homework.
A New York Times writer, Kristen Weir, wrote about how homework uses up a lot of time.
She also pointed out how there has been more homework, and it negatively affects somebody's grades (Weir).
Peter Druker, a Huffington Post writer, wrote about how homework is a waste of time, and a student could could be doing better things with their time than doing homework.
Homework is something that wastes your time, and sometimes negatively affects your grades, and usually does not affect your grades at all.
Most people know how stressful homework is.
Studies have shown that having homework is more stressing than not having homework.
John Buell, a writer from a source that gets checked a lot, Alternet, wrote that if homework was a prescription drug, FDA would have banned it a long time ago.
It is so unhealthy for you that if it was prescribed, you would get unprescribed, because is is bad for you(John Buell, Alternet).
The Huffington Post remarked, If bombarded with countless lessons at school and at home, students may feel stress and anxiety should they fail to complete the assignment on time.
If someone get a lot of homework, they are stressed.
It is because teachers give too much homework.
Some may say that homework helps you with school, teaches you time management, and helps you be independent(e-skool), but many studies have shown that homework negatively affects you.
Some think that Homework is great, but the evidence shows that that is not true.
There are few positives and a tremendous amount of negatives with homework.
It takes off free time.
An unknown author called Memphis Parent stated,  Students complete their assignments or projects on time.
Well they might learn time management, but they don't have enough time.
The people who are saying that homework is amazing are companies like Common Core and the company that makes CAASPP, SBAC, Big Ideas Math, and GO Math.
It is like saying that fast food is incredible for you if you are the C.E.O. (Chief Executive Officer) of McDonalds®.
You just want your product to make money.
Evidence backs up that homework is terrible.
Overall, homework is a useless.
It wastes free time, it is not necessary, and it stresses kids a large amount.
That is why everybody should team up and end homework.
It is just a torture device, something that is unnecessary and makes people feel uncomfortable.
Without homework, the world would be a better place.
That is why we should not have homework.
Greetings,Hello, nice to meet you.
My name is Walle.
I am from Timbuktu, and I am 49 years old.
If you don’t know what Timbuktu is, Timbuktu is the great city that is the capital of Mansa Musa’s Mali Kingdom.
As my job, I am a brewer, I make beer, I create something for people to drink.
It is very enjoyable.
I taught myself how to be a brewer after I retired from being a ferba because it sounded like more fun, and I am glad I decided to make the switch.
I live in the brewery, and as a home, it is pretty big.
When I am not doing my job, I love to cook.
The food I eat consists mostly of salted meat and vegetables.
I also pray to Allah, as I am Islamic.
Unlike other civilizations, men and women have basically the same rights.
Our Mansa, or ruler has complete power.
His name is Mansa Musa, and he is the most wealthy ruler currently.
The Hajj is an important event that affected Mali, as Mansa Musa goes on it.
I went on Mansa Musa’s Hajj.
I met Mansa Musa because I once was a government official, and went on the hajj for that reason, but about 15 years ago, I decided to become a brewer.
He treated me well besides the time when I overslept, and made a delay.
He got very mad at me, and when Mansa Musa gets mad, he gets very mad.
Thanks for reading this letter.
I hope you learned something about my culture.
Hope to see you at the Faire.
Sincerely, Walle.
In the novel Counting by 7s, the character who changed the most is Dell Duke, the counselor that does not do what he is supposed to do.
At the beginning of the book, Dell didn’t run because he was not in shape, he didn’t have a desire to run, he was a slob, and he didn’t appreciate the world like he did at the end of the book.
When the book started, the unathletic Dell didn’t run, and at the end, Dell ran, and got in shape.
Dell wasn't in shape when he first met Willow, and got inspired later.
Dell didn’t need to run at the beginning, because he wasn’t changing, and didn’t have a desire to.
Dell also didn’t know how to pace yourself when running.
He probably spirited for 5 seconds, and then got super tired.
When he ran for fourteen minutes, Dell looked like he was about to pass out, and probably didn’t even run a mile.
Next, Dell was a slob, lazy, and only ate microwavable meatloaf that was for a family when the book started.
He didn’t have real furniture, because he could just hose off the plastic garden chairs.
He didn’t see anyone, so he didn’t have to make it formalThat is what Dell thought.
When Pattie Nguyen opened up the closet, they saw the Leaning Tower of Pisa made out of underwear.
I don’t mean that literally, but he had a LOT of unwashed boxers.
He only ate meatloaf, because he didn’t have to eat anything else.
Nobody saw him, so he could eat whatever he wanted.
Lastly, he didn’t appreciate the world.
He just laid around, and didn’t care about other people.
When he met Willow, and her parents died, he thought of the world differently.
He left Cheddar at the beginning of the book.
He saw Cheddar at the end of the book beat up, and took him to his house.
So, Dell is now a much different man.
He is not a slob, because of the Nguyens moving in, and is more healthy, because he doesn't eat the meatloaf.
He also looks at the world differently.
He would still be the slob he was if not for Willow, but he changed more.
I wore a fancy outfit.
I wore that because my parents told me to.
What I wanted to wear was a Neff shirt, and athletic pants.
Why did I want to wear that.
I wanted to get a good impression, and I liked the shirt.
I don’t know why I had to wear the shirt.
My parents said that I should ‘look good’ on the first day of middle school.
But really, what I wanted to wear would of made me Look good.
Willow did not cheat because of a few reasons.
She probably didn’t know that she was going to take a test, and didn’t know of any sources.
If you want to see if she is cheating, or is just a human calculator, give her a random test.
If she got a 100 on the 6th grade test, make another one, with the same style problems, but different problems.
For example, if the problem is 2x/3z√2 3x/2z*4 = 2 + 3x / (2z * 2)^2,change the values.
Then you will see the difference.
If she Passes (100% - 70%), she didn’t cheat.
If she was below 70%, she cheated.
Describe a topic, idea, or concept you find so engaging that it makes you lose all track of time.
Why does it captivate you?
What or who do you turn to when you want to learn more?
August 21, 2019, 11:15 a.m.: I stepped into my Spanish 1 class for the first time.
Like my friends, I was not particularly enthusiastic about this new class.
What I would later come to realize is that I really enjoyed not only Spanish, but also learning about how different languages are structured.
In the end, I participated more often in Spanish class than my peers.
I started reading additional material outside of class and watching television shows in Spanish to expand my vocabulary.
Although the Covid-19 pandemic made learning Spanish in person more difficult, I think I have still made great strides during the pandemic, and my passion for learning new languages continues.
I plan on continuing to study Spanish and other languages in university and beyond.
I have even joined forums dedicated to language learning and have had very interesting conversations with those who speak primary languages other than English.
Studying languages inspires me, and I hope to become a polyglot (somebody who speaks many languages fluently).
Localization, or the process of taking content produced in one language, and adapting text, images, references, and audio to better suit it to a different audience would be a possible avenue that I would like to pursue.
Although I do not consider myself proficient enough in another language to start localizing content, I am deeply interested in that, and have read articles and watched interesting videos about the localization process.
What I find most interesting in localization is converting or rewriting puns and jokes for another language; an example being antelopes eating cantaloupes is a pun that would only work in English.
I have spent hundreds of hours reading about linguistics and studying other languages’ grammar, through e-books and articles.
Just a few years ago, I assumed that different languages were just like English but with different words; the more I dove down the linguistic rabbit hole, the more I realized that was simply not the case.
I decided to start learning Finnish because the grammar is notoriously complex, and it has 15 grammatical cases.
While I have not ‘Finnished’ mastering this language as of yet, I am proud of how much I have learned on my own and continue to work on my goal of C2 Finnish fluency.
Although some consider the two year language requirement in high school mundane, I enjoyed language classes and am thrilled to be in Spanish 4 Honors this year.
Linguistics and language learning have helped me connect with people around me, and has sparked a love of learning that goes well beyond my high school Spanish classrooms.
Change is inevitable and a slow process that can take decades to fully transpire; however, under extreme circumstances change can almost be immediate.
The adventure novel 20,000 Leagues Under the Sea, by French author Jules Verne introduces Pierre Aronnax and his limited crew as they are taken hostage.
Pierre is known as a famous explorer and his reputation becomes a point of interest to the elusive Captain Nemo and his submarine of horrors, the Nautilus, as it needs his expertise.
Unfortunately for Pierre Arronax, he, along with his servant and a Canadian harpooner, get cast overboard from a voyage they were traveling on, and find themselves being held hostage on this mysterious submarine.
Although initially intrigued, his captivation shifts to horror as he learns that this ship will be where he lives for the rest of his life, as the ship’s captain, Captain Nemo, forbids the trio to leave the ship, as to maintain the secrecy of the ship.
Arronax initially starts out as an adventurer, eager to explore the world, and ends the story wanting the opposite.
The various forms of trauma which Arronax endures is the agent of change which turns his enthusiasm for adventure into hatred towards the Nautilus and his eventual escape.
Arronax enthusiastically starts this story with a bright future ahead of him.
He is invited on a ship known as the Abraham Lincoln, on an expedition to hunt down a mysterious sea monster.
He, along with his servant, Conseil, befriends Canadian harpooner Ned Land, who will use his expertise to help bring this mystifying creature to its end.
Everybody on board believed that this monster truly was a monster, as even the captain stated that the creature is evidently a gigantic narwhal (28).
This plan, however, goes awry when Land’s harpoon cannot penetrate through the creature's steel-like skin, and after the ship gets attacked by the monster, the trio has to jump overboard to save their lives.
This would have been the end of their lives if not for the ‘monster,’ as what they landed on was actually a submarine.
At this point, Arronax still is enthusiastic about adventuring, as when he comes face to face with Captain Nemo, he eagerly tries to explain the trio’s situation to Nemo in multiple languages in hopes that he would allow them on board.
He even tried to narrate [their] adventures in Latin in hopes of Nemo’s understanding (41).
When Nemo eventually responds, the three are shocked to learn that they would travel with the crew, but would never be let off the submarine.
Although this is initially off-putting to Arronax, his inner explorer ends up engrossed in all of the intricacies on the ship, such as the organ, the advanced equipment, and the library which contains books in many of the world’s languages.
As he and his companions are still explorers at heart, this new environment is extremely exciting for the three, and at this point in the novel, they still have high hopes for their futures.
Repeated exposure to traumatizing events will cause even the strongest willed people’s perspectives to shift, and Aronnax is no exception.
Captain Nemo decides to adventure to where no man had ever gone before, the South Pole.
When he reaches the South Pole, he plants his flag, and turns back, only then to come face-to-face with a massive iceberg.
They try to venture through the iceberg, but the iceberg shifts and they become trapped.
As such, they have limited food, water, and air, and need to use it all to attempt an escape, and they narrowly escape certain death.
The stress and lack of fresh air causes Aronnax to be in immense pain, as he states that yawning dislocate[s] [his] jaws (371).
The immense pain and stress caused by almost encountering death itself fabricates a sense of wanting to escape the Nautilus, not only for Arronax, but also for Conseil, and Ned Land.
This is not an isolated incident though, there are many similar situations that have made Aronnax want to never embark on any further adventures with Nemo, and return home immediately.
Stress builds up, and there are still events to come that are far worse than this; however, the incident with the iceberg still has caused Aronnax’s perspective on travel to shift.
Even after basking in the glory of fresh air, he still is shaken from this occurrence, and trauma takes years to recover from, years that Aronnax soon will not be spending on the Nautilus.
If he did, however, the Nautilus is not a safe space for Aronnax, and would likely cause his trauma to increase.
Another example of an extremely traumatizing situation that Aronnax endures is the fight with the squid.
The nautilus encounters a large squid that would be impossible to kill simply by ramming the Nautilus into its flesh, so everybody on the ship had to go to the deck in order to fight it.
Their weapons seemed to do nothing against the squid but anger it further.
The squid decides to pick up one of the crewmates, and drags him off deck to his eventual death.
His blood curdling screams ring through the air, further scarring Aronnax mentally.
Occurrences like escaping from the South Pole and the giant squid are creating a wind of despair within Aronnax to blow, which will eventually blow out the flame that is his ambition to explore.
After experiencing trauma, change in one's opinions will transpire almost immediately.
After each brush with death, Aronnax becomes further displeased with being on the Nautilus and becomes more eager to escape the ship.
When he encounters one final near-death experience, it finally motivates him to escape the Nautilus for good, ending his adventures.
This ultimate event is a maelstrom, which is an extremely powerful whirlpool.
Aronnax, his companions, the ship’s crew, and Nemo all become extremely frightened when they realize the dire situation they are in.
This extreme stress that the crew is under is too much for Nemo to handle; he goes to his organ, plays haunting chords, and shouts Almighty God!
This situation allows Land, Conseil, and Aronnax to make a run for the dinghy, and potentially be free men once again.
The stress that Nemo had been feeling was also affecting all of them.
Escaping on a small dinghy during a violent whirlpool is a rather senseless thing for them to do; however, their plan ended up being successful, as they landed on an island off the shores of Norway.
These men were no longer adventurers, that time had passed.
They risked their lives to get away from adventure, as the immense stress and trauma had caused all of them to change forever.
Impactful events have caused, cause, and will continue to cause change in people rapidly.
Arronax, who was originally an ambitious explorer, put his and his companions' lives at risk to escape from exploration.
He upended everything he had, because changes in his life dictated that he no longer feel interested by exploration, and as a matter of fact, he felt repulsed by this very facet of his previous life.
This theme can also be reflected in the real world, for the better or the worse (think about how 9/11 both helped increase airline security, but also created widespread xenophobia against Muslims).
His journey can also be put into the context of mankind itself; as major changes, like industrialization, evolve us as humans at a rapid pace, not biologically, but informationally.
If humans did not have this generational evolution, our modernized world would not make sense to anybody, as society would have never changed to suit anything beyond the neolithic era; however, we are able to undergo this rapid change, whether as an individual, or as a whole society.
These major events have shaped our lives in the past, and will continue to do so in the future.
Luke Skywalker accepts personal responsibility for R2D2 after R2D2 ran away to try and find Obi-Wan Kanobe.
He realises he will get in a lot of trouble if he is irresponsible and doesn't find R2D2, so he accepts the task of finding him.
He also accepts personal responsibility when his remaining family dies, and he has to become a Jedi, and this shows that he accepts personal responsibility, as he knew that he had to do this to help the rebel army, as that was his responsibility.
One last time he accepted personal responsibility was when he helped the princess escape.
He realised it was his duty and responsibility to save her, as if he didn’t help her escape, she would have been killed by the Empire.
Luke Skywalker discover self motivation by going back to Obi-Wan and wanting to become a Jedi, even after storm troopers destroyed his home, and killed his aunt and uncle.
He could have given up, which he might have done saying everything he had was gone and the last of his family (that he knew of) was dead, but instead he went back to Obi-Wan, and ended up doing great things, because he was motivated to avenge his aunt and uncle.
Han Solo is a character who is very self managed.
He knows what to do and when.
We know that because he knows what to do in almost every situation, and multiple times has gotten out of trouble because of how well managed he is, like with his encounter with Greedo, and when they get sucked into the Death Star.
A time when people employed interdependence was when Luke, Han, and Chewbacka were able to get into the main control room.
They worked together as a team, Luke and Han pretending to be stormtroopers, and Chewbacka attacking everybody.
Without each other, there is no way that they would have been able to get into the main control room.
Another time they employ interdependence is when the same group, Chewy, Luke, and Han, rescue the princess by pretending that Chewy is their prisoner, and that they are taking him to a cell.
That cell happens to be the princesses cell, and they help her escape.
Han Solo developed emotional intelligence when he wasn’t even intimidated by Greedo pointing a gun at him and asking for money that he didn’t have.
He learned how to control all of his emotions, and isn’t affected by stress at all.
He also doesn’t freak out when the stormtroopers attack him.
Instead, he gets on to the Millenium Falcon, and tells Chewbacka to fly the ship out of Mos Eisley.
Luke Skywalker believed in himself when Obi-Wan put a helmet over his head, and he had to fend off attacks from a blaster ball with a lightsaber.
If he doubted himself at all, he would have gotten hit with the attacks.
But he trusted himself and the force, and was able to fend for himself.
After R2D2 and C3PO get out of the escape pod, C3PO doesn’t want to go with R2D2 in the direction R2D2 wants to go, and says he is making the wrong choice.
Later, C3PO ends up on a less than optimal path, and he blames R2D2, saying that he was trying to trick him, when really, C3PO was the one who made a mistake and not listen to R2D2.
Ever wonder about black holes?
Of course you have!
They’re the most popular topic in space for non-astronomists!
The fact that they are just a theory makes the common person wonder even more.
Well, there’s now an even bigger possibility that black holes don’t just exist… but they also have an opposite.
White holes are theoretical opposites of black holes.
Like a black hole, if you go near, space time would be warped.
The force is so strong that if you are near the white hole, you would get destroyed.
White holes are formed by the rotation of a black hole.
If a white hole connects to a black hole, you have a wormhole.
Despite just being a theory, the idea of white holes is actually over 50 years old.
A Russian cosmologist named Igor Novikov came up with the idea in 1964, as part of a solution to the Einstein field equations known as the maximally extended version of the Schwarzschild metric describing an eternal black hole with no charge and no rotation (White Holes Wikipedia).
To teach further about the discovery of white holes, as well as white holes in general, Brit Bendavid and Jacob Rosen, the people behind Space Corp., are creating their first documentary on the subject of white holes.
This documentary short is already in pre-production and set for the release date of Friday, April 28, 2017.
Although it is a direct-to-video release, you can still watch it at home.
We are not releasing this in IMAX, but instead are selling copies at Space Corp.
Our address is 29276 Circle Drive, Malibou Lake, California, 91301.
You can also purchase a copy through phone at (278) - 766 - 6691.
The cost of our product is $19.
99 or a three-day rent of $4.
represents two key questions that are addressed throughout the book, who the ghost is, and who killed King HamletThe many encounters with the ghost throughout the first act, and the very end of the first act, are testament to this fact.
What has happened around me since I returned from Wittenberg is quite crazy.
I first got asked to see if there was a ghost (to which I had my doubts).
Turns out, there was a ghost!
I could not believe myself.
After this, the state of Denmark has been getting wilder and wilder.
King Claudius announced to all that he would be getting married to his late brother’s wife’s sister, which caused much turmoil among us Danes.
After that point, poor Hamlet feels much despair, and does not want to continue his life; to him, his life is now futile, however, all of this changes when his temperament suddenly changes, and he proclaims that he is fine right now, but that his actions in the future will seem to be those of someone who has gone mad.
This has been the craziest time of my life!
Highland climate is the climate of 'high' 'land'.
So, this climate is found in high mountain areas.
There are no seasons in Highland climate.
Any seasonal differences would only be felt at low elevations, near the bottom of a mountain.
The reason mountains need their own climate type is because the climate changes as you move up the mountain.
At the base (bottom) of a mountain it might be 80 degrees and sunny, but as you climb the mountain it will get colder and be rainy.
As you keep climbing, it might be snowy and freezing cold.
In fact, the temperature drops about 3 degrees every 1000 feet in elevation as you move up a mountain.
So, the temperatures in Highland depends on the elevation.
The amount of precipitation in Highland climate depends on the elevation.
Sometimes the land around the base of a mountain is dry, but snow may cover the top of a mountain.
This happens because high mountains force warm air to rise, where it cools and creates precipitation.
The level of precipitation in a Highland climate depends on the elevation you want to measure.
The type of vegetation that grows in Highland also depends on the elevation.
At the base (bottom) of the mountain the vegetation will be the same as the surrounding climate type.
So, a mountain in the rainforest will have rainforest at the base.
As you move up in elevation it will change to plants that can survive colder weather, eventually there will be no vegetation at very high elevations.
In fact, on high mountains you can see what is called a tree line or timberline.
A tree line is a clear elevation line that divides where trees are able to survive from areas that are too harsh for them to live.
Most mountains have different tree lines depending on temperature, soil, and moisture.
Again, the types of animals will depend on the elevation and the surrounding area.
Typically, different types of goats and sheep are common in Highland climate.
Mountain Goats are common in the Rocky Mountains--their feet are adapted to walking and jumping on steep cliffs.
Large cats such as Mountain Lions and Snow Leopards are also found here.
Snow Leopards, found around the Himalayas, have large feet and big thick tails to give them better balance in mountain terrain.
Highland climates are cool to cold, found in mountains and high plateaus.
Climates change rapidly on mountains, becoming colder the higher the altitude gets.
The climate of a highland area is closely related to the climate of the surrounding biome.
Highland climates are cool to cold, varing by altitude.
Usually cooler and wetter than nearby lowlands.
Recycling, in concept, is a greatly beneficial concept, as society can properly reuse disposed goods and turn them into new products, and for a time, that is what society as a collective accomplishments.
Nowadays; however, this practice of recycling is no longer widely practiced, as China no longer accepts trash from the United States because of health concerns and fewer economic incentives.
Historically, China had ships bound from the United States to their harbors coming home empty, which wastes fuel and money.
This waste changed when China decided to buy the United States’ trash, to recycle it into new plastics.
When this practice first began there were many economic advantages to recycling the United States’ garbage, but as this practice began to evolve China no longer needed this excess plastic, as it was causing more problems than it was fixing.
The end of the American-Chinese garbage exchange is unfortunate, as not only does it cause more trash to enter the landfills, but it also causes more pollution in the world, and helps fuel oil miners to destroy the crust further.
If a domestic recycling solution were to have been implemented instead of relying on a foreign nation, this problem would have been ameliorated; however, as this course of action was not taken, our dependence on China caused what would have been the inevitable future of green plastic to plummet overnight.
The future is bleak, but if society can conjure up a solution to stop wasting useful, non-renewable, used materials, society can prosper.
The United States should continue to implement larger and larger tax cuts to those who recycle in order to incentivize people to make plastic more ‘green’, rather than purchased, filthy, mined plastic.
Change is inevitable, it happens to everything.
It cannot be avoided, but what can be is how something changes and why it changes.
Jem changes his attitude throughout the book, starting as a rebellious young boy, to a mature young man.
He starts out young and immature, thinking that he shouldn’t listen to any rules, and create his own.
He didn’t respect his father, Atticus, and those around him, doing whatever pleases his and makes him happy.
For example, Dill dares Jem to touch the Radley house.
Instead of doing the mature thing, Lee wrote that Jem, threw open the gate and sped to the side of the house, slapped it with his palm and ran back past us, not waiting to see if his foray was successful.
Even when Jem had previously asked people about Boo, and they told him to not disturb him, he still thought that he knew better, and didn’t do the right thing.
He could’ve done the correct thing, but instead, he tried to impress Dill, showing who his true character was.
His rebellious attitude could’ve ended up going wrong, luckily though, it didn’t.
Jem just isn’t the most mature he could be, and it shows.
However, he does change, for the better, throughout the rest of the story.
Change does not just happen, as something needs to cause it.
Jem isn’t static; for he changes with the environment around him.
He starts out not knowing what is right and what is wrong.
He changes though, standing up for what is right.
He does this when Mrs.
Dubose is being toxic about Atticus’s case.
Jem shows her that she is a terrible racist when he cuts her hedges.
This is because she said that Atticus was, No better than the [n-word]s and trash he works for (Lee, 135).
Was this a reactive decision?
But it is the motives that drove him that shows that he is changing.
He believes in his father, and instead of disobeying him, he stands up for him.
This change in attitude is no accident; it is him adapting to changes in his environment, as his father's stress about the upcoming case is reflected in Jem’s attitude.
He changes how he acts because he moves with the tide.
He gets very affected by his environment.
There are many inventions that were made in Ancient Greece, but there are few that we still use.
Some of the ones we use are, democracy, coins, and boats.
Democracy back in the Greek days was run by the assembly.
The assembly voted on laws, and voted on a ruler.
government, we vote on either a Democratic or Republican candidate.
There are three branches of the government, the Legislative, Judicial, and Supreme.
The president and vice president are in the supreme court.
The legislative has the Congress, which has the House of representatives, and the senate.
The senate writes the laws.
The judiciary has the supreme court.
Another invention is coins.
In the olden days of Greece, there were three main types of currency: Talent, Mina, and Drachma.
We now have pennies, nickels, dimes, quarter, half dollar, and dollar coin, as well as paper dollar bills, 1, 5, 10, 20, 50, 100, 1000, and 10000 dollar bill.
There is now inflation, and different countries have different currency.
Last of all, we have boats.
The old boats were not simple, but they were much simpler than they are now.
When now boats are run by steam, pistons, or motors, back then they were had a sail, and they were run by oars.
Olden boats were usually made of wood, and were hand made, when boats now are made from metal, an extremely strong plastic, or an alloy of many strong materials.
Greek boats did not have a large capacity, but now boats(mostly cruise ships) don’t just have a extremely large capacity, they can have things like theatres, pools, restaurants, gift shops, and even suites.
Gravity can be pulling an object even when it is not falling.
The Force of gravity of an object is different when the object is in different places in the solar system?
What determines the strength of gravity?
Mass: The amount of matter within a object.
(takes up space)Weight: The force of gravity acting on your mass.
(scale is squeezed)He makes the connection between the audience by introducing himself as an African American student.
This establishes ethos because it shows the reader that this is somebody who has experienced prejudice firsthand.
He uses pathos, logos, and ethos throughout, one example of pathos is giving anecdotes, and another example would be his use of allusions.
He changes the pronouns he uses to create a sense of disconnection, something that an African American student faces often.
His use of the n-slur makes this part of the piece emphasized, although some formality, unfortunately, is lost.
Manhattan, New York City, New York State.
Felix and Antonio, the Amigo Brothers, were walking with each other, and were talking about the fight that they were going to fight against each other.
They decided to split up to balance their mind sets, and make the fight fair.
They didn’t want to fight like they were friends.
Felix decided to pass the time by training, and he also watched a boxing movie.
He tried to not think about Antonio.
Antonio passed the time by training, and also went to his Aunt’s house in Bronx.
While he made the journey, there were some gang members he avoided to prevent conflict.
It was time for the fight.
They saw each other when they dressed and waved to each other.
When they went into the ring, the crowd was roaring and cheering with an excited atmosphere.
They fought hard, but they were fighting with care.
They fought not hard enough, and fought with care.
That changed in round two.
They fought with viciousness, like two starving bears fighting for food.
But they fought each other way too hard.
The round ended, and they felt as if they had just climbed a mountain with somebody punching them while they climbed.
They were bleeding, and when they went over to their coaches, they said to not fight as hard.
The last two rounds tied, so this was sudden death.
This was the moment that would change them.
They fought out their hearts, and fought with all of their energy(that was remaining).
The round was over and the fight was too.
This was the climax, with everything rising up until now.
The announcer was attempting to talk in multiple dialects.
Ladies and gentlemen, Senoras y senores The announcer was announcing who the winner was.
And that is how the story ends.
That was the end of a development, and the end of the story.
The Amigo Brothers had left the ring, and would never know the winner.
Charlie wants to become a true father to his daughter again.
He leaves in order to spend time with her and to try to convince Honoria that he can be a sufficient father and should be given back.
He used to have a toxic relationship with his daughter and left her, and his wife, because his wife kissed another man at a party.
He means that at every party that he and Helen got more distant from each other, and that the constant partying was pulling him away from her.
Marion resents Charlie and therefore puts all of her negative feelings and anger on him.
He likes to believe that he is a good person, but what he did when he abandoned Helen obviously has taken a toll on him, and he regrets his previous actions.
She becomes sickened because of what she saw.
Charlie does not want to think that he is a weak, lazy, and bad man, and paints a picture of Helen in his mind that is unrealistic.
He does this because he needs to find comfort in an otherwise extremely bad situation.
The difference between a company doing good and a company doing well is that the company doing good is contributing positively to society and making, or at least trying to make a positive impact while the company doing well is doing whatever it must in order to make a large profit and nothing more.
Mattel diversifying Barbie dolls is Mattel doing good by making sure that anyone can find a Barbie doll that looks like them, and promotes a broader sense of what it means to be fashionable rather than the generic blonde Barbie doll.
This topic is more suitable to the daily rather than the monthly as although it is a feel-good piece, not much change will occur from this.
Yes, many people will be able to buy Barbies that more accurately reflect themselves, but this will not make that big of a difference in the grand scheme of things.
The two authors’ points of view are similar in the fact that they both despise how a cookie-cutter model is cast among so many young people in an unhealthy way that promotes unhealthy bodily standards.
The value that these toys bring is outweighed by the harm they may cause, and luckily, the diversity helps with this by making Barbies more fit for everyone.
He uses a serious tone, as this subject is no laughing matter.
The voice he uses in each passage is one that has undergone or is undergoing a considerable amount of pain.
He could have used an omniscient third person point of view, which would give the same plausibility, but also could put one in the shoes of the waterboarders too, rather than just the waterboarded.
Hitchen means that the interrogators, although questioning this man and torturing him, still had sympathy for what he was undergoing.
The effect of putting this towards the end of the essay changes one's view of the interrogators, from being soulless tormentors to being real people who were trying to get an answer, but could see the person on the other side.
Although this was just a simulation, people went through this unsimulated, and probably got sympathy from the interrogators.
He looks like he is screaming out in pain, and cannot escape.
This angle shows his face and body foremost, although you cannot see his legs, which are probably flailing out of pain.
The difference is that Hachiya uses tone to convey grimace, while Hitchen uses tone to convey pain and suffering.
They do both; however, use serious language, and go into vast quantities of details.
Hachiyas essay makes me sad, while this essay makes me feel angry.
What he gains from this is that he makes it clear that everywhere on Earth, there is racial discrimination, but what he risks is people not taking his work seriously because of a rather simple analogy.
His use of figurative language, in my opinion, makes the text more understandable and vivid and makes the reader more engaged.
By addressing his son, you realize that this is a problem that is multigenerational, has lasted a long time, and may not end anytime soon, unfortunately.
By mentioning those two who got shot, it puts perspective into the average day of an African American, and the reader contrasts their lives with that of an African American and begins to understand, and maybe even feel passionate about, the issues of African Americans.
The issue that Coates gets by assuming that schools are always against him is that there are definitely schools that are supportive of African American students.
The educational system is unfair, but it is to everybody, not just one specific group, as it was designed to make students break.
When the author states that The West isn’t a place that gives up easily, she means that the west is not a place where people often sit back, and people do stand up for what they believe in.
The narrator describes Homer’s job to show that people who moved to the west had an affinity towards nature.
She describes the way that Homer and her make love in-depth and with a lot of detail.
This is to make the reader understand that she and Homer did not have a very close relationship and a strong connection to one another and that she loved others.
He was the man she truly loved.
She never cared for Homer, she only cared for his body, but she did love Monte.
Homer was soft and would not hunt animals, and cared for nature.
The narrator views this as a weakness.
She realizes that Homer is who she really loved and that she shouldn’t find love in people’s body.
She isn’t happy with the outcome of the events that unfolded.
Heidegger, I would believe that this water causes more harm than good to society.
The men become extremely immature and ended up destroying the vase that contained the water in which they cherished due to its youthful properties.
These characteristics describe characters who although are annoying, will be gone in the next 5 years.
These characters are greedy, as they all try to take the woman for themselves, gluttonous, as they all devour the water, and lustful, also pertaining to their actions caused by the woman.
The part where the flower suddenly youthened revealed the science fictonious nature of this work.
The motivation was to create a cure for aging and becoming an old geezer.
The mirror represents the fear that might be caused by people and speculation of what something like this means for the greater society.
The guests first hand learned what greed can cause.
I know this because they saw themselves shift into better versions of themselves, only to be reverted shortly thereafter.
Coronavirus came out of nothingness and changed everything.
Suddenly we had to adapt to these new circumstances that were cast upon us.
For school, we started by doing Distance Learning, and all classes were held virtually.
There have been many positive experiences in this new world of learning, such as the ability to more reflect our sleeping schedules to how our bodies seek fit, as we can now wake up 30 seconds before class starts, as opposed to the 30 minutes required prior.
There also is much more free time to get homework done during nutrition and lunch that won't have to be done after school is dismissed.
Everything isn't perfect though, as many students miss seeing their friends in person, which can no longer be done.
That being said, the students can communicate through chatting apps like Discord, so not all hope is lost.
Overall, Distance learning is very good, and even though there is the elephant in the room that is the social aspect to school, most of this program has gone swimmingly.
Sentiment analysis is a powerful tool that can be used to evaluate if a given sample has a positive, negative, or neutral sentiment.
Within this paper, a bidirectional RNN trained to analyze sentiment was given sentences from Wikipedia article revisions over a certain time period corresponding to before and after a major current event which involved the subject in question.
Based on the findings, analysis suggests that political issues were more likely to experience a negative shift in sentiment.
This sentiment change was often sharp and sudden.
The use of language on the internet has been tied to people’s sentiments since the very early days; however, in today’s world, with people more than ever connected to the internet, this phenomenon has real world ramifications.
One of the largest websites, Wikipedia, a crowdsourced online encyclopedia, gets updated constantly.
While Wikipedia intends for its articles to be politically neutral, articles can oftentimes exhibit political skew (Greenstein & Zhu, 2012).
By evaluating Wikipedia articles over time, and tracking the sentiment for each article section, the correlations between current events and the way language is used can be observed.
Tracking these articles over time can reveal several details about the use of language on Wikipedia.
Wikipedia was specifically chosen as it has a somewhat consistent writing style, and is one of the largest worldwide sources of information.
Additionally, it has been posited that articles on Wikipedia undergo high levels of change during newsworthy times, which create articles that are noticeably different from the prior versions (Ferron & Massa, 2011; Keegan et al, 2013).
Such findings can include which topics are the most contentious, which types of topics seem to be affected the most by current events, and if shifts in the type of language used happen suddenly or gradually.
Researching these shifts in language use reveals information about both Wikipedia as a platform, and the people who use it.
As such, this paper aims to address three main research questions. 
RQ1: To what extent does sentiment change occur in relation to current events on Wikipedia?
RQ2: Is the aforementioned sentiment change a universal phenomenon on Wikipedia, or does it only occur on certain issues?
RQ3: Is the change in sentiment gradual or sudden?
In order to obtain such data, it is important that a computational model be applied, as manually sifting through (in some cases) over a thousand article revisions per article to search for subtle differences in tone would be an infeasible task.
To achieve this, a bidirectional recurrent neural network (BiRNN) was trained on labeled tweets from Twitter, and then was subsequently fed sentences from multiple article revisions of a single article to track the change over time.
Although domain shift of this type has the possibility of introducing noise into the model, a qualitative test inputting Wikipedia sentences into the model suggests that the model performs appropriately on sentences from Wikipedia.
To examine this data, computational sentiment analysis was performed.
Sentiment analysis, or opinion mining, is the computational process of using natural language processing to determine sentiment within a given text (Pang & Lee, 2008).
Rather than utilizing a simple sentiment analysis via cosine similarities (which can be used as a tool to determine similarity and therefore sentiment for individual words), this approach uses a BiRNN to learn more complexity about sentences and account for context.
Learning about language use over time through revisions as is done here is important as it shows how the information which people consume affects the sentiment of online writing.
It has been proven that the information that people intake can affect their sentiment on a topic (Mutz et al, 1997).
Since the way Wikipedia is written both reflects and affects how people feel, understanding these changes can show how people feel and can help us predict how people will feel in the future.
A corpus of tweets marked as positive, negative, neutral, or irrelevant was selected for this process.
This dataset was chosen due to the wide variety of topics discussed on Twitter.
Many corpora use online reviews as topics, but since those usually just talk about commercial products (whereas Tweets discuss a larger variety of topics), they would be less appropriate for classifying Wikipedia text.
All tweets marked as irrelevant were removed.
Each tweet was made lower-case and then it was embedded with spaCy’s pretrained embeddings (en_core_web_lg) with an embedding dimension of 300.
The labels were mapped to 0 for Positive, 1 for Negative, and 2 for Neutral About 17% of the data was removed due to being marked as irrelevant, and the remaining data was well distributed, with 22,542 negative samples, 20,832 positive samples, and 18,318 neutral samples.
Sequences were padded by the model at batch time, not during preprocessing.
No additional preprocessing (such as removing mentions or URLs) was applied.
A bidirectional RNN was trained on these embeddings.
An RNN was selected for this task due to its relative speed and effectiveness.
Training data was batched, with a batch size of 32 that was padded and packed.
The model consisted of two hidden layers, both with a size of 48 in each direction (or 96 per each timestep).
From there, this output was passed into two feedforward layers, the first of which had 128 features and the second of which had 64.
Dropout was used on the training with a value of 0.
3 in the BiRNN, and 0.
5 in the feedforward part of the model.
Training used cross-entropy loss, and Adam was used as the optimizer with a learning rate of 10^-4.
During training, the model was trained on batches of 32 shuffled tweets.
The model was trained for 144 epochs (at which the training started to plateau).
The model, after 144 epochs, ended up with a training loss rate of 0.
26, and a validation accuracy of 0.
Overall, the model performed with an F1 score of 0.
The validation set contained 1000 samples.
The validation samples were provided within the dataset used for training as a separate file.
Overall, due to the variability of the dataset, and the subjective nature of sentiment analysis, an accuracy/F1 of 0.
84 was deemed to be sufficient.
In order to test a wide variety of concepts, both pop culture and political topics were chosen for evaluation.
This choice was made, as politics and pop culture are among some of the most talked about aspects of modern society, and these topics were likewise edited frequently on Wikipedia.
These will provide data that shows many edits on ever-changing topics, which is optimal for this analysisDue to the varying amount of edits made to each individual article in comparison to the other articles, no uniform date range around a given topic was chosen, and instead a case-by-case evaluation was made to ensure that there were enough edits made before the event (at least a month), and enough time and edits made after the event (at least half a year and at least several hundred edits).
The chosen articles were as follows: Charli XCX, Dave Grohl, Russia, Ukraine, Zionism, and the Wikipedia entry about apples as a control.
Apple (the fruit, not the company) was chosen as a control due to apples rarely appearing in the news, and it not being a charged topic within current discourse.
All of these following topics were in the news and relevant in society one way or another, hence their inclusion.
British musician Charlotte Aitchison’s (Charli XCX) 2024 album Brat was a cultural phenomenon in the English speaking world, making her music and her fame grow.
This album was so popular that the summer of 2024 was retroactively named Brat Summer by some (Planey, 2024).
As such, a period of time spanning approximately half a year before the release of the album to about a year later was chosen for this article.
Foo Fighters singer and former Nirvana drummer Dave Grohl is an important figure in the music world, but to the dismay of his fans, he announced on Instagram in September of 2024 that he had fathered a child outside his marriage (Grohl, 2024).
Because this was a contentious time for his image, a period of time spanning about two months before this incident to ten months after the incident was chosen to see if and how this controversy would affect his main Wikipedia article.
Following the Russian invasion of Ukraine on February 24, 2022, both Russia and Ukraine were in the news.
These articles can help illustrate the differences in views between these two sides, and how the war has affected them.
For the Russian article, the span of dates goes from about a month prior to the invasion to a year after, and the Ukrainian data was collected from about two months before the invasion to a year afterwards.
The topic of Zionism has been controversial following the attacks on Israel on 2023-10-07 and the subsequent war.
Since it has been such a debated topic, this article can provide insight about how articles with contentious topics are edited.
The date range chosen for this begins three months before the October Seventh attacks and ends slightly over a year afterwards.
Overall, all of these topics provide ample grounds for evaluating the types of topics most associated with sentiment shift on Wikipedia, and provide insight into how these edits are accumulated, and how they last over time.
Graphs were constructed for each as a line graph, with the option of displaying all the sections in the article as a single graph.
Sections that did not exist in every revision tested were not used to ensure all sections existed for the whole graph.
Due to the sheer number of revisions used, a parameter was set on each graph to average a fixed number of revisions per data point.
Additionally, to ensure that the graphs show data over time elapsed instead of revision number, the averaged datapoints were placed according to the date of the last revision in the batch of averaged revisions.
Since (excluding the control) all the articles have relevant dates in which a current event occurred, there is an option to place one or multiple vertical lines on specific dates.
This approach, rather than putting a dot on the graph at the location, helps the multi-section graphs show where on the graph the event occurred, and if or how long after any change in sentiment occurred.
The control (apple) found that within an article that was not related to current events, sentiment tended to stay relatively consistent throughout the duration of the data, and stay around 0.
5 on the graph, ± approx 0.
One important thing to note about the graph for apple is that the etymology section tends to spike and dip substantially more than other sections — this is due to the length of the section, which is substantially shorter than other sections.
Even very subtle changes within the section can have drastic effects on the graph.
Therefore, while using this approach, very short sections are not good candidates for evaluating language change.
Within the article for Russia, most of the sections looked consistent, with two notable exceptions.
The first exception was the main contents of the article, at the top.
Additionally, the section for government and politics was also affected.
Both of the aforementioned sections showed declines in sentiment, although the way in which they declined was different.
While the main contents had a more gradual decline that seemed to start from the beginning of the data (prior to the invasion), the government and politics section had a sharp and sudden decline.
Additionally, the main contents tended to vary more throughout the trial, suggesting that subsections tend to show stronger sentiment trends than the main section.
In contrast to the article for Russia, the article sections for Ukraine tended to show no or positive change in sentiment, with the exception of etymology and orthography.
Similarly to the article of apple, the short length for this section causes even small changes in language to have large effects on the graph, explaining this change.
The rise in the contents section can be attributed to subtle changes in tone between revisions — for example, the model classifies the sentence The country is a member of the United Nations, the Council of Europe, the OSCE, the GUAM organization, and the Lublin Triangle, and is one of the founding states of the CIS, even though it never became a member of the organization (Wikipedia Contributors, January 2022) to be negative, due to the sentence ending with a negative implication.
In the newer revision, this sentence was changed to say It is a member of the United Nations, the Council of Europe, the OSCE, the GUAM organization, the Association Trio, and the Lublin Triangle.
(Wikipedia Contributors, March 2022), which the model classifies as neutral (due to there no longer being a negative connotation at the end).
Revisions similar to this were made to the article by Wikipedia editors following the start of the war, causing the sentiment within this section to become more positive.
In addition to the war between Ukraine and Russia, another major conflict was the war between Israel and Palestine.
Rather than looking at a country in this scenario, a philosophy was looked at instead, namely Zionism.
Certain sections, like the contents and the terminology tended to fluctuate, while most other sections remained relatively stable, with the notable exception of history and beliefs.
History was steadily declining ever since the start of the war, and beliefs tended to fluctuate, until September 2024 on the graph (which was when Israel launched attacks on Lebanon).
Unlike other articles where after a major dip in sentiment, the sentiment starts to steadily rise to a steady level, the sentiment stayed relatively stagnant for the remainder of the graph.
The graph for Dave Grohl shows relatively consistent sentiment over time (with the exception of Musicianship and equipment, which is too short to suggest stable estimates).
The dip in all sections about 40 days in was due to the formatting shifting within the article, and was deemed as meaningful due to the short length.
The graph for Charli XCX shows a spike in approximately October of 2024, which corresponds to an addition in the section about her heritage, specifically that she has pride in her Indian heritage.
The decrease in the early life section corresponds to an addition in the section about racism she experienced as a child being half-Indian and living among white people.
The sections for career and the main section tended to stay relatively stable, with a slight dip around the beginning of November 2024 making it more neutralThis was because the size of the section expanded, which increased the amount of neutral sentences that were compared, lowering the average that was above neutralFor the same reason as previous etymology sections, the tours section is ignored due to insufficient length causing too much noise.
Another pattern that is observed within the graphs was that when significant changes occurred, most often the sentiment ended up returning near the level at which it was prior to it changing drastically.
The notable exception to this is the main contents of the Ukraine article, which went up, and did not return to the original levels.
The two mentioned sections in the Zionism article did not return to normal levels, but this was because the dips happened very late in the collected range of revisions and therefore there was insufficient time for them to return to their mean position.
Overall, the findings indicated that political topics tended to experience the effects of current events in terms of sentiment change the most, whereas pop culture topics seemed to experience less sentiment shift.
The time taken for this change in sentiment to occur varied, but the change typically was pronounced.
The use of a BiRNN was effective for creating data, as the model was able to understand context within a sentence, however, a major limitation to this approach involved the dataset: there is a mismatch between what was used to train (Tweets) and what was evaluated (Wikipedia sentences).
While at first this may seem alarming, this did not appear to be a limitation or an issue based on qualitative analysis, as the samples tended to average around neutral sentiment (which a quick read through several Wikipedia articles matches), suggesting that the use of training data that is less formal than Wikipedia has a negligible effect on the quality of sentiment analysis performed by the model.
Future research can improve on the methodology used by this paper by ensuring that the model is trained on data from the source that is being evaluated.
These findings indicate the type of edits made to Wikipedia articles, specifically in regards to current events.
As mentioned previously, the issues that tended to receive the most attention within the sampled articles in terms of edits within were issues that were mainly ongoing political topics.
As such, it can be suggested that topics related to current day political events tend to see strong emotional shifts during current events on Wikipedia.
According to Wikipedia themselves, Wikipedia has an internal policy which states that articles must be written from a neutral point of view, which has the goal of representing fairly, proportionately, and without bias, the significant points of view that have been verifiably published by reliable sources on a topic (Wikipedia contributors, October 2025).
This is in line with what was noticed by one of the Wikipedia cofounders, who suggested that Wikipedia is no longer a neutral source and may now skew left politically (Sabur, 2021).
While this subjective and anecdotal claim cannot be proven based on the testing that occurred (in regards to the political leanings of Wikipedia), more emotional writing has been correlated to stronger influence on ones view of an issue (Baum and Rahman, 2020), and therefore the effect that emotional writing has on Wikipedia readers is a genuine question.
This paper does not aim to address political leanings and therefore this paper cannot prove or disprove the co-founder’s statement; however, it can provide insight as to why he might believe this.
It has been proven that when discussing political topics, people tend to use more emotional speech on issues relevant to their beliefs (Gennaro and Ash, 2021).
This is consistent with the data, as edits that change the overall sentiment of political topics tend to cluster soon after a major event.
An example of this can be seen in the Zionism article: while the section for the main contents seems to vary, it overall has fallen since the start of the war, and sentences like Zionists wanted to create a Jewish state in Palestine with as much land, as many Jews, and as few Palestinian Arabs as possible (Wikipedia Contributors, October 2025), which was not present in the article prior to the war.
Prior to the war, this section appeared to contain analytical and descriptive information, rather than emotionally based claims about what Zionists historically wanted.
The role of moderation is very important to this factor, whereas while the sentiment could change rather quickly, it usually returned to the ordinary sentiment level of the section prior.
The fact that the sole exception to this rule was a positive change in sentiment raises the possibility that Wikipedia editors may be more likely to moderate lexical shift in the negative direction than the positive direction.
Unfortunately, the only other similar datapoint (the personal life section in the Charli XCX article) did follow the aforementioned trend, that being said, it did level off at a similar yet higher mean sentiment, therefore this trend involving the moderation is inconclusive, and more research should be done involving this subject.
Due to the fact that current events are shown to impact the sentiment of Wikipedia articles, some may question the accuracy of Wikipedia as a source of information.
As previously mentioned, when people intake information, it does affect their perception of issues (Mutz etal, 1997), and Wikipedia has shown sentiment shift in areas relevant to online political discourse, which can suggest the type of edits made on Wikipedia.
That being said though, every editor of any news source will perceive the issues they are writing about and will in turn take a stance, which can be reflected in their writing, meaning that all news sources are susceptible to this phenomenon.
Despite these findings, Wikipedia still remains a reliable source of data (Mannix, 2022), despite the emotional writing style of political topics.
This is because sentiment and accuracy are different measures, whereas sentiment reflects emotion, accuracy reflects truthfulness.
This is an important and nuanced distinction, as even though it is safe to say the information provided on Wikipedia is likely true, that does not mean that people will be swayed one way or the other by the way that said information is portrayed on Wikipedia.
Overall, the findings from the data show several trends about Wikipedia (specifically these set of articles).
First, section that received sentiment shift tended to be topics that are politicalWhile these types of edits were made, typically the large changes in sentiment tended to return back to a steady level after some time, suggesting that the moderation on Wikipedia is effective within keeping overall sentiment consistent.
Additionally, Wikipedia has been proven to be an accurate source of information, although accurate does not always equate to unemotional, and the way political topics are discussed tends to be more emotional, despite the fact that Wikipedia has an internal policy of neutrality (Wikipedia contributors, 2025).
Despite this, Wikipedia is usually a good source for obtaining information, although caution should be taken when approaching political topics on Wikipedia.
Summertime is almost here!
The weather is great (finally)!, the birds are chirping, and yet, you are still eating your disgusting plain instant ramen.
However, I offer to you a solution — here are two delicious and medium effort recipes that you can make for yourself and share with your friends (this assumes you have them).
These recipes are healthy, delicious, and taste MUCH better than whatever chemicals you are ingesting with your ramen from its dissolving styrofoam container.
Recipe 1: Poke Bowls (serves 3)Did you know that Norwegian Farm raised salmon is safe to eat raw?
According to FDA Food Code References: 3-402.11, Aquaculture Fish, such as Salmon, that are served raw or undercooked are exempt from [flash freezing requirements], given that they are served feed that does not contain live parasites, which Norwegian salmon feed is free of due to heat treated feed.
Additionally, farm raised salmon (at least bought on the west coast) is almost always flash frozen as well due to traveling across the country.
In short, this means that Norwegian farm raised salmon is safe to eat raw.
Recipe 2: Hearty Pasta Salad (Serves 6)I know many a college student who eats pasta plain, no creativity, no sauce (or even worse, canned sauce), and no fun.
A great way to make your past more interesting and healthier is to turn it into a delicious, hearty salad.
Summertime is wonderful, and with two brand new and delicious recipes for you, it will be even more wonderful!
Hopefully, those of you who are hesitant to cooking due to the amount of effort required now can see that healthy and delicious food can be made for very low effort.
My artistic point of view about this production is characterized by struggle through contrast — struggle in adjusting to life in the US, struggling to retain ones culture, and struggle against oppression.
The acting style for this production should generally be loud and full of emotion but should at times also be reserved — themes such as violence and abuse should have one side’s being portrayed as strong and full of power (the abuser), and one side small and lacking power (the abused), symbolizing the dissonance between the two sides.
A good example about this would be when the two men grab the young girl.
When they grab the girl, the lighting should cast shadows that make them look bigger than they are, and the girl should be screaming and struggling, but not with full effort (to symbolize the lack of power she has in this situation).
This mindset of expressing struggle through dichotomy can be applied to other aspects of the production.
For example, the background can be representative of Medea’s struggle to assimilate to American culture, having the American elements of the Victorian household look neglected while the Mexican elements look colorful and full of life, representing her view about life in America.
This mentality can also be applied to the costume design, for example, when Medea kills Acan, his American outfit can look faded and dull, the only color being the blood.
This contrast once again represents the struggle for Medea to adapt, but also how Medea is losing Acan.
This contrast could also be shown through the music — within the same scene, there could be a medley of strong uptempo Mexican music when Medea is talking, which switches to slow, American music when Acan is talking.
Overall, my artistic point of view is that this production needs to emphasize struggle through contrast in the acting, set design, clothing, and music.
This production should be for a wide demographic but is especially important to happen now.
In such a tumultuous time, especially with the hostility to immigrants, it is important to watch a production like this to remind the audience both about the importance of keeping ones culture, the importance of assimilation, and the importance of accepting both of these.
This conflict is what leads to the play, and in a time where people are starting to be hostile to those who are different I believe this play is a good reminder of acceptance and empathy.
I want the audience to gain an understanding of the importance of kindness and understanding — the tragedy of this play (the killing of Acan), and many other misfortunes that the characters endure due to lack of kindness and understanding.
Many people in today’s world do not accept others and treat each other without kindness, and I want the production to show what outcomes such behaviors cause.
I hope that the audience walks away feeling bad for the characters, especially those who lost their lives, like Armida and Acan.
I believe that this is important, as the aforementioned importances of kindness and understanding play directly into the theme of struggle, and when characters don’t receive kindness and understanding.
A character such as Armida, despite being in the grey area within this story due to her influence, was still an active victim, just like Acan, whose death is equally tragic, showing that approaching struggle (in this case between Mexican and American culture) without kindness and understanding can have major consequences.
My central image is the border between the United States and Mexico.
I believe that this represents the transition and struggle between Mexican and American culture in Medea’s life, and also the trauma from getting across the border.
I chose this specific image, as I am fascinated in man made structures and their effects on people.
This would also be seen in the costumes — those in the story who had to pass through the dusty border, especially those who were negatively affected by this experience like Medea, would wear dusty clothes.
I would motivate the lighting to be bright — this image shows the day in full light, but I would also include shadows (since the wall creates a large shadow), placing characters who are facing negativity in the dark.
This ties back to the artistic point of view, as it really highlights the characters undergoing struggle, and accentuating their differences.
I would use this scenery directly when the characters are crossing the border, as it shows the true size of the wall, symbolizing the contrast between Mexican and American lives, and the struggle to incorporate both together for Medea.
I would also incorporate this into sound design, because even though the house in Boyle heights is in the city (a loud place), the border represents the inner struggles of the story, which is a quiet place, meaning when conflict arises in the production, the background noise should be reduced to a minimum.
In these quiet moments, sounds that might be heard at the border (such as screeching of birds, or the rustling of wind) can be heard, timing.
This represents both the dirtiness of the border, and the emotional aspect of it.
To the characters in the story These all help me convey my artistic point of view by accentuating differences between characters, showing how the differences between them cause struggle, providing contrast between the sides of the border, both figuratively and literally.
I plan on using the elements of design to construct sets which portray environments in a fashion that accentuate certain characteristics that show contrast.
Some examples would be using lighting to make the difference between bright spots and dark spots much greater than typical, or making sure that characters that are in conflict are closer to each other, but making sure that the person who is portrayed as stronger is physically elevated to show the power dynamic.
Choices like these show how differences between characters cause struggle.
Line: I would use lines for positioning, but also some backdrop elements, such as individual lines for the fence at the border.
These lines shouldn’t be straight though, to promote disorder (which represents the disorder in the characters lives).
Shape: I would use shapes for shadows in the backgrounds, and for props in front of the backdrop that are supposed to receive less attention from the viewerForm: I would use geometric objects to make objects within a scene that should get more attention from the viewer, as opposed to the 2 dimensional ones that will be further back.
Value/Contrast/Color: I will use this in full force, I will make sure that things full of life are brighter and have more vibrancy, and things that are on the receiving end of struggle are void of color and are darker.
Texture: I would use texture mainly within costume design, making the clothes of those receiving struggle to look battered and unkempt, showing how it is affecting their life.
Scale/Proportion: I would make characters who are facing oppression or struggle look smaller by placing them lower down on the set or by placing them further away from the audience.
Motion: I will make scenes that involve movement have the characters constantly moving, and the scenes that do not involve movement having the characters stay in place to increase the contrast between the two.
Emphasis: As previously mentioned, done through form, and additionally through light.
Balance: There will be unbalanced sets, especially in scenes with conflict.
Rhythm: Like with balance, there will be irregular rhythm, as it shows the disorder in the characters life.
Space: I will put important things closer to the stage and less important things further away (as mentioned in form)Unity: The scenes should be slightly chaotic, but especially chaotic whenever conflict arises.
Overall, most of the production decisions fit within my analysis of the script.
The big decision I feel needs to be rethought is spacing — I think that since there is a lot of motion, having characters in conflict be closer together would be very difficult to choreograph, so perhaps placing less emphasis on the movement, and more emphasis on other forms of contrast.
Throughout the last couple of years, the Jewish community has faced many positive and negative experiences, and we as a community have grown stronger and changed in many ways.
Although there were many challenges in the past couple of years, one benefit is the creation of a Jewish student-led magazine — Fig and Vine.
This year, many members of the Jewish community were able to contribute and write some phenomenal articles.
One such member is Shani Levy-Richards, who has just graduated this year.
I had the pleasure of being able to interview her and to share her thoughts to the rest of the community.
The first question I asked Shani was how she felt being a Jewish student at UCSB.
She said that overall, she liked it (due to the large Jewish community).
By percentage, UCSB has the largest Jewish community of any UC, leading to our community to be large and strong.
She also mentioned that she liked the various events that happen within the UCSB Jewish community.
Having such a great Hillel and Chabad here at UCSB is very helpful because these events are fun, engaging, and a great way to meet friends.
Some of her favorite memories that made her feel like a part of the community have been Pesach at Hillel, and various events at Chabad.
She says that she feels like the community here is more welcoming than other UCs.
When I asked her about what brought her to Fig and Vine, Shani told me that she is a history major, and that Jewish history has always interested her.
She told me that there is so much to talk about and learn within Jewish history, hence her fascination.
I agree — we have existed as a people for the last 3000 years, and within that time so many events have occurred.
She specifically told me that one interesting part was how holidays evolved over time.
I followed up with Shani and asked her if she had taken any Judaic study classes at UCSB.
I have taken many, and have learned how Jewish practice and laws have evolved over time.
She told me that unfortunately she had not had the time to take any classes despite the fact that she wanted to, but she had many friends who had taken Judaic studies classes, and who had enjoyed them.
Personally, I have taken several of these classes with Professor Medina, and have learned a lot about our history, so I am very sad for her that she did not have the opportunity to take those classes.
After that point, I inquired about her experiences as being an author with Fig and Vine.
Shani told me that she enjoys documenting the history — to her, it felt similar to a diary.
She also mentioned that she feels more confident as an author than as a speaker, making it easier for her to express her thoughts.
She told me that homework does not feel empowering to create, but since she started writing on her own volition, she feels very empowered in her writing.
Following, I asked Shani about her plans post college.
She told me that she plans on taking a gap year between college and grad school.
She wants to do a masters in English (potentially with history).
Furthermore, she told me that although her time is done at UCSB, she still has a twin sister here, so she will be visiting.
The penultimate question which I asked her involved how the Jewish community has evolved at UCSB.
She told me that over time, the community, through events such as October 7th, the rise in anti-semitism, and incidents of hate directed at specific individuals, has gotten more tight knit.
The Jewish community at UCSB, and within the greater world has been through a lot, so in order to ensure that we are able to prosper, our communities have gotten stronger.
I then asked her if she had anything else to share, and she told me to let every know that she wishes them good luck.
Overall, Shani has contributed greatly to the community by teaching us all about the history, and she will be dearly missed.
Lets all wish her a successful time at graduate school, and hope that she visits Santa Barbara very soon!
Throughout history, there have been many examples of indigenous groups that face oppression from the government which is supposed to help them.
Oftentimes, when this happens, the oppressed people do not take kindly to being oppressed and rebel, sometimes gaining de facto independence from their oppressors.
One such example are the Zapatistas, which are formed by several indigenous groups within Chiapas, Mexico.
They historically faced oppression, such as having diminished land rights, and other racist policies.
This caused the indigenous people within Chiapas to revolt, and the day that the NAFTA agreement was being signed, the Zapatistas declared themselves as an independent state.
This caused conflict within the region, leading to a massive military response from the Mexican government, a governor getting kidnapped by the Zapatistas and being forced to do hard labor (Delpama), failed and disingenuous attempts of peace being proposed by the Mexican government, a massacre by the Mexican government (Fisher), and many more such tragic and horrific events.
The reason for these forms of violence by the Mexican government was because they wanted to quell any sort of rebellion put upon them.
This is because they wanted to appear to the broader global stage as a strong, capable country, being that his was a time of great growth for Mexico, and they were finally no longer considered a 3rd world country.
Even though efforts have been made to ensure peace in the region over the last three decades, no such peace has been formed, due mostly in part to the Mexican government’s refusal to address the issues that caused the Zapatistas to rebel in the first place.
Such a rebellion is not unique — there are many such examples an or many indigenous groups banding together in order to fight back on an oppressor, leading credence to the validity of such movements.
Overall, the Zapatista’s struggle for peace and security only exists and is exasperated by the actions of the Mexican government.
For over 2000 years, the Mayan population has thrived within Chiapas, keeping the land healthy and their home.
In 1523, however, their peace and prosperity was destroyed when a colonizer, Luis Marín, started conquering the area.
He was able to take over the land of many of the indigenous people, but he ran into issues fighting the Tzotzil indigenous people (Schmal).
He fortunately was not able to conquer all of the land — however, the Spaniards would not give up their goal of conquering all of the Mayan’s land, so they kept fighting.
By 1528, all of the land had been conquered, and the last of the indigenous people that were free were now under control of the Spanish crown.
The Spanish instituted a rule of slavery for the indigenous people, further diminishing the rights of those who belonged to the land, until it was abolished.
Even though this rule of slavery had ended, the Spaniards (and later the Mexicans) were always at odds with the indigenous people.
Historically, Chiapas’s indigenous people have been excluded from the governmental decision-making process as well as from enjoying basic human rights and services such as education and healthcare (Godelmann).
Additionally, the government wanted to gain land rights, so it fabricated a situations where it appeared as if government simply was resolving land disputes between different indigenous groups (Krinsky), and seized all of the land, despite the fact that there were many other indigenous groups that were not the Lacandron Community.
(Krinsky) As can be seen, both the Spanish and Mexican governments have historically treated the people within the Chiapas state extraordinarily unfairly, leading to deep resentment within the people, and an eventual liberation movement.
This aforementioned tension within the region in regards to the indigenous people boiled over on New Years Day of 1994 (Mora).
This day was going to be the day that Mexico joined the NAFTA (North American Free Trade Agreement), and therefore would be considered in the eyes of the greater world no longer a third world country.
This came about through the Ejército Zapatista de Liberación Nacional’s (EZLN), known in English as the Zapatista National Liberation Army’s declaration of war against the state of Mexico.
This was not simply preformative, as the Zapatistas briefly occupied six towns in the state’s central highlands, before retreating to the jungle to avoid further conflict (Harvey, Arnson, 129).
This fighting continued, and further tensions with the Mexican military required the Zapatistas to resort to more brutal means of enforcing their rule, such as kidnapping Absalon Castellanos Dominguez, the 70-year-old former Governor of Chiapas (Depalma).
Despite peace talks occurring, this tension still exists to this day (albeit less severe), as the Mexican government does not want to listen to the demand of the Zapatistas.
These demands are as follows: truly free and democratic elections, the incumbent federal executive and the incumbent stare executives… must resign, Recognition of the EZLN as a belligerent force, the end of neoliberal policy, General elections for the whole state of Chiapas, among many others (CCRI-CG of the EZLN).
Since these demands have never been met, the EZLN still holds itself as a de facto state, and is and continues to be at odds with the leadership and military of the Mexican government.
It is important that this be known, as these demands exist due to the abuse of both the Spanish and Mexican governments, and until the Mexican government rectifies its past wrongdoings, the people of Chiapas will never accept them as a valid state and will continue to posit their own independence.
The reason that all of this conflict exists lies solely on the actions (and the lack of actions) of the Mexican government.
The actions that the Spanish government did in the past were horrible — enslaving indigenous people, running them off of their land, and many other things.
When Mexico gained independence, they finally had the opportunity to right the wrongs that the Spanish crown had done to the indigenous people — a fresh start.
This fresh start, however, was (as previously mentioned) not such a fresh start, and in fact worsened the lives of the indigenous people.
Moving towards the modern day, the modern version of the government has acted against these people.
Even though their anger is rightful and justified, the government treats them as a bunch of rebellious hoodlums, and treats their demands as irrelevant and unjustified.
This government treatment of indigenous people is not unique to the state of Mexico — the United States, Canada, Australia, and so many other countries have treated their indigenous people unfairly.
That being said, many wrongs do not make a right, and Mexico treating their actions as justifiable just because others have been able to get away only serves to worsen the situation in Chiapas.
Had the Mexican government treated their demands as they should have, not only would there be peace within the region, countless lives would have been saved, but this would only help show the rest of the world that they in fact were a competent power, and worthy of being considered a first world country.
Their treatment of the indigenous Mayans whom inhabit this land has shown and continues to show the world that the government is cruel, and only serves to give more attention and positive press to the EZLN.
Overall, had the Mexican government treated these indigenous people fairly, such a situation would have never occurred in Chiapas, and had they chosen to listen to the reason provided by the EZLN, this crisis for the government would be a situation of the past and not of the present.
As previously mentioned, the reasoning for the Mexican government’s oppressive rule over the Mayans and their Zapatista movement was to make their country appear more competent and powerful.
While Mexico had been seen as somewhat strong during the founding of their nation, throughout the 20th century, due in part to drugs, gang activity, and weak economics, Mexico was seen as a Third World country, and was not taken seriously on the global stage.
Because of this, they felt that they needed to look ‘ready’ and ‘presentable.’ 
The obvious humane solution to this would be to get in talks with indigenous people and figure out what would help them feel like they could peacefully live within Mexico; however, as has been seen previously, this was not the approach taken by the government.
Likely the reason why this was was because of public sentiment towards indigenous people — at the time (due to previous propaganda), many Mexicans viewed the indigenous people as inferior, so taking such an approach would risk the non-indigenous Mexicans getting angry and starting an uproar.
This would obviously look bad for the Mexican government as well, and as the indigenous people had fewer resources, less power, and a weaker voice, they chose to oppress the indigenous people.
As has been mentioned, this oppression was the basis of the creation of the Zapatista movement.
At this point, the government had had their reputation smeared — instead of showing the world that they were a competent first world country, the Zapatista movement instead showed a darker and more sinister side to the government.
In response, the government retaliated against the Zapatistas, furthering the point that the Zapatistas were making about the government.
At any time, the Mexican government could have accepted defeat, which would have made the current government look better, and would have better the lives of both the Mayans and the general Mexican population, but the tyrannical leaders at the time chose violence, hurting the image of Mexico, which in turn led to a spiral of more violence and tyranny.
This type of struggle, while having unique facets, unfortunately is not unique to just this localized instance.
A great example of this comes from geographically close to Mexico, Guatemala.
The indigenous Mayans of Guatemala had also faced many hardships for generation upon generation, and despite the changing world of the 21st century, this trend continued and continues into the new millennium.
For example, there is currently a fourth invasion going on, which is distinguished from three previous invasions: first, the Spanish invasion and colonization; second, the creation of the plantation economy in the late nineteenth and early twentieth centuries; and, third, the state-sponsored genocide that occurred during the Guatemalan Civil War (1960–96) (Batz, 3).
This invasion is due to the government deciding that they are going to, without the permission of the indigenous people whom live there, build a hydroelectric plant.
This is very similar to the aforementioned example of the Mexican government seizing land for their own purposes — just because a country is within a certain area does not mean that it has full autonomy and privilege to force people off the land for its own gains.
After this, the government sent in the military to destroy the protests that started because of this declaration of construction.
This too is very similar to the aforementioned example in Mexico, where in 1994, the government sent in the military to attempt to (and not successfully) end the defect state that the Zapatistas created.
This is a common theme worldwide — governments believe that they are always righteous, and therefore possess the right to do whatever they want to quell rebellion and dissenters because they always must be ‘wrong.
’ Unfortunately, this line of thinking has been implanted in the minds of many, so one must fight against it rather than waiting for this line of reasoning to disappear.
Although situations like the two that were previously mentioned are grim and dire, it is important to remember that fighting for rights is important, and has, does, and will continue to positively benefit the people who are fighting.
A great example of this comes from the United States of America, specifically Colorado.
Many indigenous tribes lived along the Colorado river for millennia upon millennia, until the descendants of white Europeans, specifically members of the Church of Latter-Day Saints, colonized this area and forced the native people out (Sakas & Bures).
This river was essential to these indigenous people — their farming and tradition was built upon it.
In 1922, the river was divided among many states — though this did not give the indigenous people any rights to the river.
This left the indigenous people of that area in a very tough spot — they now had even less access to the river they barely had access to for the past couple of centuries.
They then protested this, and fought tirelessly for decades to ensure that they got access to the river which they called home.
Finally, in the 1990s, all the indigenous tribes of this region finally got what they so desperately needed and deserved.
This example shows the efficacy of fighting against the government to get what is needed.
In this situation, the time required to get the government to meet the demands of the indigenous people exceeds the time which the independent Zapatista movement has existed.
While people may look at the Zapatista movement and believe that it is just a waste of energy for the people there, it is important to remember that the United States government is just as unwilling to help indigenous as the Mexican government, but even for a self centered government like Mexico, eventually, they will give in, hence the importance of fighting on continually.
In conclusion, indigenous people all over the world face oppression by their governments, and the indigenous groups that form the Zapatistas are no exception.
The indigenous people of Chiapas during the rule of the Spanish crown fell victim to oppression, slavery, and mistreatment.
This mistreatment did not stop during Mexico’s independence from Spain, and continues on into the modern day.
Because of this, the indigenous groups of Chiapas banded together, and declared independence from Mexico on January 1st, 1994.
Rather than listen to the demands of the Zapatistas, the Mexican government sent the military in and unleashed violence upon the people who were living there.
The reason that the Mexican government unleashed such a strong response was because they wanted to be seen as a strong, 1st world nation, and by being oppressive to its subjects, ended up doing the opposite.
The whole reason that there is conflict in this region in the first place is entirely the fault of the Mexican government and its inability to listen to the demands of the people who they oppress, and is the responsibility of the Mexican government to fix.
Had previously the Mexican government listen to the needs of the indigenous people of Chiapas, no such de facto state would have been formed, and had they listen to the needs of the indigenous people after the de facto state had been formed, the ensuing violence, death, and embarrassment for the government would have never happened.
This situation is not just unique to the Zapatistas — many movements like this exist within the world (in smaller scale or larger grandeur), and many movements by indigenous people have been successful (such as the aforementioned example with the Colorado river).
Overall, the existence of the Zapatista state and the violence that has ensued due to it are the fault of the Mexican government, and until they address the issues that the Zapatistas ask, no positive change will occur.
There are many features of Rabbinic literature that are very different from the Hebrew Bible.
First of all, whereas revelation was something that occurred through prophets in the HB, within Rabbinic literature, it was carried out through knowledge and interpretation of Jewish texts.
Two additional features that distinguish Rabbinic literature are makhloket and safek.
Makhloket refers to disagreements, where multiple Rabbis would have opposing opinions on an issue and would have them debated, and safek refers to doubt or uncertainty, which was not discussed as often within the HB.
The emergence of the Torah scholar contributed to this phenomenon of disagreement because rather than just understand the Torah, Torah scholars now both understood and interpreted the Hebrew Bible.
Since the Hebrew Bible was not ultra specific, and oftentimes vague, this could lead to disagreements in interpretation.
The sages disallow this.
Even when a divine voice spoke out to the sages, they still stated that it was required to show proof about a matter in order to have it count for debate.
The Tannaim and Amoraim have very different views about Torah Scholars.
Whereas the Tannaim believed it best for one to understand the Torah, the Amoraim placed the importance of Torah study on debate and interpretation.
Additionally, the Tannaim view Torah study as something that is done in addition to a job, whereas the Amoraim see it as a job within itself.
There are two very good examples of this, one of which is the gathering of manna (or specifically when manna should be gathered), and the story of Rav Shimon Bar Yochai in the cave.
The first example comes from the gathering of Manna.
It was stated by the Tannaim that it should be done once a day, every day.
While this might seem cut and dry, the Amoraim suggested an interpretation about this that it is permissible to collect manna for the day and for the day that follows, as is done for Shabbat.
This shows that the Amoraim view a Torah scholar as one who can interpret information from Jewish texts, whereas the Tannaim are just recounting events and laws.
The second example comes from the Babylonian Talmud.
In this story, Rav Shimon Bar Yochai emerges from a cave, and his son in law sees that he is in very poor physical condition and starts to cry.
Rav Shimon Bar Yochai states that he is happy, because although he is low in physical health, his knowledge and understanding of the Torah is much higher.
This is important, because it shows how the Amoraim view Torah study as work (as he locked himself away in a cave to study Torah), rather than a Tannaitic Torah Scholar who would have lived within their town and had an additional job.
Makhloket is a very important part of Rabbinic law, as it serves to show differing opinions of Rabbis.
The jug-breaking case is a great illustration of this, as the original law simply states that an owner is liable for damages for a jug left within a public area.
The makhloket surrounding this discusses in which cases this law is considered and several opinions are presented, two examples of which are this applying when the jugs completely block the road and when it is around the corner.
This reveals that makhloket presents differing opinions of Rabbis.
The Mishnah preserves minority opinions even if they aren’t the ones being followed because in the future, a court might change its view.
If in the future what once was a minority opinion becomes the majority and it is viewed as being of a higher spiritual status, then this opinion is the one that is followed rather than the previous majority opinion.
This shows that the Rabbi’s held specific situations and common views as important facets of laws.
A contemporary example of this that is important is the discussion surrounding Roe V.
While this ruling had been previously followed, the supreme court in 2022 decided to overrule the previous decision of the court.
The minority opinion was stated in the original ruling, which was cited in the new ruling, showing a modern example of the importance of recounting minority views.
This principle is important in declaring a husband dead because the consequence of a wife remarrying does not change based on how long she waits (should she be wrong about the death), but the certainty of the husband being dead goes up over time.
For this reason, the Rabbi’s decided that a 3 day waiting period would be sufficient for determining a husband dead.
Even though an example of somebody being alive after 3 days was brought up, the Rabbis still held that there was high enough certainty after 3 days for a wife to remarry.
The contrary example to this is saving a life on Shabbat.
If somebody were to be covered in rubble, the Rabbis deemed it to be permissible to remove rubble so long as they did not find that the person was dead (at which point they should stop).
This is for two reasons — unlike the previous example, this example is time dependent, for which the Rabbis allow more leniency.
Additionally, the mitzvah of Pikuach Nefesh supersedes all other mitzvot, hence why the Rabbis allow this example.
Within this passage, the majority view a vessel as being complete when it is acts as a functional vessel, whereas Rav Eliezer views it as complete when it appears as a vessel (with the caveat of something being repaired with sand not being considered as appearing as a vessel).
This is important, as it shows two different legal approaches — one where the law is strictly defined — absolutism (as in Rav Eliezer’s case), and one where it is defined by the view of the majority.
A modern example that illustrates the difference between these views this would be the speed limit on the freeway.
Somebody who helped a strict interpretation of the law would state that anybody going 66mph or faster would be breaking the law and should be fined, whereas the majority view (and what the majority of people do) Is follow the speed that traffic is moving on the freeway rather that driving at exactly 65 mph.
Throughout its history, Jewish law evolved a great dealIn the beginning, the Torah was simply a partially known tool that was used for teach — however during the second temple period it became both fragmented (due to the sects) and statutory.
After the temple was destroyed, the law became much more interpretive and united as compared to the 2nd temple period.
Jewish law began during the biblical period (First Temple).
In this time period, the law was viewed as pedagogical rather than being statutory.
In other words, this meant that the laws were meant to teach, rather than to be strict ways of life that must be followed.
Additionally, for much of the first temple period, the Torah was hidden, so the exact laws were not followed (a good example of this would be the King David and the spoils of battle).
Within the 2nd temple period, a sharp difference emerged based on how the law was interpreted.
Laws were followed much more strictly, and were now seen as a legal code rather than teaching.
Additionally, sects (such as the Pharisees and Sadducees) forms, each having their own interpretation of the law.
Within the Qumran community for example, they did not view the 2nd temple as being valid, and therefore set up their own community near the Dead Sea.
These changes occurred due to the Torah being better known, and questions about the validity of the Maccabees as leaders of the Kingdom of Israel.
After the fall of the 2nd temple, the Rabbinic period began.
The Rabbis did not want a repeat of the sects from the 2nd temple period, so they allowed more flexibility within interpretation of the law.
To be more specific, when the Amoraim constructed the Gemara, they recounted multiple views so that future courts and different communities could have flexibility under the law.
Additionally, within this time period, the law evolved too — starting at the beginning with the Tannaim, the Mishnah was created which was a recount of the oral law, and later during the period, the Amoraim made the Gemara, which provided further interpretation of the Mishnah, both of which make up the Talmud.
Overall, throughout Jewish history, the use of the law, and how the law was used has changed significantly.
In the beginning, the law was a tool for teaching and was partially unknown, after which (during the 2nd temple period) became a strict statutory code while also being fragmented by the multitudes of sects, before Judaism became united again during the Rabbinic period, where interpretation of the law and court rulings became important.
During this course, we went much more into depth about various interpretations and discussion/debate about the law (makhloket) than we did during the Judaism class during the fall.
My takeaway from this is that multiple opinions can be correct because of the many context — this relates to how multiple opinions were recounted so that future courts could change the interpretation of the law if they saw fit.
This is important in life because although the world may seem very black and white, this simple fact shows that it in fact is not so black and white, and in many cases, a different perspective is not an incorrect perspective, but simply just a different perspective.
Journal 1: Reflecting on this journal, I feel that I have learned a lot about how people are educated about indigenous people.
Going into this class, I had very little experience studying the lives of indigenous people (specifically how it relates to the state), and having learned that the phenomena of indigenous people’s interactions with the dominant power of an area often being misconstrued is a very common occurrence.
I feel that it is very important for people not only to learn that specific examples are wrong (such as the pilgrims being good guys example), but that this practice is very common, and understanding why it happens.
Journal 2: I think that what I wrote in this journal rings very true.
As a linguistics student, I can also think of many examples about colonization and power dynamics affecting language (think about how the large bed room where a slave owner could have slept is called a master bedroom).
While these racist and prejudiced pasts to certain parts of various languages lexicons may not be apparent in every day speech anymore (for example the greeting long time no see originally was a mimicry of what English speakers thought that Chinese speakers speaking English sounded like), but should be taught about.
Journal 3: I feel that this was the journal I forgot most about throughout the course, so revisiting it was very nice to do.
As a linguistics major, I have learned about how certain words can evoke certain emotional responses, and an otherwise trivial difference in word choice can be the difference between the message that you want to have sent being sent and not being sent.
I believe that the choice of words being rage rather than fear can evoke an extraordinary difference in emotional response, and changes the role of the women from sounding like weak and helpless to strong and powerful — which they are.
Journal 4: I consider myself to be an environmentalist — and as somebody who considers himself an environmentalist and cares about how indigenous people are treated, this journal hits home on two different fronts.
For example, the anecdote from Batz about the dam constructed on Mayan land without the Mayan’s consent bothers me both because this dam can potentially harm the environment in drastic ways, but also because it was an infringement on their land.
I believe that both of these factors are very important when holding governments accountable — oftentimes people like to focus on one or the other, but both are meaningful and important issues.
Journal 5: I think that blood quantum are a very important topic to be discussed.
I did this journal entry before we watch the video about the Indigenous people in Canada, and how they were not treated as a part of their community, and I believe that that is horrible, and the culture that you grow up in determines your ethnicity and what community you belong to, not your genetics.
Additionally, oftentimes these blood quantum are only a problem for women who intermarry rather than men, not just making society more xenophobic but also more misogynistic.
Overall this is an important issue that I hope more people will start to discuss.
Journal 6: I think assuming that western education, and the systems that surround and govern it are the default and nest way to educate is a very dangerous assumption to make, and a myth that needs to be distilled.
For example, Finland has some of the highest test scores in the world, and they spend significantly less time learning.
Additionally, many forms of western education are competitive rather than collaborative, which can lead to heightened anxiety and an overall worse outcome for science as compared to those who are not stressed.
It is important that these myths get distilled, and that we all learn the truth.
Journal 7: This journal was speaking about how certain indigenous communities in Latin America have been fighting for their rights, in regards specifically to their culture.
This is a very important part of learning about how indigenous people get affected by the state — indigenous people are not idle ducks — they will fight a gains unfair laws and orders put upon them.
It is very important to remember that learning about these fights against tyrannical laws are just as important as learning about the tyrannical laws themselves — having information about both helps us better understand the context of ever indigenous struggle in the world.
Journal 8: I really enjoyed this talk — I am a very big dan of music (I have been playing piano since age 5, I love listening to music and making music), so I can appreciate the importance of music to both individuals and to a larger culture.
I enjoyed learning about all the ways that the indigenous communities in Oaxaca preserved their culture, but the music part specifically stood out to me for the aforementioned reason.
Overall, my main takeaway from this week’s journal entry is that there are many important facets to one’s culture, and finding ways to preserve them is very important (such as with the music)Journal 9:I really resonated with this journal entry because I really care about nature and I do my best to be in harmony with it.
I love going on hikes with my friends, and loving nature for what it is and has to offer shows me that the oftentimes overly urbanized capitalist society does not appreciate where we came from.
I always do my best to keep nature clean and safe — there only is one Earth upon which we live, and taking care of it is very important.
That being said, the Earth (as discussed in the reading) can benefit sometimes from being taken from, so it is important to be a part of the planet and to love the planetJournal 10:I believe that state intervention is a major problem when it comes to indigenous issues.
Although this is not necessarily about indigenous rights, what has been currently going on with ICE and the Trump administration is a good demonstration about the problems with state intervention.
By the state not listening to the demands of the protestors and instead sending in the ICE and making the situation more dangerous for everybody, they end up aggravating people and showing people that this aforementioned kind of intervention (rather than listening) is a major problem in our society today, in regards to the state.
AbstractIntroduction.
Patient evaluation and diagnosis is important in cases of aphasia to provide good treatment and support.
A YouTube video of an aphasic is watched and a diagnosis is provided from said video.
A transcription was made for the video, and it was found that the patient had difficulty forming long sentences, and oftentimes fell back on familiar language.
It was determined that the patient in question has Broca’s Aphasia.
Correctly diagnosing and treating Aphasias is important to help the patient, and give closure to them and their family.
Learning about aphasias and the type of symptoms that they produce is important as it can help patients receive the care they need, and help those around them understand what is happening and why.
Additionally, since aphasics can get better over time, it is important to diagnose it so that they be in an environment conducive to brain recovery, such as language rehabilitation.
While many people might rush to diagnose the aphasia quickly, getting language rehabilitation at any point will help about the same amount (Moss & Nicholas, 2006).
One of the major types of aphasia, Broca’s Aphasia, is caused by damage in BA 44 and 45 and patients often have issues with speaking fluently, finding the correct words, articulating well, repetition, and the production of complex sentences (Dronkers etal, 2007).
Because other language tasks, such as semantics, and comprehension, occur in other parts of the brain, even though they are closely intertwined with tasks associated with Broca’s Aphasia, they are not affected in Broca’s Aphasics (Schwartz etal, 2009).
Case StudyThe person being discussed is an elderly woman, Jan, who is having a discussion with her daughter, Carly, on a YouTube channel they share dedicated to informing people about aphasia.
The specific video is about how to have a good conversation with an aphasiac.
ResultsWithin the conversation, she had difficulty creating complex sentences, and saying the name of a number (although she could write it down).
Additionally, her daughter (who also is her caretaker) already had a sheet on hand with many activities on it, further indicating that she had problems expressing her thoughts.
Jan additionally had multiple phonemic paraphasias.
Patient also extensively used the word yeah, as this was the only interjection used by Jan (at least during this recorded snippet).
Additionally, almost every utterance she said was very short, lending to the fact that it was difficult for her to form long utterances.
DiagnosisBecause Jan’s speech was at times telegraphic, and she had word finding difficulty, the diagnosis that is given is Broca’s Aphasia.
Jan showed a high level of comprehension (ruling out Wernicke’s Aphasia and Global Aphasia).
There were also many signs of Broca’s Aphasia, such as a common word or phase being often used (in this case yeah), and the knowledge of a number, but not being able to say the name of it.
DiscussionAs previously discussed, the diagnosis and understanding of diseases like Broca’s Aphasia are important as they can help get patients the best treatment that they deserve and give closure to both them and those close to the patients.
Since certain aphasias are caused by certain areas of the brain (Schwartz etal, 2009), being able to know what is and what is not affected for patients like Jan is essential because treating her like a Wernicke’s Aphasiac will not help her in any meaningful way, and would be mentally taxing for her.
Giving her the correct treatment could, as previously discussed, bring back language functionality to her brain (Moss & Nicholas, 2006), further stressing the importance of correctly identifying and treating aphasias.
Additionally, knowing a rough idea of the area in the brain in which the the stroke occurred is helpful because it can predict other symptoms of the stroke.
Since Broca’s area is near BA 44 and 45 (Dronkers etal, 2007), other related processes that occur in the frontal lobe nearby will likely be affected.
Overall, correctly diagnosing and treating aphasias is vital to the efficacy of short term care of patients, the potentiality of long term restoration of language skills, and the closure that the patient and their family deserves.
Judaism and food have a long and complex relationship — we have entire holidays dedicated to abstaining from eating.
We have certain foods we can’t eat, and certain foods we must eat.
And finally, lets not forget, gefilte fish.
Jewish culinary culture is formed by these very laws, making Kashrut (kosher laws) an important facet of our food and our culture.
Growing up, I kept no form of kosher, and up to going to school here, I continued this trend.
If you were to ask me what my all time favorite food was, I would have enthusiastically said shrimp!
Having come to college, I formed a deep connection with my Judaism, and in turn, with Kashrut.
While I might not be fully orthodox kosher (I will still eat at non-kosher restaurants as a pescatarian with a shellfish allergy), I do keep a somewhat strict degree of Kashrut.
I may be a Jew in IV who keeps kosher, but I am not the only Jew in IV that does, so I wanted to see what others’ experiences with Kashrut in IV and at UCSB were.
The first person to whom I posed the question of Kashrut in IV was Nicole Zar.
She told me that she grew up keeping kosher, and that this was important to her, so she kept keeping kosher in college.
To her, it is totally doable to keep kosher here, despite the many obstacles.
She said that she often relies on Chabad for meals (especially on Shabbat).
Although it is not simple, she told me, it is a very rewarding part of Judaism that [she] takes pride in and a mitzvah that [she] enjoys keeping!
As Nicole said, UCSB Chabad besides being a keystone to the Jewish community is also one of the best sources of Kosher food in Isla Vista.
Miri Klein, the Rebbetzin at UCSB Chabad, is the mastermind behind all that is food at UCSB Chabad and works tirelessly to ensure that Jewish mouths in IV are fed.
She told me that the biggest challenge for providing so many people food is that there are many difficulties getting large quantities of food — either the foods need to be bought from a kosher market, or comes in very small quantities.
She suggests that anyone who wants to start keeping kosher contact her with any questions, and go to Trader Joes.
I was able to get another perspective from Jake Nguyen, another student who keeps Kosher.
He said that personally, he has not had to make any concessions with Kashrut when coming here, and that the only downside to a kosher diet is that it is more expensive.
All of his friends are also supportive of his diet, making the social aspect of living in IV easy.
He also recommends Trader Joes to anybody who wants to keep kosher, as it is the only place nearby with kosher meat.
Like for Jake, Trader Joes is also an important part of keeping kosher in IV.
Most of the groceries I buy are produce (which don’t require kosher certification), meat, fish, bread (which is more expensive), rice, and pasta.
Thankfully (besides produce), Trader Joes has all of this, so if anybody reading this wants to keep kosher, Trader Joes is a great place to start!
Another student, Idan Fierstein, shared his experiences with me.
He grew up in New York City, where there were many kosher restaurants, and now living here is very difficult in that regard (the closest kosher restaurant is over 70 miles away).
As such, he eats out in restaurants (similarly to how I do), but he says that he appreciates kosher meals more than he used to due to the complexities involved with eating kosher in IV.
He hopes that we can get a kosher food cart in IV like they have in SLO.
Eating out kosher style in non-kosher restaurants is something I understand — even if there were kosher restaurants in IV, it would be very hard to stay social especially with how often social plans revolve around a restaurant.
A way I have found to bridge the gap between keeping kosher and staying social is having plans revolve around cooking in my apartment — even though my apartment isn’t kashered, everything I use for cooking is kosher (including hard-to-come-by kosher items like meat and balsamic vinegar), and as of right now, that’s kosher enough for me!
I do try my best to keep kosher, even if I am not following the laws to the dot of how they are in the book.
Others, however, take a more liberal approach to Kashrut.
When asked about his favorite kosher dish to cook, one Max Bloom responded: I’ve gotta go with the cheeseburger.
For those not in the know, cheeseburgers have not been, are not, and will never be kosher, unless you are Max Bloom.
More power to him!
As you can see, keeping Kosher in Isla Vista is no easy feat — the odds are stacked against you, and unless you live exclusively with others that keep kosher, there will be some difficulties.
While not every Jew keeps kosher, those who do feel a strong connection to the food they eat, and are willing to make sacrifices to keep their diet how they wish.
Overall, Kashrut in Isla Vista comes in many colors, shapes, and forms, and is an important part to the Jewish community in IV.
While The 13 Colonies have not legally been colonies for 100s of years, and do not operate the same way they did when they were considered colonies, the effects of colonialism still affect our everyday lives in many ways.
A good example of this is the perception of race.
Within the reading for this week, it is stated that Cherokee identity is socially and politically constructed, stating that much of what and who is considered Cherokee is largely not decided by the Cherokee  (Strum, 2).
These harmful labels affect every ethnic group in America — anybody of European descent is clumped together as white, or anybody of African ancestry is labeled black, even though Africans who migrated here in the past couple of years have completely different culture and struggles than African-Americans whose lineage has been in the United States since the slave trade.
Another example of colonialism has to do with who lives where.
I go to school at UCSB, and although it is a nice institution, 1000 years ago the Euroamericans who constructed this institution did not even know about the land on which this university sits.
This is not exclusive just to UCSB, as according to the textbook, between 1808 and 1835, Cherokees were increasingly confronted with U.
expansion into their territory in the Southeast (Strum, 52).
Both the Chumash people who lived in the land which now holds UCSB, and the land in which the Cherokee people lived is now land owned by wealthy descendants of colonizers, even though in both example the indigenous people were here first.
This is yet another example of how colonialism still affects our lives today.
In conclusion, while the US is not actively expanding its territory, and many colonial practices are illegal (such as slavery), the scars which these activities left on society many years ago are still visible today.
Race labels are oftentimes arbitrary, and don’t reflect the actual experiences of ethnic groups, and yet they are still used today.
Additionally, the land which many of us inhabit belonged to indigenous people for thousands of years before it belonged to us, and yet we are the ones who live there, not the indigenous people who were forcibly removed.
These examples illustrate how the effects of colonization still remain with us today.
My first memory about Indians was from 1st grade, when we were learning about Thanksgiving.
The holiday was introduced to us, and we learned the typical story — Christopher Columbus accidentally discovered North and South America when he tried to get to India by sailing west.
He called the people there Indians (which was the explanation as to why they were called Indians even though they were not from India.
After Europeans learned about these new continents, they wanted to move there.
One of those groups of people from England, the Pilgrims, ended up caught in a storm and it took too long to get to America, and they ended up landing in the wrong place.
By the time they got off the ship, they were all starving, and there was a nice group of Indians there who helped them, and taught them how to keep the land.
They ended up having a great big feast, which is why we celebrate Thanksgiving today.
This narrative went unchallenged until middle school.
We first learned that no, not only was Columbus not the first person to discover the Americas, he wasn’t even the first European to discover the Americas, just the most famous.
Additionally, we were told another important part about him — mainly that he did horrible things to the indigenous people that he encountered.
We also didn’t learn about the fact that both Indigenous people and Indians do not like when Indigenous people are referred to as Indians, and how that diminishes their identity.
Calling them Indians while mentioning that they aren’t actually Indian implies that they are somehow not worthy of having an endonym to refer to them.
Additionally, the story of the first Thanksgiving is a gross misrepresentation of what actually happened.
While the Indigenous people did help the Pilgrims, there was no feast, and the second that the Europeans were able to kick the Indigenous people out in order to colonize the land.
Architecture has been an important part of history, being with the human race ever since the beginning of civilization, and continuing to be an important part of humanity to today.
As society started to progress, architecture started to prioritize form over function; however, recently, this trend has started to reverse.
A prime example of this is Manzanita Village at UCSB, which was designed to house 1100 UCSB students.
(UCSB Campus Housing, 2024).
Being built in 2001, it is relatively recent, and features a minimalistic approach to its housing, using its space to maximize occupants rather than feature complicated designs.
Simple shapes such as rectangles and triangles are used, as these shapes are cheaper and more space efficient than more ornate designs.
In addition, architectural features that provide little to no functionality to the building such as complicated pillars or statues are not used, opting instead to make every architectural piece of Manzanita Village to contribute to the practicality rather than the aesthetics of the site.
Overall, Manzanita Village is a great example of the trend within contemporary architecture to prioritize function over form known as functionalism, saving both money and time for contemporary architectural works.
Manzanita Village’s architectural designs were implemented to maximize space, as having more space per building means that more students can be housed.
Rather than using just wood or steel, Manzanita Village is built out of primarily reinforced concrete, while incorporating wood and steel as a minority materialConcrete as a material has many advantages, such as low cost, durability, and the ability for the rooms to cool themselves at night (Bell and Rand).
Advantages such as those can help maximize occupancy — for example, a low cost means more buildings can be built, and the rooms being able to cool themselves at night removes the need for air conditioning infrastructure to be build, further maximizing space.
Additionally, the buildings of Manzanita Village are built with several buildings in the shape of a ‘U,’ allowing for three buildings to share a communal lawn, rather than all the buildings sharing a lawn.
With one large communal lawn, a significant amount of money would need to be allocated to watering the grass, which is money that cannot be spent on domiciles.
These are examples of functional contemporary architecture, as rather than allocating money towards unnecessary additions such as a complicated cooling system, or a large, luxurious lawn, simplicity and efficiency were paramount while still allowing social functionality.
Additionally, Manzanita Village features a minimal design, opting to use basic shapes rather than intricate designs.
Functionalism is an important part of contemporary architecture, with contemporary architecture rejecting the notion that complicated intricacies are what make beautiful architecture.
Since the industrial revolution, emphasis has been placed on function over form, leading to simpler designs meant to be functional rather than be flashy (Giedion).
The most utilized shape within the construction of Manzanita Village is the rectangle, which is the shape used for the buildings itself, the windows, window shades, and the rooms within.
Triangles are also put to use, as they are the primary shape of the roofs.
These shapes provide important structural integrity, as do more elegant designs, but rather than be complicated, the beauty lies within the simplicity of the design.
When designing this housing complex, intricate designs could have been used, but the designer opted for simple geometry, choosing a functional approach over a flashy approach.
This trend is very common in contemporary architecture, where the beauty comes from simple shapes laid out elegantly (ie a decorated shed), rather than using intricate shapes that are both cost and shape inefficient.
Furthermore, every part of Manzanita Village provides some sort of functionality in addition to visual appeal, which is a defining trait of contemporary architecture.
Many older architectural works contain details such as fountains and statues that do not add any functionality to the building, which is a characteristic which seldom exists within decorated-sheds, which is the type of functional buildings that are used for Manzanita village.
As previously mentioned, Manzanita Village is made up of several buildings for housing, several lawns for students in which students fraternize, and paths connecting the buildings to each other and to the rest of UCSB.
Every part of each individual building, and the village as a whole, contribute to the functionality in some way, such as the roof slant being pointed towards the grass to redirect precipitation to water the crops, or the colored awnings shading the windows as to not let in too much light.
Within older styles of architecture, many ornaments or architectural features may have contributed to the allure of the work, and yet not contribute to the functionality of the constructed object.
Many of the features of Manzanita Village provide visual appeal, and yet they still contribute to the functionality of the housing complex.
As such, Manzanita Village is a great example of the ability of contemporary architecture to provide both visual appeal and functionality to a constructed object.
In conclusion, within the long history of architecture, there have been many trends, and within contemporary architecture, one important trend is functionalism, the minimalist trend of prioritizing functionality over aesthetics (while still taking those into account) within architectural works.
Manzanita Village at UCSB, being built in the 21st century, is a contemporary architectural work that is reminiscent of this functionality of form mindsetManzanita was built in such a way that occupancy (the main focus of the building) was maximized by minimizing cost of materials, and using the environment, not AC, to cool the building.
Additionally, simple geometric shapes were preferred over more complicated shapes that require more space and more money to construct.
Furthermore, every part of the architecture contributes to the functionality of the buildings somehow, with none of the architecture solely being built for aesthetic purposes.
While previous trends in architecture aimed to add ornate details to add beauty, contemporary architecture (such as Manzanita) shows that beauty can exist in solely functional architecture.
Overall, Manzanita Village shows the trend of functionality within contemporary architecture, opting for designs that optimize cost and provide to the building without sacrificing the aesthetic beauty of the building.
Sociolects are a part of everyday usage of language that very few people besides linguists are aware of —the way we speak around others is shaped by the interactions and relations that we have with said people.
Within a romantic relationship, a sociolect forms very quickly, with speakers having mannerisms, slang, and types of language use unique to the relationship.
This sociolect is unique to the couple, as it reflects their shared identity within the relationship, and their history.
Within a relationship, sociolects can be used to relate information faster — a common understanding between two people means that less information needs to be transmitted.
An experiment was performed by Herbert Clark and Deanna Wilkes-Gibbs in their 1986 paper, where two participants had to solve a puzzle involving tangrams without seeing what the other person was doing.
Clark and Wilkes-Gibbs’s paper was the basis for the experiment in this paper.
In the experiment done for this paper, participants are given the same puzzle, and the time taken for them to correctly place each tangram was noted.
The test was given to see if throughout the experiment, the overall time required to determine a tangram decreased (after the initial go where the puzzle was new to both participants), or if the time remained consistent.
The time taken is an indicator for sociolectal development, as a shared sociolect will be used to consolidate dialogue between participants, shortening speech.
For testing, two people in a couple (A and M) were given either a picture with tangram shapes on it in a specific order, or were given tangram pictures individually to sort.
There were three trials, and for each one, the person who gave instructions and who sorted pictures switched — for the first trial for example, A gave instructions and M sorted pictures, and they switched for the second trialAll the puzzle pieces were the same for each one, just in different orientations.
Participants’ views were blocked, so that they could not see what the other person was seeing, forcing them to rely only on vocalizations to communicate how to orient the puzzle pieces.
Participants were told that they were going to be recorded, but were not told for what the recording was being used.
After the trial, the recording was transcribed, and times for each puzzle piece were noted for each trialWhile times for the first pass were noted, they were not used in the results, as the purpose of the experiment was to evaluate if further development occurred between turn two and turn three.
The results of the aforementioned experiment were as follows: between trials two and three, the average time taken for the describer to correctly explain to the matcher the location of each tangram more than halved.
Additionally, the maximum time taken for a single picture also decreased between trial two and three (0:37 vs 0:26).
Throughout all trials, the matcher always correctly placed every single tangram.
While in trial 2, several of the shapes had been given impromptu names from trial 1 such as eroding IV cliff, by trial 3, most tangrams had been given names and/or identities that were shared among both participants (there was little to no confusion from the matcher as to whom was being referred by the describer).
In conclusion, the findings of this experiment show that even after having had seen and heard all pictures described once, further development of the sociolect of A and M occurred.
It is already known that the number of turns after each trial gets lowered (Clark and Wilkes-Gibbs), showing that this affect exists within the general population, but after conducting this experiment, it is shown that already fleshed out sociolect (between a couple) can still be developed due to new circumstances.
This supports the hypothesis that sociolects continuously evolve, as a non-evolving, already developed sociolect would likely have very similar average times for pictures, especially after all the pictures had been described once and the participants were comfortable with both the pictures and the experiment.
Although these are interesting results, there is a major limitation with this study, as only evaluating one couple is much too small a sample size to have conclusive data.
That being said, a similar experiment that should be performed would be to evaluate how strong this effect is on multiple different types of relationships (ie have co-workers, a boss and employee, a couple, and two random strangers), and see which one’s average time decreased the most between trial two and trial three.
Overall, these findings show that sociolects, even in individuals whom are well acquainted, continuously evolve over time due to new circumstances and conversation topics.
The memorial I will be constructing will be in Agoura Hills, CA, and will memorialize everybody who unfortunately lost their lives in the Woolsey Fire during late 2018.
Specifically, this memorial will be in Chumash Park, as this is a very frequented area with Agoura Hills (it is right next to Agoura High School, and many events, such as the Great Race of Agoura take place here).
It will be built out of wood and concrete, and it will be approx 4ft x 2ft x 7ft.
The bottom 3 feet will be the base, whereas the top four feet will be the statue itself.
The concrete will be smooth, and there will be a small notice on the bottom that says Please do not touch.
The monument will be a concrete statue, the left half is a heart design, while the right half will an anatomical heart, representing the physical life and emotional connections that were lost respectively.
Below that, there will be a placard with all of the names of those who lost their lives.
The names on the placard will be in the following format: In loving memory of: as the title, followed by: (Birth Year) - 2018.
The wood holding up the memorial will be light birch wood, and the placard will be black with golden writing (there will be braille below each line as well), with a golden border.
The platform on which the statue will be and that contains the placard will be a trapezoid, with 3 sides being rectangular, and the last (the one that contains the placard) being slanted.
The monument will be about 250 ft away from Argos street, to ensure that drivers passing by don’t look at the monument too intently, as to cause an accident, and there will be a concrete path leading to the monument itself.
There are many memorials in the world, most of them focusing on either just one person, or an enormous amount of people (think wars, terrorist attacks).
My memorial will fit somewhere in the middle, focusing on those who lost their lives during the Woolsey fire.
The message of this memorial is two fold — to honor those who lost their lives, and to educate those who see it about the dangers of fire and fire prevention.
This location is a strong location for the memorial, as it covers the area affected by the fire.
In addition, the form would be a great way to convey the meaning of the monument, as the statue in conjunction with the placard will convey the importance of the memorialIn addition, this monument serves as a counter-memorial, as it serves to remind people about the lives that were lost due to the negligence of Southern California Edison.
My monument is going to convey two messages: first, this monument is going to act as a memorial for all of those who died tragically during the Woolsey fire.
While many people know about the Woolsey fire and that people died during it, having the names written on a placard will remind people that families got destroyed and souls who should be with us are no more.
While the more recent LA fires may have been more destructive, the Woolsey Fire impacted Agoura Hills specifically, hence the choice to commemorate that event.
The second message that this is going to convey is a warning about the dangers of fire.
This fire was caused because Southern California Edison (a power company) failed to shut off the power during extraordinary high winds, causing a major fire in SoCal’s history.
In addition, many people chose to reside in their dwellings despite evacuation notices, causing them to unfortunately pass away.
This monument will remind people of both of those, that they should be appreciative and not angry when the power company shuts off the power during high winds, and that listening to evacuation notices, no matter the circumstances, is crucial and lifesaving.
In addition, this monument is going to be located at Chumash Park because that location serves as a meeting point of the community in which it resides, and also happens to lay in the wake of where the fire passed.
Within this park, many events take place, and besides organized events, this park serves as a wonderful picnic spot for many within the community.
Because this is such a frequented area, having the monument located here allows many more within the community view the monument, furthering the spread of the message of the monument.
Additionally, Agoura Hills was one of the locations that was evacuated during the Woolsey fire, hence why it should be located here.
Many people within the community know somebody who lost their home during the fire, therefore having this monument here can serve as a reminder to those people of the importance of preventing fires.
Additionally, constructing the monument with the aforementioned design is a great way to convey the meaning of the memorial, as the placard serves to honor those who have passed, with the statue further reenforcing this meaning.
Although on a conceptual level, everybody knows that people have passed, many people (especially after so much time) have forgotten what that truly means.
Having the names of the people who passed during the Woolsey fire serves as an important reminder that those who we lost were people with names, families, and interests, and that they will never get to experience what life has to offer anymore.
Additionally, the imagery of the statue serves as a reminder to what was actually lost: the anatomical heart shows that people who were once living are no longer living, and the heart design represents the emotional connections and love that were lost during this horrific fire.
Furthermore, this monument serves as a counter-memorial rather than a memorial, as it both serves to commemorate those who were lost, while acknowledging the negligence of the power-companies.
A memorial is defined as something that celebrates life and glorifies an event, whereas a counter-memorial is defined as a monument that recognize darker events and serve to teach, rather than glorify (Kerby et al).
The purpose of this monument is to teach about the dangers of fire, and the importance of power companies shutting off the electricity due to high winds.
If this monument were to be constructed as a memorial, it would glorify rather than villainize Southern California Edison.
Since the purpose of this monument is to hopefully prevent such a disaster from ever occurring again, it is important that this monument take the form of a counter-memorial rather than a memorialIn conclusion, the monument which I planned servers as a way to remember those who regrettably passed during the Woolsey fire, to teach about fire safety, and to inform about the role that SCE played in this disaster.
The selected location, Chumash Park in Agoura Hills, CA, acts as a frequented locale that exists within the affected areas of the fire, reminding residents of Agoura Hills and visitors alike the importance of fire safety.
Additionally, the design of the monument reminds viewers that there were people with real names that died during this fire, and connections they had with those around them will never be able to be replaced.
Furthermore, this monument being a counter-memorial serves to explain the role of SCE in this disaster, as having the monument be a standard memorial would not hold them responsible.
This monument is an important way to teach about the dangers of fire, and commemorate those who were lost.
The Salem witch trials had taken place in 1692, only thirty five years earlier.
Identify five details describing the setting that suggest something sinister and supernaturalThe reason that Dritto does not explain anything is to get the reader to want to read more of this passage.
Dritto not explaining anything is a hook because it doesn’t give away any plot details, and makes the reader curious as to what will happen next.
Uora-Uora was put on lookout duty because he was well dressed and would look less suspicious than anybody else.
Baby had not eaten a piece of cake in a long while, perhaps even since the war.
He felt a desire so strong to eat those cakes that he could not focus on the task on hand, and he also feared the inevitable end of this cake feasting extravaganza, whether that be due to getting caught or because of becoming full.
The manner in which they were eating the cake was very animalistic and non-human-like, which is why the author chose to use animal similies for this animal-like feast.
GluttonyEating Cake represents sexual intercourse.
Licking cream has strong sexual connotations, and the manner and place in which this occurred (on a bed) gives the impression that something sexual, or sex itself, occured.
We feel sorry for her, and Katie pities her.
NoHe invites them because he is attracted to Katie.
Yes, as he likes Katie more, and therefore would like Martha to be more like her.
Furthermore, being with Martha would radically simplify things for Martin.
Martha puts up with Martin because she has a small child, but she cries when Jenny has her first period because she realizes that Jenny isn’t young anymore.
The reason that Dritto does not explain anything is to get the reader to want to read more of this passage.
The features on computers and the internet that make for what the world is today are ease of use and speed.
Paired with a good operating system, anyone can use a computer, phone, or any computer powered device easily.
Having something that is fast means you can get work done quickly and efficiently, without having to wait (although the internet in its early days suffered this fate).
These two inventions have changed the way my life is, from entertainment, to how I am writing this assignment.
Ordinary folks were the dreamers, the people who sought change and had brilliant minds and ideas.
Isaacson states that the great connectors are usually the inventors, and their inventions.
A CEO might invent something that makes themselves a lot of money, but a dreamer will invent something that changes the world.
The difference between these two works is how they frame social media as a tool.
While Isaacson states that technology will change the world and revolutionize people, Gladwell thinks that since people still had revolutionary ideas and spread them without Twitter, that social media will play little to no role in politics.
The candles, the darkroom, the pit, the pendulum, the robed judges, and the compacting walls.
The pit symbolizes the fear of falling in, the pendulum symbolizes that time is running out for the main character, who has little time to escape his destined fate.
Even though the protagonist is still alive when the rats crawl over him, these rats symbolize that the man has already reached a certain end, and time was just ticking for when it would arrive, however, he was eventually saved.
Poe is condemning the overpowered nature of the Spanish inquisition, and how someone's life could be completely upturned just because of their religion.
He paints the inquisition to be bad, as he wants to warn people of what would happen if people were discriminated against by religion.
Yes, and I don’t like Disney-esque ‘happy endings.
’ This ruined the story for be because it made the message seem less valuable, if all of that happened, just for none of it to matter because he was saved, for some reason.
He is not in a good state of mind and he is slowly devolving into insanity.
The entire story shows his rapid demise, and this further proves it.
He made them twins as it would further contrast them; with nothing different between them, any slight difference is that much more noticeable.
The way he acts changes throughout the entire story, and this is proof enlight that he is going crazy.
I think that he is hallucinating, and his insanity is slowly killing him.
There is no way he should be seeing her, but he does because his mind is deteriorating.
The title Some Dreamers of the Golden Dream embodies Lucille’s struggle for happiness.
This struggle occurs because of her distance from her husband, and the pain and shame she dealt with for being a cheater, and a murderer.
While other people lived happy lives: The Golden Dream, she did not.
I think that hearing Arthwell’s story puts perspective on what it is like to be in a contributor's shoes.
His story shows that he got a lot of shame, and guilt, and although many people think he didn’t suffer, I know he did.
Lucille’s Golden Dream was to live a happy life with a man she loved.
This never happened, which caused this drama.
Writing a non-fiction story often has less bias, as one does not slather their opinion throughout the work.
This changed how people read this story, because they’re opinions were not swayed by the author.
He offers to support his judgment of considering his efforts a failure because he realizes that his efforts were not successful, and he should not pretend that they are.
He uses the phrase self-annihilation to express the process of creating a mistake that makes you worse, and everybody around you worse.
The significance of the quote is that it represents who he is.
Tizon looks up to his father, as any son does.
This is evident throughout the essay, and his choice to mainly include his fathers examples does a good job at explaining this, and Tizon himself struggled too if he was close enough to his father to have this many stories.
This also works to show the difficulties of being an immigrant (in a figurative and/or literal sense), and not fitting in.
Whiteness affects both Authors by making their culture feel different from a perceived American ‘Normal’.
There is, however, no normal lifestyle, as people's lives vary greatly.
The approach taken differers from this essay to that essay in that this essay is much more cynical while that essay was much more open.
This provides detail, because it justifies his point, and the way he says it makes one want to read more of this written work.
He now means that instead of somebody getting labeled on what they say, they get labeled on the platform they say it.
I disagree with this, as people criticize whatever the other side is, even if they are using the same platform as themselves.
He states this to emphasize his point that tweeting behind a screen isn’t put yourself in any harm, and is almost cowardly.
He describes high risk activism as any form of activism that puts you, your reputation, and your integrity at risk.
The only risk you put yourself at when constantly sitting behind your computer 24/7 and tweeting is the risk of a heart attack.
I think this point was to show that the author does think that social media is good, and has its use cases; however, trying to be an activist solely on social media won’t make the cut.
I agree with him here.
He makes the distinction in order to establish that being motivated by a cause, and actually participating in it are two separate occurrences altogether.
The distinction between the two is that the motivation is something that most people can get behind, while the participation is oftentimes more limited.
His first example is of MLK.
I think that his choice of this example is very smart, as many people know about it, and know that he simply preached, he never incited violence, and most importantly, he never twittered.
Another example was Ivanna, which shows that social media is powerful, but that you shouldn’t use it for protests.
Activism:He grants the benefit of helping people in need.
I think that saying digital evangelism is a little extreme, and it really is just a hub.
He uses modern example to show that people very similar to us didn’t need the internet to make their points go through.
I think they are similar because they both use simple language to understand, and showed how people could be judgemental, and helpful.
The narrator wants to get validation from the couple as a way to make him feel like he is living a better life than he is.
He is crippled, and not living a great life, but wants to think he is extremely important in the grand scheme of things.
Helping him will shine a light on the fact that he is crippled and will make him feel worse about his current condition.
The wife does not want the illusion that her husband sees to be broken.
Hubert means that he exists only behind the scenes and does not want to be front and center as to upset the man.
The author is shining light on the fact thaHaving no names included makes the reader try to fill in gaps about these people, which engages the reader further.
This rhetorical choice is a smart strategy because it ensures that the reader pays attention.
She would do that to try and understand Hubert’s perspective of life, and in turn, felt herself shift into that.
This embodies greed.
A White Russian is an alcoholic beverage.
Alcohol is addictive, and people can be greedy to obtain alcohol.
The gray zone she speaks of are the Muslims who are genuinely good people, and also happen to be Muslim.
These people are not terrorists, but often get clumped together with them, just because of their race.
The gray zone not only resists hate and terror, something that the minority of Muslims who are terrorists feast on, but also the notion that they are terrorists.
The group’s religious message is that they believe that the only way to worship Allah is to kill any non-believers and try to scare those who can be changed into preaching Islam the same way they do.
This parallels what most Americans think terrorists are: destructive religious extremists.
It can be inferred that Lalami is emphasizing the point that a notable chunk of the American population believes that anybody who is Muslim is automatically a terrorist, and Muslims get treated unfairly because of that.
Her choosing to put this essay as a personal narrative helps the points get through because this essay appeals to pathos.
Both essays share the fact that they want to ensure that their daughter(s) will know that their life will not be normal, as women who are in a minority.
This emphasizes that judging people based on their religion or ethnicity doesn’t affect just the terrorists who happen to be of that faith, it doesn’t just affect adults who have to deal with this unwarranted prejudice, but also children, who will have to grow up like this.
He was feeling extreme anger and harsh resentment.
This anger kindled him.
This shows the innocence of a child.
This is important to understanding the plot and everything about the story.
Deb, who was once put into this awkward situation, gained someone who trusted her and felt comfortable around her.
The narrator’s anger ended up hurting their relationship with Deb.
She points out the fact that these people were taken from their land, and their kids were not allowed to go back.
This was continued, for an extremely long time.
She uses pathos to make readers feel as if they are her and experiencing this racism themselves.
These tactics make the reader feel resentment towards racists, and her explanation of the microaggressions makes the police, and the average citizen seems unknowing and ignorant to the issues that these people face.
She conveys this message this way to show that there is hope, and another way, ie to move away.
The strategy that is employed most is pathos, which is used to tie examples together and make the reader be put in the shoes of those who were forced to deal with these horrific racial incidents.
She makes tragedy seem sad, while he makes it feel bad.
This difference in tone makes both pieces have very different overall feels, while this one was one that made you feel bad for the African Americans, his made you want to do something about the issues.
He loves gazing at La Grande Bretèche for he loves the beauty it possesses.
This, unfortunately, is frowned upon by everyone else as he has trespassed there.
Furthermore, the late owner of the establishment forbade anybody from entering.
He wants to explain the severity of his passion for making sure that nobody enters the manor.
This provides the reader with insight as to the nature of this house, and the lore that surrounds it.
She did, as she felt guilty about how the events, or rather, the tragedy that occurred impacted the Spaniard.
Although cheating is highly immoral, suffering that fate because of cheating is a bit extreme.
Even though she was implicated in his death, she was never tried nor found guilty.
This story takes place in France, and the allusion to the biggest city in that country can be an easy place to compare from.
Using a well-known city with well-known tropes as a tool for comparison is an effective rhetorical device because it gives the reader context.
She is too afraid that her husband will get mad, and call her a sinner for lying under the oath of a cross.
She believes that the household became cursed after her negligent decision-making processes were tested, and wants to ensure that the household is no longer cursed.
She probably also wants the rotting smell of a dead body to not be in the house, for obvious reasons.
The word victim is used in the opening sentence as Staples is trying to show that occurrences like the one mentioned at the beginning of the story are so common that he has resorted to satirizing them in order to cope with them.
A reader might interpret the first sentence as a shock, but after reading further, one will understand why Staples decided to use such wording.
He shifts his point of view when the story is explained from the perspective of the victim, to show his point of view and his intentions.
His purpose in doing so is to show the reader that he is not doing anything wrong and that he is getting blamed for the color of his skin.
Staples behaves in a mature and safe manner, and although annoyed by the woman’s fear, does not react to it in any way that would escalate the situation.
This is similar to Muñoz’s essay, by the fact that people, especially minorities, are just supposed to accept what somebody who is alike in every way except for race and/or gender are going to define the status quo for them, and they will get no say.
I do not think this is true, as, for years, many teachers of mine have tried to get me to enjoy reading.
I have started to enjoy it more in my high school years, but I was always inspired to read something at any age.
I cannot say for certain if others have the same experience as I do, but I assume that most people are similar to me, as I was never extraordinary, nor was I falling behind.
She uses such strong language to get a reaction out of those who agree with her.
This technique, while effective, will turn off non-like-minded individuals from reading what was written.
I found that she also puts her opinion front and center, rather than making this an unbiased piece.
This, once again, will attract those who agree and detract those who do not agree with her.
Prose means that by limiting the reader’s exposure to other sources in the class, you are stifling their imagination, and that constriction leads to them forming an unnuanced opinion.
I tend to disagree, as students aren’t trapped inside of a small room, and can see the outside world, and form their own opinions from that, not just from what is taught in the classroom.
She is implying that everyone in the classroom forms their opinions based on only what is taught in the classroom.
They loathe it because it feels forced and unnecessary.
From my personal experience, I find: She suggests reading more books and using fewer lesson plans.
The origin of the name is based on the main character, who is a dwarf who is often mocked.
The title is ironic because he was the one who got the last laugh, as he engulfs the cabinet in flames and murders all of them.
The only name that the main character gets called is Hop-Frog, as the king, and everybody else did not take him seriously whatsoever.
This reveals that he was often thought of as unimportant, which can be very taxing on somebody mentally, which may cause them to act in such a manner as he did in the end.
The author, Poe, assumes that the royalty suddenly changes their stance on Hop-Frog from being a joke to being somebody for whom they could ask.
Poe portrays Trippetta as being short, similar to Hop-Frog.
This is significant as it forms a rhetorical connection between Trippetta and Hop-Frog.
Furthermore, this connection stretches into her actions, as when the king tries to get Hop-Frog to become extremely drunk, Trippetta protests, showing that, unlike the king, she actually has heart.
Since Trippetta is held in high esteem by the king, one might assume that she is like-minded; however, she ends up going against the king.
The irony in the story is that Hop-Frog, who was oftentimes mocked and harassed by the king, ended up becoming victorious and got revenge on the king for all of his misdoings to him.
The chains and the fire symbolize how the king and his men treated Hop-Frog: with disrespect and fuego.
The phrasing basal beasts paints the men out to be relentless, unforgiving, greedy men.
The characters engage in a lot of fantasy.
They are not happy with whom they are with, and have lustful fantasies about other people.
YesThe narrator is a victim of their own doing.
Yes, never having enough is definitely a characteristic of gluttony.
Hubert and his wife like to live a so-called ‘eloquent’ life, and having each other’s way minimizes discourse that could get in the way of that dream.
No, he is irresponsible and a bad parental figure.
Women in society are often seen only as mothers.
If staying together benefits the child more than it hurts the parents, then yes, otherwise, no.
He feels as if Deb is an impostor.
Society weighs heavily towards couples staying together, even if it negatively affects somebody or both people in the relationship.
Men, as traditionally, were the ones to ask out the women, and the ones to propose.
Balzac tells his story from an outsider’s perspective as that would be the most relevant vantage point for the audience.
These are intermingled in the story through the main character’s strong desire to visit a forbidden place, even if somebody specifically requested not to.
Even though he was no longer creating music, his death made his popularity rise, as the death of such an influential figure made waves in the music community, causing many people to discover his music.
We then moved about a mile away, up a hill and across our street, there is a hiking trail.
I live in my house with my brother, mom, dad, and my cute dog, Suki.
I went to Willow, from kindergarten to 5th grade, I was at the GATE program at A.
Wright in 6th grade, and of course, in 7th grade, I attend Lindero.
I think the turning point in the poem is line 20: But yet sufficient for us leftLine 45: With glory richly furnished.
People today are much more materialistic and less intrinsic as the past, so we would probably take longer to get over the loss while being less affected by it.
The house is nature, it is made by God (hence the capitalized Him), and it is better because to her because it was a gift.
In comparison to what she now has, what she once had is gross and unwanted.
What she wrote was lost.
This was her life, this was her.
Losing this was like losing everything about you.
She had nothing left of what she had written.
She doesn't want to move and does not want to accept defeat.
He said that he had sold the house and he suggests that everyone should just take the money.
He would essentially be accepting defeat.
White and black people should have the same opportunities, and by giving up hope, he is pushing that away.
Walter tells Linder that he isn’t going to move, and Linder appeals to mama in an attempt to reason with her.
She gives the same response that Walter had given to Linder.
The importance is that everything is coming full circle and she has accepted that it is time to move on and live life.
His arrival changes the mood because it prompts Bertha to remove her headdress, revealing a nearly bald head, and all of those who were around her were appalled to see her head in this condition.
Walter looks down upon George and thinks that he is lesser in class and in intelligence compared to himself.
She has put a down payment on the house.
Ruth rejoices while Walter feels despair and anger.
George wants to marry somebody who is simple yet intelligent, and he refuses to listen to her ideas because he thinks that he is more important than her.
She helps end the argument and stands behind her daughter.
She gets called, and he was wandering into clubs instead of working.
He envisions a future where his family lives in a utopia with only happiness and smiles to go around.
They aren’t white and the neighborhood around them does not appreciate their presence.
He tells him no and tells him to leave.
Gardening tools, as she was at the time attending to her plant.
His money had been stolen, and he was angry.
The apartment itself is very small, unfurnished, and old.
These lead to the living conditions for the Younger family to be sub par at best, and horrid at worst.
Ruth is reluctant to give money because she does not see the value of spending money unless it is absolutely necessary, while Walter gives the dollar because he understands the importance of this to Travis.
Walter thinks that women should employ roles in society traditionally held by women, and should act womanly.
Bertha wants to study medicine, which goes against Walters views on a woman’s role in society.
She takes her plant, which symbolizes her nature as one who cares and doesn’t take, unlike everyone else.
Ruth returns from the doctor, for whom had told her she was pregnant.
He brings traditional Nigerian clothing and music as gifts for her.
He made Ruth so mad that she locked herself in the bathroom, and he carelessly spends their money.
Mama influenced him by telling him the truth, that Ruth is pregnant.
He was the only person to vote innocent right off the bat, wanting to give the defendant a fair chance.
He feels for the defendant and understands that he had a rough life and sometimes a second chance will make you better.
He knew exactly what to say in order to try to sway some of the other jurors.
He is grumpy about how long the trial is, even before it starts.
He was abusive, even to his own son.
He says that he would be happy to perform the execution.
"""
```

```python
textb = """
Hi, my name is Jacob Rosen, and I just want to say that it is very crazy, that it is the time to culminate out of 5th grade.
It feels like on the very first day of kindergarten, I started out with an empty tool belt.
Every lesson, grade, thing I learned, and test felt like adding a brand new tool into my belt.
I have so many tools in my belt, I could build a bridge.
That bridge will carry me to middle school, and with all of these tools, I am very prepared.
Willow has taught me so much, and I have gone through kindergarten, 5th grade, and every grade in between.
The bridge I am going to build is huge, but Willow gave me the tools to walk across the bridge.
It all began when I first walked into the kindergarten classroom.
So much has changed since that day, academically and not.
I barely knew anything, I didn’t have braces or glasses, and I wasn’t very mature.
I was very lucky, I came to Willow the final year there was a kinder campout, and the last year of reciting the Finding Dracula poem too.
But we are talking about kindergarten, not about how lucky I was.
The kindergarten campout taught me a lot, like how to be brave in situations you have never been in before.
With that tool in my toolkit, I will use it to be brave enough to build that bridge, and cross the gap into middle school.
Starting a new experience, when I went to 1st grade, I got to go to the big yard, with the 2nd graders and the giant 3rd graders, which taught me about kids who were older than me.
What gave me an even better tool was Turkey-Tea.
Turkey-Tea is when the 1st graders go up on stage around Thanksgiving time, and sing some songs.
That was one of the first times I ever went on stage.
When you are that young, it takes a lot of guts to do that.
I guess that gave me a tool of perseverance, which lets me keep building the bridge, even when it is embarrassing, or not going well.
Surprisingly, 2nd grade taught me a very big lesson.
I was in Fersht’s class, and we were reading a book called Strega Nona.
Now I don’t remember much about the book, but what I do remember is that there was a character named Big Anthony.
One thing I remember about Big Anthony is that he went to a magic pot, said a spell to make pasta, said the spell to stop it, but what he did not know was that he had to give the pot 3 kisses to stop it.
Eventually, Strega Nona came back and stopped the pasta from coming out.
But Big Anthony had to eat around 5 tons of spaghetti, and you could guess what happened.
So the tool is to follow directions, and don’t be a Big Anthony, I repeat Don’t Be A Big Anthony!
Next up is 3rd grade.
3rd grade has a very big event, Ellis Island.
Ellis Island is when you go into the M.P.R. and dress up and pretend you are a different person.
The person you pretend to be is an immigrant from another country.
You have to get through security, not be sick, and you can buy things.
You can also write letters to your family back in your old country.
That gave me an important tool, to be thankful for where I live, and that I am lucky to be born into the family I have, and not live in someone else’s shoes.
Now we will talk about 4th grade.
Probably the biggest thing you do in 4th grade is the 4th Grade Fiesta.
In the 4th Grade Fiesta, you sing songs, dance, and eat Mexican food(Yummy Yummy)!
Most 4th graders would probably agree that they did not like the dancing.
I definitely did not either.
The tool I got from that is that even if you don’t like something, endure through it and you will get rewards(Ex. The end of the line is 5th grade).
Most kids would probably talk about Outdoor Ed, but I don’t really get what tool you would get from that.
Surviving in the wild?
Learning about the outdoors?
I will talk about the State Report instead.
In this report, you have to learn about a state, write interesting facts on a trifold poster, and memorize a speech you made.
It is hard work to glue everything onto the trifold, find the facts, make it look nice, and create and memorize a 3 minute speech.
But if you put a lot of work into it, it will look amazing.
That is a good tool.
Finish everything first, and then you have time to do what you want.
Thank you siblings and parents and aunts and uncles and grandparents or whatever relationship you are to your culminating 5th graders, for coming out, changing your schedules, working around your everyday lives, and thank you teachers for helping us get the tools to build a bridge to middle school, and I hope you enjoy the rest of your day.
5th graders, yay!
We’re culminating!
The bear who lived in the forest heard a loud and strange call.
Who could it be, he thought to himself.
He realized it was the Jade Emperor.
Calling to all the animals in the forest.
The year has shifted in an unknown way, and the Zodiac is ruined.
We will have to do it again, or else there will be no time.
We will add a new animal, the Bear.
The bear celebrated that he was chosen.
He was the special one.
The next day, the bear arrived at the starting line.
The ice on the ground shined like the stars.
You could feel the cold breeze.
You could see the bear zoom off at the start.
The race had begun.
Suddenly, the horse caught up with the bear, and the bear attacked the horse.
Nobody passes me, the bear howled at the fallen horse.
The bear was as vicious as a black widow.
Yelled the horse.
A little farther back, the ox saw the fallen horse, and helped him up.
He asked who did it, and the horse said the bear.
The ox sprinted up to the newly formed zodiac, and said, Not cool.
I am stopping you.
The bear rammed into him, and that is when everything changed.
Why did you bite me?
If that’s what you want to do, here is a taste of your medicine, The bear bit into the ox’s leg, and blood poured out like water from a river, and the bear’s teeth, like razors, pierced the poor ox.
As the raven once quoted, the bear said, Nevermore.
You will not pass me now.
As the bear gracefully swam through the dangerous river, and sprinted toward the finish line, a flash appeared in front of him.
Blending in with his brown fur, the rat jumped off the bear’s back, and beat him with the same trick used on the ox that was now bleeding, thousands of years ago.
With all the viciousness of the bear, and the insanity of the race, the bear was tricked at the end,.
He was announced as 2nd place.
To fit in, wear the same clothes that the other girls are wearing.
Hang out with girls, and go play hopscotch.
You should not hang out with the boys and run around.
Native Americans use natural resources in their everyday lives.
My chosen tribe was the Apache, who live in the Southwest USA.
Today, the Apache would live in Arizona and New Mexico, which are very warm.
In this report, you will learn about the Apache’s traditions, food, tools, clothing, shelter, and 3 facts.
Natural resources are very important for Apache food.
Food is necessary for survival, and they get it through hunting and gathering.
One way the Apache get food is by hunting.
Buffalo is one animal they hunt.
People also hunt other animals though.
The Apache also gather acorns and berries, which are very important staples.
They can cook inside their wikiups and their tepees.
The Apache made great use of their food resources.
The Apache built their homes using natural materials.
One type of home was a wikiup.
A wikiup was also known as a wigwam.
The wikiup had a dome shape.
Another type of home was a tepee.
A tepee was a temporary structure.
A tepee was also lightweight.
The Apache were nomads, which means they had no permanent home, and nomads usually traveled in search of food.
That is how they used natural resources to build homes, why they were nomads, and what their homes were like.
The Apache also used natural resources to make clothing and artifacts.
Apache clothing was sometimes made of leather.
The shoes the Apache wore were called moccasins.
Tools were very helpful in Apache life.
One tool they used was a travois, which was used by nomads.
They also used horses, which were important because they helped with transportation.
The weapons they used included arrowheads and a bow and arrow.
These tools were important for everyday life, hunting, and travel.
My tribe had many interesting customs, traditions, and artifacts.
One tradition they had was eating roasted agave.
Agave comes from cacti.
Agave was a food they enjoyed a lot.
Apache women stayed at home.
They made artifacts.
One type of artifact was sewn baskets.
The men were warriors.
Finally, the extended family stayed together.
These facts make the Apache interesting to learn about.
I am about to share 3 interesting facts about the Apache.
When an Apache woman marries a man, the man joins the woman’s family.
Also, the Apache word for corn is maize.
Finally, a famous Apache was Geronimo.
The Apache had unique customs and facts about them.
All Native American tribes use the natural resources around them in their everyday lives.
The Apache used natural resources in many ways, have an interesting culture, and live interesting lives.
There are many things to learn about the Apache.
They are a special tribe with strong characteristics.
You hear the bell ringing.
You just got out of school.
It is a cold winter day, and you hear the pitter patter of rain outside.
When you get home, you want to play your new FIFA game, and drink hot chocolate.
You sit on the couch, and right before the Xbox turns on, you realize you have two hours of homework.
You just spent an hour learning something in class, and chances are you will remember it.
How could you possibly forget it?
Homework eats up free time that could be spent with family, wastes time, and cuts into sleep because kids stay up late doing it.
It also raises stress levels for students.
Homework should be banned.
It seems strange that after learning something for an hour, schools expect students to forget it, and so they assign homework to “remember” it.
According to apa.org, homework barely makes a difference.
Test scores don’t improve much with or without homework (APA).
This means someone could spend hours on homework to maybe get a 96% on a test, when doing no homework might get them a 95.
It’s not a few minutes of homework, it’s a huge amount.
A PBS writer said the average student spends 81 minutes on nightly homework.
That’s 1 hour and 21 minutes.
John Buell from the Huffington Post explained that some research even shows homework can make grades worse (Buell).
One hour and twenty-one minutes is a massive chunk of time.
Someone could be spending that time with family instead.
If you multiply that over 180 school days, it’s 180 x 81 minutes, which equals 14,580 minutes, or 243 hours.
243 hours could be spent with family.
In 2017, everyone is so busy that homework takes up almost all free time.
It’s a lot of work without making a real difference.
An extra hour of homework might only earn one more question correct.
Homework wastes time, both personal and family time.
Scholastic pointed out that homework steals family time just to meet arbitrary school rules, not because it’s truly needed for learning.
This means free time is grabbed by homework.
Time is scarce, and homework eats it all.
Kristen Weir from the New York Times said homework takes up a lot of time, and that increasing amounts of it can negatively affect grades.
Peter Drucker from the Huffington Post called homework a waste of time, arguing students could use their time better.
Homework often doesn’t help grades, and sometimes it makes them worse.
Most people know how stressful it is.
Studies show homework stresses students more than no homework does.
John Buell from Alternet said if homework were a prescription drug, the FDA would have banned it because it is so unhealthy.
The Huffington Post wrote that too much schoolwork and homework causes stress and anxiety.
Stress comes from teachers giving too much homework.
Some argue homework teaches time management, independence, and helps schoolwork (e-skool), but studies show homework has mostly negative effects.
Few positives, massive negatives.
It steals free time.
Memphis Parent wrote that students might learn time management, but they don’t get enough time overall.
Those claiming homework is amazing are often organizations like Common Core or companies making standardized tests.
It’s like the CEO of McDonald’s saying fast food is healthy—just promoting profit.
Evidence shows homework is largely useless.
It wastes free time, isn’t necessary, and stresses students.
That’s why homework should be eliminated.
It’s an unnecessary form of torture that causes discomfort.
Without homework, life would be better.
Greetings, Hello, nice to meet you.
My name is Walle.
I am from Timbuktu and I am 49 years old.
If you don’t know, Timbuktu was the capital of Mansa Musa’s Mali Kingdom.
I am a brewer; I make beer and create drinks for people.
It’s very enjoyable.
I taught myself to be a brewer after retiring from being a ferba because it sounded more fun.
I live in the brewery, which is pretty big.
When I am not working, I love to cook.
I mostly eat salted meat and vegetables.
I also pray to Allah, as I am Muslim.
In my society, men and women have mostly equal rights.
Our ruler, Mansa Musa, has complete power and is currently the wealthiest man alive.
The Hajj is an important event, and I went on it with Mansa Musa.
I met Mansa Musa when I was a government official, but fifteen years ago I switched to brewing.
He treated me well except for the time I overslept, which made him very angry.
Thanks for reading my letter.
I hope you learned something about my culture.
Hope to see you at the Faire.
Sincerely, Walle.
In Counting by 7s, the character who changes the most is Dell Duke, the counselor who doesn’t do his job.
At the start, Dell didn’t run because he wasn’t in shape, had no desire, and was a slob who didn’t appreciate the world.
By the end, Dell runs and gets in shape.
When he met Willow, he was unathletic and uninspired.
He didn’t need to run because he had no desire to change.
Dell also didn’t know how to pace himself when running.
He might sprint five seconds and get tired fast.
When he ran for fourteen minutes, he looked like he might pass out and probably didn’t run a full mile.
Next, Dell was lazy and ate only microwavable meatloaf.
He had no real furniture and just used plastic chairs.
He saw nobody, so he didn’t care about appearances.
When Pattie Nguyen opened the closet, they saw a tower of unwashed boxers.
He only ate meatloaf because he could.
Nobody watched him, so he ate whatever he wanted.
Lastly, he didn’t appreciate the world, just laying around and ignoring others.
When he met Willow, and her parents died, he began to see the world differently.
He left Cheddar at the start of the book, but at the end, he rescued him.
Dell became a much different man.
He isn’t a slob anymore because of the Nguyens, and is healthier, no longer eating just meatloaf.
He also sees the world differently.
He would have remained a slob without Willow, but he changed.
I wore a fancy outfit.
My parents told me to.
I wanted a Neff shirt and athletic pants.
Why?
I wanted to make a good impression, and I liked the shirt.
My parents said I had to “look good” on the first day of middle school.
But really, my outfit would have looked fine.
Willow didn’t cheat for several reasons.
She probably didn’t know about the test, or any sources.
To test if she cheats, give her a new random test.
If she scores 100 on a new test, she didn’t cheat.
If below 70%, then maybe she did.
Describe a topic or concept you find so fascinating it makes you lose track of time.
Why does it captivate you?
Who or what do you turn to when learning more?
August 21, 2019, 11:15 a.m.: I stepped into Spanish 1 for the first time.
Like my friends, I wasn’t excited at first.
Later, I realized I really enjoyed Spanish and learning how languages are structured.
I started participating more than my peers.
I read extra material and watched Spanish shows to expand vocabulary.
Even during Covid-19, I made progress and my passion continues.
I plan to study Spanish and other languages in college and beyond.
I joined language learning forums and had interesting conversations with native speakers.
Studying languages inspires me; I hope to become a polyglot.
Localization—adapting content for another audience—is something I want to explore.
I am not yet proficient enough to localize, but I have read and watched materials on it.
I find it fascinating to translate jokes and puns, like antelopes eating cantaloupes.
I’ve spent hundreds of hours reading linguistics and studying grammar.
A few years ago, I thought languages were just English with different words.
Learning Finnish challenged me with its 15 grammatical cases.
I haven’t mastered it yet, but I’m proud of my progress.
Though some think high school language classes are boring, I enjoyed them and am thrilled to be in Spanish 4 Honors.
Language learning has helped me connect with people and sparked a love of learning beyond Spanish class.
Change is slow, but extreme events can force immediate change.
In 20,000 Leagues Under the Sea, Pierre Aronnax and his crew are taken hostage.
Aronnax is a famous explorer, and Nemo wants his expertise.
They are thrown overboard and trapped on the Nautilus.
At first, Aronnax is intrigued, but horror sets in when he realizes he can’t leave.
Aronnax begins eager and adventurous but ends repulsed by the Nautilus.
Trauma shifts his perspective from excitement to fear.
He is invited on the Abraham Lincoln to hunt a sea monster.
He and his companions, Conseil and Ned Land, attempt to stop it but fail.
They jump overboard and land on a submarine.
Still excited at first, Aronnax tries to reason with Nemo in multiple languages.
They learn they will travel but never leave the Nautilus.
Aronnax is fascinated by the ship’s organ, equipment, and library.
He and his companions are explorers, so this excites them.
Repeated trauma changes even the strongest perspectives.
Nemo takes them to the South Pole, where they become trapped by an iceberg.
Limited food, water, and air create extreme stress.
Aronnax experiences pain and wants to escape.
Other dangerous events continue, like the giant squid, which kills a crewmate.
These traumas erode Aronnax’s desire to explore.
Each near-death experience makes him want to leave the Nautilus.
The final event is a massive whirlpool, frightening everyone.
Nemo panics, plays haunting chords, and shouts.
Aronnax, Conseil, and Ned escape on a dinghy and land on an island.
They are no longer adventurers; stress and trauma have changed them.
Impactful events can rapidly change people.
Aronnax, once ambitious, risks everything to escape.
He rejects exploration after these events.
This mirrors real-life change, like 9/11 impacting security and society.
Generational changes shape humanity, just as events shape individuals.
Luke Skywalker accepts responsibility when R2D2 runs away.
He also accepts responsibility after his family dies, becoming a Jedi.
He saves Princess Leia, knowing it’s his duty.
Luke finds self-motivation by returning to Obi-Wan and wanting to become a Jedi.
He could have given up, but he persevered.
Han Solo demonstrates self-management, knowing what to do and when.
His skill helps him survive dangerous situations like with Greedo and the Death Star.
Luke, Han, and Chewbacca show interdependence when they work together to reach the control room.
Without teamwork, they wouldn’t succeed.
Another example of interdependence occurs when Chewy, Luke, and Han work together to rescue the princess by pretending that Chewy is their captive and that they are taking him to a cell.
Coincidentally, that cell happens to be the princess’s, and they help her escape.
Han Solo demonstrates emotional intelligence when he remains unfazed by Greedo pointing a gun at him and demanding money he doesn’t have.
He shows remarkable control over his emotions, remaining calm under stress.
Even when stormtroopers attack, he does not panic.
Instead, he boards the Millennium Falcon and instructs Chewbacca to fly them out of Mos Eisley.
Luke Skywalker shows self-belief when Obi-Wan places a helmet over his head and he must defend himself against a blaster ball with a lightsaber.
Had he doubted himself, he would have been struck.
Instead, he trusts both himself and the Force, allowing him to act confidently.
After R2D2 and C3PO exit the escape pod, C3PO resists following R2D2, claiming he is choosing the wrong path.
Eventually, C3PO ends up in a less optimal location and blames R2D2, despite it being C3PO’s own failure to listen.
Ever wondered about black holes?
Of course you have!
They are one of the most intriguing topics in space for non-astronomers.
The fact that they remain theoretical only increases the curiosity.
Now, there is speculation not just about black holes, but their theoretical opposites.
White holes are considered the counterparts to black holes.
Like black holes, approaching a white hole would warp spacetime.
The force of a white hole is so intense that anything nearby could be destroyed.
White holes are theorized to form from the rotation of black holes.
If a white hole connects to a black hole, it could create a wormhole.
Though theoretical, the idea of white holes has existed for over fifty years.
In 1964, Russian cosmologist Igor Novikov proposed the concept as part of a solution to Einstein’s field equations, specifically for an eternal, non-rotating, uncharged black hole (White Holes Wikipedia).
To explore this concept further, Brit Bendavid and Jacob Rosen of Space Corp. are producing a documentary on white holes.
The short documentary is in pre-production and slated for release on Friday, April 28, 2017.
Although it will be a direct-to-video release, it can still be watched at home.
It will not appear in IMAX, but copies will be sold by Space Corp.
Their address is 29276 Circle Drive, Malibou Lake, California, 91301.
Copies can also be ordered by phone at (278) 766-6691.
The cost is $19.99 or a three-day rental for $4.
Two central questions drive the book: who the ghost is and who killed King Hamlet.
The multiple appearances of the ghost throughout the first act, including its final appearance in the act, highlight this mystery.
Since returning from Wittenberg, I have witnessed remarkable events.
I was first asked to investigate a ghost, which I doubted.
To my amazement, the ghost was real.
Afterward, the state of Denmark became increasingly chaotic.
King Claudius announced he would marry his late brother’s widow, causing turmoil among us Danes.
Poor Hamlet fell into despair, seeing his life as futile, though he later declares he is fine and plans to act in ways that may appear mad.
This period has been the wildest of my life!
Highland climates are found in elevated mountainous regions.
They do not experience traditional seasons.
Any seasonal effects are mostly noticeable at lower elevations.
Mountains have distinct climates because temperature changes with elevation.
At the base, it might be 80 degrees and sunny; higher up, it becomes colder and rainy.
Climbing further, snow and freezing temperatures can appear.
On average, temperature drops about three degrees for every 1,000 feet of elevation.
Precipitation in highlands also varies with altitude.
Some mountains have dry bases while their peaks are snow-covered.
This occurs because high elevations force warm air upward, cooling it and creating precipitation.
Vegetation in highlands changes with elevation.
At the base, plant life mirrors the surrounding region.
On a rainforest mountain, the base would support rainforest species.
Higher elevations support plants adapted to colder climates, eventually giving way to barren areas.
Mountains often feature a tree line, marking the limit of tree growth.
Tree lines vary depending on temperature, soil, and moisture.
Animals in highlands also vary with elevation and surrounding ecosystems.
Mountain goats are common in the Rockies, with feet adapted to steep terrain.
Large cats like mountain lions and snow leopards inhabit these regions.
Snow leopards have broad feet and thick tails for balance in mountainous terrain.
Highland climates range from cool to cold, depending on altitude.
Temperature changes rapidly with elevation.
The highland climate mirrors surrounding biomes but is usually cooler and wetter.
Recycling was once widely practiced, turning waste into new products and benefiting society.
Today, this practice has declined, especially since China stopped importing U.S. trash due to health and economic concerns.
Previously, U.S. ships traveled empty to China, wasting fuel and money.
When China purchased U.S. waste for recycling, it created economic benefits.
Over time, however, China no longer needed the excess plastic, causing more problems than it solved.
The end of U.S.-China recycling has increased landfill waste and environmental harm.
A domestic recycling solution could have mitigated this issue, but dependency on China caused the collapse of green plastic initiatives.
Society’s future depends on finding ways to reuse materials sustainably.
Tax incentives for recycling could encourage the use of eco-friendly plastics.
Change is inevitable, but its direction depends on our choices.
Jem matures throughout the book, evolving from a rebellious boy into a more responsible young man.
Early on, he resists rules and disrespects authority, acting selfishly.
For example, Dill dares Jem to touch the Radley house.
Instead of showing maturity, Jem rushes to the house, slaps it, and runs back.
Despite warnings about Boo Radley, he acts recklessly to impress Dill.
His actions reveal his immaturity, though fortunately, nothing tragic occurs.
Jem changes positively over time.
Change requires external influence.
Jem adapts to his environment, learning right from wrong.
He stands up for what is right, as when confronting Mrs. Dubose’s racist comments about Atticus.
He cuts her hedges to demonstrate that her views are unacceptable (Lee, 135).
This was a reactive decision, driven by his growing sense of justice.
He believes in his father and acts accordingly.
Jem’s attitude reflects his father’s stress during the case, showing environmental influence on behavior.
Many ancient Greek inventions remain in use today, including democracy, coins, and boats.
Greek democracy relied on an assembly voting on laws and leaders.
Today, we vote for Democratic or Republican candidates.
Modern government consists of three branches: Legislative, Judicial, and Executive.
The president and vice president belong to the executive branch.
The legislature (Congress) writes laws, while the judiciary (Supreme Court) interprets them.
Ancient Greek coins included Talent, Mina, and Drachma.
Modern currency includes coins and paper bills in various denominations.
Ancient boats relied on sails and oars, while modern boats use motors or steam engines.
Wooden boats were handcrafted; today’s boats use metal, strong plastics, or alloys.
Ancient boats had small capacities; modern cruise ships feature theaters, pools, restaurants, and suites.
Gravity affects objects even when they are not falling.
Gravity’s strength depends on an object’s mass and the gravitational environment.
Weight is the force of gravity on mass, measured by the compression of a scale.
An African American student introduces himself to connect with the audience, establishing ethos.
He uses pathos, logos, and ethos, including anecdotes and allusions.
Shifts in pronouns convey a sense of disconnection experienced by African Americans.
The use of the n-slur emphasizes the passage, even at the cost of formality.
Felix and Antonio, the Amigo Brothers, walk together discussing their upcoming fight.
They separate to mentally prepare and ensure fairness.
Felix trains and watches a boxing movie, avoiding thoughts of Antonio.
Antonio trains and visits his aunt, avoiding gangs along the way.
On fight day, they greet each other and enter the ring.
They initially fight with caution, which escalates in round two.
By the final rounds, they expend every ounce of energy.
The announcer struggles with multiple dialects while declaring the winner.
The Amigo Brothers leave the ring, never knowing the official result.
Charlie aims to rebuild his relationship with his daughter.
He leaves to spend time with her and convince Honoria of his worth as a father.
He regrets abandoning his family after his wife’s indiscretion at a party.
Marion directs her anger and resentment toward him.
Charlie avoids seeing himself as weak or bad, idealizing Helen in his mind for comfort.
Doing good differs from doing well; the former prioritizes positive impact, the latter prioritizes profit.
Mattel diversifying Barbie dolls promotes inclusivity and healthier beauty standards.
This topic is better suited for daily coverage than monthly, as its impact is gradual.
Both authors criticize cookie-cutter representations in toys, highlighting potential harm.
Diversifying Barbies mitigates these harms, making dolls accessible to more children.
The tone is serious, reflecting the importance of the issue.
The voice conveys pain and personal struggle.
An omniscient third-person POV could provide similar plausibility but might shift perspective to the oppressors.
Hitchen shows interrogators’ sympathy despite the simulated torture, humanizing them.
Although simulated, participants likely felt real distress.
Hachiya conveys grimace through tone, while Hitchen conveys suffering.
Both use detailed and serious language; Hachiya evokes sadness, Hitchen anger.
Hachiya illustrates racial discrimination, though figurative language risks diminishing credibility.
Addressing his son emphasizes multigenerational, systemic issues.
Mentioning victims highlights the harsh realities African Americans face.
Coates assumes schools are adversarial, but supportive institutions exist.
The educational system is flawed, affecting all students.
The West’s resilience reflects a culture of standing up for beliefs.
Narrative detail shows a love for nature and a critical view of physical attraction.
Homer’s softness and aversion to hunting represent perceived weakness.
The narrator realizes her true love is Homer, not physical allure.
She is dissatisfied with the outcome of unfolding events.
Heidegger’s water causes harm due to immature, greedy behavior of those consuming it.
The story explores human traits like gluttony, lust, and greed.
The rejuvenating flower symbolizes the quest for youth.
Guests witness the consequences of greed firsthand.
Coronavirus disrupted life, forcing adaptation.
Distance learning allowed flexible schedules and more free time, though social interaction suffered.
Students adapted using digital communication, maintaining some social connection.
Overall, distance learning proved effective despite social limitations.
Sentiment analysis evaluates whether text expresses positive, negative, or neutral sentiment.
This paper applies a BiRNN to Wikipedia revisions, examining shifts before and after major events.
Political issues experienced sharp negative sentiment shifts.
Online language reflects real-world emotions, amplified by connectivity.
Wikipedia’s frequent updates allow tracking of sentiment changes.
Articles aim for neutrality but often display bias (Greenstein & Zhu, 2012).
Evaluating revisions reveals language trends and reactions to events (Ferron & Massa, 2011; Keegan et al., 2013).
Research questions: extent of sentiment change, universality across topics, and speed of change.
Manual review is infeasible; computational models are required.
A BiRNN trained on Twitter data tracked sentence sentiment over time.
Despite domain differences, qualitative tests showed the model performed adequately on Wikipedia.
Sentiment analysis uses context-aware BiRNNs rather than simple cosine similarity.
Language intake affects perception and sentiment (Mutz et al., 1997).
The dataset contained positive, negative, neutral, and irrelevant tweets; irrelevant samples were removed.
Remaining data: 22,542 negative, 20,832 positive, 18,318 neutral samples.
Sequences were padded at batch time; no additional preprocessing was applied.
The BiRNN had two hidden layers of size 48 per direction and two feedforward layers (128 and 64 features).
Dropout was applied at 0.3 in BiRNN and 0.5 in feedforward layers.
Training used cross-entropy loss and Adam optimizer (learning rate 10^-4).
Batches of 32 were trained over 144 epochs, achieving 0.26 training loss and 0.84 F1 score on validation.
Topics analyzed included Charli XCX, Dave Grohl, Russia, Ukraine, Zionism, and apples as a control.
Date ranges were chosen to capture sufficient edits before and after events.
Graphs averaged revisions and mapped data points to the date of the last revision in each batch.
Control (apple) showed stable sentiment, except for short etymology sections.
Russia showed gradual negative sentiment decline in the main contents, and sharp decline in government/politics.
Ukraine showed stable or slightly positive sentiment, except for short sections.
Zionism displayed declines in history and fluctuating beliefs; changes remained stagnant after September 2024.
Dave Grohl’s article remained stable, with minor dips due to formatting.
Charli XCX experienced spikes related to heritage and racism sections.
Significant sentiment changes usually returned to prior levels, except for Ukraine’s main contents and late dips in Zionism.
Political topics experienced the most sentiment shifts, whereas pop culture was less affected.
The BiRNN successfully captured context, though training on Twitter may introduce minor domain mismatch.
Future research could train models directly on Wikipedia data.
Wikipedia remains reliable despite emotional writing, as accuracy and sentiment are distinct.
Political topics tend to have more emotional edits, which Wikipedia moderation generally stabilizes.
Caution is advised when interpreting politically charged content, though overall reliability persists.
Summer is approaching, the weather is pleasant, and you are still eating plain instant ramen.
Here are two simple, tasty recipes you can try.
They are healthy, delicious, and far better than instant ramen.
Recipe 1: Poke Bowls (serves 3)
Norwegian farmed salmon is safe to eat raw due to parasite-free feed and often being flash-frozen.
Recipe 2: Hearty Pasta Salad (serves 6)
Many students eat plain pasta; turning it into a salad adds flavor and nutrition.
With these two recipes, summer becomes more enjoyable and cooking easier.
My artistic perspective emphasizes struggle—adapting to life in the U.S., preserving culture, and resisting oppression.
The acting style for this production should generally be loud and emotionally charged, but at times it should also be restrained.
Themes such as violence and abuse should be represented through a clear power dynamic, with the abuser shown as strong and dominating, and the abused as small and powerless.
A good example of this is when the two men grab the young girl.
During this moment, lighting should cast shadows that make the men appear larger than life, while the girl struggles and screams without full strength, symbolizing her lack of power.
This method of expressing struggle through contrast can be applied throughout the production.
For instance, the background can reflect Medea’s difficulty assimilating into American culture, with American elements of a Victorian household looking neglected and dull, while Mexican elements are colorful and vibrant, representing her perspective on life in America.
This contrast can also extend to costume design.
For example, when Medea kills Acan, his American clothing can appear faded and dull, with the only color being the blood.
This emphasizes both Medea’s struggle to adapt and the loss of Acan.
The contrast can be mirrored in the music as well, with lively, uptempo Mexican music playing when Medea speaks, switching to slow, somber American music when Acan speaks.
Overall, my artistic vision emphasizes struggle through contrast in acting, set design, costumes, and music.
The production should be accessible to a wide audience, but its message is particularly timely.
In the current climate of hostility toward immigrants, it is important to show the significance of maintaining cultural identity, assimilating, and accepting both.
This tension drives the play, and in a time of growing intolerance, the production serves as a reminder of the importance of empathy and acceptance.
I want the audience to understand the consequences of a lack of kindness and understanding.
The tragedy of this play, including the deaths of characters like Acan and Armida, illustrates the harm caused by cruelty and indifference.
Many people today fail to treat others with empathy, and the production should reflect the outcomes of such behavior.
I hope the audience leaves feeling compassion for the characters, particularly those who suffered or died.
This emphasis on kindness and understanding directly reinforces the theme of struggle, showing what happens when characters are deprived of it.
Armida, though morally complex, is still a victim, as is Acan, whose death highlights the consequences of approaching cultural conflict without compassion.
My central visual motif is the border between the United States and Mexico.
This symbolizes both the cultural transition in Medea’s life and the trauma of crossing the border.
I chose this image because I am fascinated by man-made structures and their psychological impact on people.
The motif extends to costume design, with those affected by the harsh border crossing, like Medea, wearing dusty, worn clothing.
Lighting should be bright to reflect the full daylight at the border, but shadows should fall over characters experiencing hardship.
This reinforces the artistic goal of emphasizing struggle and contrasting the experiences of different characters.
The border scene can be directly represented in the set, showing its immense scale and highlighting the difficulty of combining Mexican and American identities for Medea.
Sound design should support this contrast, reducing urban background noise during moments of conflict to evoke the quiet tension of the border.
In these moments, natural sounds such as wind or birds can underscore both the physical and emotional atmosphere.
These design choices highlight differences between characters, illustrating how these differences cause struggle and reinforcing the broader theme of contrast.
I plan to use design elements to construct sets that accentuate differences and conflict.
Lighting will enhance contrast between bright and dark areas, and characters in conflict will be positioned to visually emphasize the power dynamic, such as elevating the stronger character.
Line can guide positioning and create symbolic backdrops, like the uneven lines of a border fence, reflecting disorder in the characters’ lives.
Shape will be used in shadows and props to control visual focus.
Form will emphasize three-dimensional objects in the foreground, drawing attention to important elements.
Value, contrast, and color will be employed boldly, with vibrant, lively elements contrasting with dark, muted ones to represent struggle.
Texture will primarily appear in costumes, showing worn or battered clothing for those enduring hardship.
Scale and proportion will visually diminish characters facing oppression, placing them lower or further from the audience.
Motion will reinforce contrast, with constant movement during active scenes and stillness during quieter moments.
Emphasis will be achieved through form and lighting.
Balance will intentionally be uneven, particularly in conflict scenes.
Rhythm will be irregular to illustrate chaos in the characters’ lives.
Space will prioritize important elements closer to the audience and less important elements further away.
Unity will maintain slight chaos, escalating during conflict.
Most production decisions align with my script analysis.
The primary adjustment I plan to make involves spacing, as intense movement and close proximity between characters in conflict may complicate choreography.
I may reduce emphasis on motion and highlight other contrasts instead.
Over the past few years, the Jewish community has experienced both challenges and growth.
One positive development is the creation of the student-led Jewish magazine, Fig and Vine.
Many community members contributed excellent articles this year.
One contributor, Shani Levy-Richards, graduated this year.
I interviewed her to share her experiences with the community.
I asked Shani how she felt being a Jewish student at UCSB.
She said she enjoyed it, largely due to the size of the Jewish community.
UCSB has the largest Jewish community among UC schools, creating a strong and active community.
She also appreciated the various events hosted by Hillel and Chabad.
These events were engaging, fun, and helped her meet friends.
Some of her favorite memories included Pesach at Hillel and other Chabad events.
She said UCSB’s community felt more welcoming than other UC campuses.
I asked her why she joined Fig and Vine.
She said her interest in Jewish history drew her to contribute.
She enjoys exploring Jewish history, as there is so much to learn and document.
She compared writing for Fig and Vine to keeping a diary.
Shani feels more confident expressing herself in writing than in speaking.
Writing empowers her, unlike traditional homework.
I asked about her post-college plans.
She plans to take a gap year before pursuing a master’s in English or history.
Although she has left UCSB, her twin sister remains, so she plans to visit.
I asked how the Jewish community has evolved at UCSB.
She noted that events such as October 7th and rising antisemitism have strengthened the community.
The community has grown tighter to survive challenges.
She wished everyone good luck and thanked the community.
Shani has contributed significantly and will be missed.
We hope she enjoys graduate school and returns to Santa Barbara soon.
Throughout history, many indigenous groups have faced government oppression.
When oppressed, these communities sometimes rebel, occasionally gaining independence.
The Zapatistas in Chiapas, Mexico, are an example.
They faced diminished land rights and racist policies.
These injustices led them to revolt on the day NAFTA was signed.
They briefly occupied six towns before retreating to the jungle.
The Mexican government responded with military force, violence, and massacres.
The government sought to appear strong and capable internationally.
Despite peace efforts, the underlying issues remain unresolved.
The Zapatistas demand free elections, recognition as a belligerent force, an end to neoliberal policies, and general elections for Chiapas.
Their demands stem from historical abuses by the Spanish and Mexican governments.
Indigenous communities in Chiapas have been excluded from government decision-making and basic services.
Land seizures were often disguised as dispute resolutions between groups.
These injustices fueled the Zapatista rebellion.
The conflict escalated on January 1, 1994, when Mexico joined NAFTA.
The Mexican government sought global approval, while the Zapatistas demanded independence.
The rebellion continues today due to unmet demands.
The conflict illustrates the consequences of government oppression.
Similar struggles exist worldwide, including Guatemala and the United States.
Indigenous groups often fight for their rights over land, water, and resources.
Persistence in resistance has historically led to positive outcomes, even if delayed.
The Zapatista situation exemplifies the ongoing need to respect and address indigenous rights.
Rabbinic literature differs from the Hebrew Bible in several ways.
Revelation in the Hebrew Bible occurred through prophets, while in Rabbinic literature, it is transmitted through study and interpretation.
Two key features are makhloket and safek.
Makhloket refers to disagreements among Rabbis, and safek refers to uncertainty.
The emergence of Torah scholars enabled interpretation, not just understanding.
The Hebrew Bible’s vagueness often caused disagreements in interpretation.
The Tannaim and Amoraim had different approaches to Torah scholars.
Tannaim emphasized understanding the Torah, while Amoraim emphasized debate and interpretation.
Tannaim viewed Torah study as secondary to a job, while Amoraim treated it as a profession.
Examples include gathering manna and the story of Rav Shimon Bar Yochai.
The Tannaim said manna should be gathered daily, while the Amoraim allowed preparation for Shabbat.
Rav Shimon Bar Yochai valued Torah knowledge over physical health.
Makhloket preserves minority opinions for future legal flexibility.
Modern examples include the Supreme Court’s treatment of minority opinions.
Jewish law evolved significantly over time.
Initially, it was a teaching tool, then became strict and statutory during the Second Temple period.
Sectarian divisions emerged, such as the Pharisees and Sadducees.
After the Second Temple, Rabbinic interpretation allowed flexibility.
The Mishnah and Gemara preserved multiple views to guide future courts.
The evolution of law reflects a shift from pedagogy to strict statute, and finally to interpretive flexibility.
Multiple perspectives are often valid, showing that life is not strictly black and white.
Journal 5: I think that blood quantum are a very important topic to be discussed.
I did this journal entry before we watch the video about the Indigenous people in Canada, and how they were not treated as a part of their community, and I believe that that is horrible, and the culture that you grow up in determines your ethnicity and what community you belong to, not your genetics.
Additionally, oftentimes these blood quantum are only a problem for women who intermarry rather than men, not just making society more xenophobic but also more misogynistic.
Overall this is an important issue that I hope more people will start to discuss.
Journal 6: I think assuming that western education, and the systems that surround and govern it are the default and nest way to educate is a very dangerous assumption to make, and a myth that needs to be distilled.
For example, Finland has some of the highest test scores in the world, and they spend significantly less time learning.
Additionally, many forms of western education are competitive rather than collaborative, which can lead to heightened anxiety and an overall worse outcome for science as compared to those who are not stressed.
It is important that these myths get distilled, and that we all learn the truth.
Journal 7: This journal was speaking about how certain indigenous communities in Latin America have been fighting for their rights, in regards specifically to their culture.
This is a very important part of learning about how indigenous people get affected by the state — indigenous people are not idle ducks — they will fight a gains unfair laws and orders put upon them.
It is very important to remember that learning about these fights against tyrannical laws are just as important as learning about the tyrannical laws themselves — having information about both helps us better understand the context of ever indigenous struggle in the world.
Journal 8: I really enjoyed this talk — I am a very big dan of music (I have been playing piano since age 5, I love listening to music and making music), so I can appreciate the importance of music to both individuals and to a larger culture.
I enjoyed learning about all the ways that the indigenous communities in Oaxaca preserved their culture, but the music part specifically stood out to me for the aforementioned reason.
Overall, my main takeaway from this week’s journal entry is that there are many important facets to one’s culture, and finding ways to preserve them is very important (such as with the music).
Journal 9: I really resonated with this journal entry because I really care about nature and I do my best to be in harmony with it.
I love going on hikes with my friends, and loving nature for what it is and has to offer shows me that the oftentimes overly urbanized capitalist society does not appreciate where we came from.
I always do my best to keep nature clean and safe — there only is one Earth upon which we live, and taking care of it is very important.
That being said, the Earth (as discussed in the reading) can benefit sometimes from being taken from, so it is important to be a part of the planet and to love the planet.
Journal 10: I believe that state intervention is a major problem when it comes to indigenous issues.
Although this is not necessarily about indigenous rights, what has been currently going on with ICE and the Trump administration is a good demonstration about the problems with state intervention.
By the state not listening to the demands of the protestors and instead sending in the ICE and making the situation more dangerous for everybody, they end up aggravating people and showing people that this aforementioned kind of intervention (rather than listening) is a major problem in our society today, in regards to the state.
AbstractIntroduction.
Patient evaluation and diagnosis is important in cases of aphasia to provide good treatment and support.
A YouTube video of an aphasic is watched and a diagnosis is provided from said video.
A transcription was made for the video, and it was found that the patient had difficulty forming long sentences, and oftentimes fell back on familiar language.
It was determined that the patient in question has Broca’s Aphasia.
Correctly diagnosing and treating Aphasias is important to help the patient, and give closure to them and their family.
Learning about aphasias and the type of symptoms that they produce is important as it can help patients receive the care they need, and help those around them understand what is happening and why.
Additionally, since aphasics can get better over time, it is important to diagnose it so that they be in an environment conducive to brain recovery, such as language rehabilitation.
While many people might rush to diagnose the aphasia quickly, getting language rehabilitation at any point will help about the same amount (Moss & Nicholas, 2006).
One of the major types of aphasia, Broca’s Aphasia, is caused by damage in BA 44 and 45 and patients often have issues with speaking fluently, finding the correct words, articulating well, repetition, and the production of complex sentences (Dronkers etal, 2007).
Because other language tasks, such as semantics, and comprehension, occur in other parts of the brain, even though they are closely intertwined with tasks associated with Broca’s Aphasia, they are not affected in Broca’s Aphasics (Schwartz etal, 2009).
Case StudyThe person being discussed is an elderly woman, Jan, who is having a discussion with her daughter, Carly, on a YouTube channel they share dedicated to informing people about aphasia.
The specific video is about how to have a good conversation with an aphasiac.
ResultsWithin the conversation, she had difficulty creating complex sentences, and saying the name of a number (although she could write it down).
Additionally, her daughter (who also is her caretaker) already had a sheet on hand with many activities on it, further indicating that she had problems expressing her thoughts.
Jan additionally had multiple phonemic paraphasias.
Patient also extensively used the word yeah, as this was the only interjection used by Jan (at least during this recorded snippet).
Additionally, almost every utterance she said was very short, lending to the fact that it was difficult for her to form long utterances.
DiagnosisBecause Jan’s speech was at times telegraphic, and she had word finding difficulty, the diagnosis that is given is Broca’s Aphasia.
Jan showed a high level of comprehension (ruling out Wernicke’s Aphasia and Global Aphasia).
There were also many signs of Broca’s Aphasia, such as a common word or phase being often used (in this case yeah), and the knowledge of a number, but not being able to say the name of it.
DiscussionAs previously discussed, the diagnosis and understanding of diseases like Broca’s Aphasia are important as they can help get patients the best treatment that they deserve and give closure to both them and those close to the patients.
Since certain aphasias are caused by certain areas of the brain (Schwartz etal, 2009), being able to know what is and what is not affected for patients like Jan is essential because treating her like a Wernicke’s Aphasiac will not help her in any meaningful way, and would be mentally taxing for her.
Giving her the correct treatment could, as previously discussed, bring back language functionality to her brain (Moss & Nicholas, 2006), further stressing the importance of correctly identifying and treating aphasias.
Additionally, knowing a rough idea of the area in the brain in which the the stroke occurred is helpful because it can predict other symptoms of the stroke.
Since Broca’s area is near BA 44 and 45 (Dronkers etal, 2007), other related processes that occur in the frontal lobe nearby will likely be affected.
Overall, correctly diagnosing and treating aphasias is vital to the efficacy of short term care of patients, the potentiality of long term restoration of language skills, and the closure that the patient and their family deserves.
Judaism and food have a long and complex relationship — we have entire holidays dedicated to abstaining from eating.
We have certain foods we can’t eat, and certain foods we must eat.
And finally, lets not forget, gefilte fish.
Jewish culinary culture is formed by these very laws, making Kashrut (kosher laws) an important facet of our food and our culture.
Growing up, I kept no form of kosher, and up to going to school here, I continued this trend.
If you were to ask me what my all time favorite food was, I would have enthusiastically said shrimp!
Having come to college, I formed a deep connection with my Judaism, and in turn, with Kashrut.
While I might not be fully orthodox kosher (I will still eat at non-kosher restaurants as a pescatarian with a shellfish allergy), I do keep a somewhat strict degree of Kashrut.
I may be a Jew in IV who keeps kosher, but I am not the only Jew in IV that does, so I wanted to see what others’ experiences with Kashrut in IV and at UCSB were.
The first person to whom I posed the question of Kashrut in IV was Nicole Zar.
She told me that she grew up keeping kosher, and that this was important to her, so she kept keeping kosher in college.
To her, it is totally doable to keep kosher here, despite the many obstacles.
She said that she often relies on Chabad for meals (especially on Shabbat).
Although it is not simple, she told me, it is a very rewarding part of Judaism that [she] takes pride in and a mitzvah that [she] enjoys keeping!
As Nicole said, UCSB Chabad besides being a keystone to the Jewish community is also one of the best sources of Kosher food in Isla Vista.
Miri Klein, the Rebbetzin at UCSB Chabad, is the mastermind behind all that is food at UCSB Chabad and works tirelessly to ensure that Jewish mouths in IV are fed.
She told me that the biggest challenge for providing so many people food is that there are many difficulties getting large quantities of food — either the foods need to be bought from a kosher market, or comes in very small quantities.
She suggests that anyone who wants to start keeping kosher contact her with any questions, and go to Trader Joes.
I was able to get another perspective from Jake Nguyen, another student who keeps Kosher.
He said that personally, he has not had to make any concessions with Kashrut when coming here, and that the only downside to a kosher diet is that it is more expensive.
All of his friends are also supportive of his diet, making the social aspect of living in IV easy.
He also recommends Trader Joes to anybody who wants to keep kosher, as it is the only place nearby with kosher meat.
Like for Jake, Trader Joes is also an important part of keeping kosher in IV.
Most of the groceries I buy are produce (which don’t require kosher certification), meat, fish, bread (which is more expensive), rice, and pasta.
Thankfully (besides produce), Trader Joes has all of this, so if anybody reading this wants to keep kosher, Trader Joes is a great place to start!
Another student, Idan Fierstein, shared his experiences with me.
He grew up in New York City, where there were many kosher restaurants, and now living here is very difficult in that regard (the closest kosher restaurant is over 70 miles away).
As such, he eats out in restaurants (similarly to how I do), but he says that he appreciates kosher meals more than he used to due to the complexities involved with eating kosher in IV.
He hopes that we can get a kosher food cart in IV like they have in SLO.
Eating out kosher style in non-kosher restaurants is something I understand — even if there were kosher restaurants in IV, it would be very hard to stay social especially with how often social plans revolve around a restaurant.
A way I have found to bridge the gap between keeping kosher and staying social is having plans revolve around cooking in my apartment — even though my apartment isn’t kashered, everything I use for cooking is kosher (including hard-to-come-by kosher items like meat and balsamic vinegar), and as of right now, that’s kosher enough for me!
I do try my best to keep kosher, even if I am not following the laws to the dot of how they are in the book.
Others, however, take a more liberal approach to Kashrut.
When asked about his favorite kosher dish to cook, one Max Bloom responded: I’ve gotta go with the cheeseburger.
For those not in the know, cheeseburgers have not been, are not, and will never be kosher, unless you are Max Bloom.
More power to him!
As you can see, keeping Kosher in Isla Vista is no easy feat — the odds are stacked against you, and unless you live exclusively with others that keep kosher, there will be some difficulties.
While not every Jew keeps kosher, those who do feel a strong connection to the food they eat, and are willing to make sacrifices to keep their diet how they wish.
Overall, Kashrut in Isla Vista comes in many colors, shapes, and forms, and is an important part to the Jewish community in IV.
While The 13 Colonies have not legally been colonies for 100s of years, and do not operate the same way they did when they were considered colonies, the effects of colonialism still affect our everyday lives in many ways.
A good example of this is the perception of race.
Within the reading for this week, it is stated that Cherokee identity is socially and politically constructed, stating that much of what and who is considered Cherokee is largely not decided by the Cherokee  (Strum, 2).
These harmful labels affect every ethnic group in America — anybody of European descent is clumped together as white, or anybody of African ancestry is labeled black, even though Africans who migrated here in the past couple of years have completely different culture and struggles than African-Americans whose lineage has been in the United States since the slave trade.
Another example of colonialism has to do with who lives where.
I go to school at UCSB, and although it is a nice institution, 1000 years ago the Euroamericans who constructed this institution did not even know about the land on which this university sits.
This is not exclusive just to UCSB, as according to the textbook, between 1808 and 1835, Cherokees were increasingly confronted with U.
expansion into their territory in the Southeast (Strum, 52).
Both the Chumash people who lived in the land which now holds UCSB, and the land in which the Cherokee people lived is now land owned by wealthy descendants of colonizers, even though in both example the indigenous people were here first.
This is yet another example of how colonialism still affects our lives today.
In conclusion, while the US is not actively expanding its territory, and many colonial practices are illegal (such as slavery), the scars which these activities left on society many years ago are still visible today.
Race labels are oftentimes arbitrary, and don’t reflect the actual experiences of ethnic groups, and yet they are still used today.
Additionally, the land which many of us inhabit belonged to indigenous people for thousands of years before it belonged to us, and yet we are the ones who live there, not the indigenous people who were forcibly removed.
These examples illustrate how the effects of colonization still remain with us today.
My first memory about Indians was from 1st grade, when we were learning about Thanksgiving.
The holiday was introduced to us, and we learned the typical story — Christopher Columbus accidentally discovered North and South America when he tried to get to India by sailing west.
He called the people there Indians (which was the explanation as to why they were called Indians even though they were not from India).
After Europeans learned about these new continents, they wanted to move there.
One of those groups of people from England, the Pilgrims, ended up caught in a storm and it took too long to get to America, and they ended up landing in the wrong place.
By the time they got off the ship, they were all starving, and there was a nice group of Indians there who helped them, and taught them how to keep the land.
They ended up having a great big feast, which is why we celebrate Thanksgiving today.
This narrative went unchallenged until middle school.
We first learned that no, not only was Columbus not the first person to discover the Americas, he wasn’t even the first European to discover the Americas, just the most famous.
Additionally, we were told another important part about him — mainly that he did horrible things to the indigenous people that he encountered.
We also didn’t learn about the fact that both Indigenous people and Indians do not like when Indigenous people are referred to as Indians, and how that diminishes their identity.
Calling them Indians while mentioning that they aren’t actually Indian implies that they are somehow not worthy of having an endonym to refer to them.
Additionally, the story of the first Thanksgiving is a gross misrepresentation of what actually happened.
While the Indigenous people did help the Pilgrims, there was no feast, and the second that the Europeans were able to kick the Indigenous people out in order to colonize the land.
Architecture has been an important part of history, being with the human race ever since the beginning of civilization, and continuing to be an important part of humanity to today.
As society started to progress, architecture started to prioritize form over function; however, recently, this trend has started to reverse.
A prime example of this is Manzanita Village at UCSB, which was designed to house 1100 UCSB students.
(UCSB Campus Housing, 2024).
Being built in 2001, it is relatively recent, and features a minimalistic approach to its housing, using its space to maximize occupants rather than feature complicated designs.
Simple shapes such as rectangles and triangles are used, as these shapes are cheaper and more space efficient than more ornate designs.
In addition, architectural features that provide little to no functionality to the building such as complicated pillars or statues are not used, opting instead to make every architectural piece of Manzanita Village to contribute to the practicality rather than the aesthetics of the site.
Overall, Manzanita Village is a great example of the trend within contemporary architecture to prioritize function over form known as functionalism, saving both money and time for contemporary architectural works.
Manzanita Village’s architectural designs were implemented to maximize space, as having more space per building means that more students can be housed.
Rather than using just wood or steel, Manzanita Village is built out of primarily reinforced concrete, while incorporating wood and steel as a minority material.
Concrete as a material has many advantages, such as low cost, durability, and the ability for the rooms to cool themselves at night (Bell and Rand).
Advantages such as those can help maximize occupancy — for example, a low cost means more buildings can be built, and the rooms being able to cool themselves at night removes the need for air conditioning infrastructure to be build, further maximizing space.
Additionally, the buildings of Manzanita Village are built with several buildings in the shape of a ‘U,’ allowing for three buildings to share a communal lawn, rather than all the buildings sharing a lawn.
"""
```
